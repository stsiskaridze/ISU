<h2 align="center">შეფასების კლასიკური მეთოდები</h2>

- სტატისტიკური დასკვნა შედგება იმ მეთოდებისგან, რომლებზეც დაყრდნობით ხდება დასკვნები განზოგადოებია პოპულაციის შესახებ.

<h3 align="center">წერტილოვანი შეფასება</h3>

- <h3>განმარტება 12.1:</h3>
   
  პოპულაციის $\theta$ პარამეტრის **წერტილოვანი შეფასება** ეწოდება $\hat{\Theta}$ სტატისტიკის ერთ კონკრეტულ $\hat{\theta}$ მნიშვნელობას. 



- მაგალითად, $\overline{X}$ სტატისტიკის $\overline{x}$ მნიშვნელობა, გამოთვლილი $n$ ზომის ნიმუშიდან, არის პოპულაციის $\mu$ პარამეტრის წერტილოვანი შეფასება. 

- რათქმაუნდა, მოსალოდნელია, რომ ვერ მოხერხდეს შეფასდეს პოპულაციის პარამეტრი შეცდომის გარეშე. 


- ჩვენ არ ველით, რომ $\overline{X}$ ზუსტად შეაფასებს $\mu$-ს, მაგრამ, რა თქმა უნდა, იმედი გვაქვს, რომ ის შორს არ არის შეჭმარიტი მნიშვნელობისგან. 


- კონკრეტული ნიმუშისთვის შესაძლებელია $\mu$-ის უფრო ახლო შეფასების მოპოვება შუამავლის $\widetilde{X}$ ნიმუშის შემფასებლის გამოყენებით. 


- მაგალითად, განვიხილოთ $2$, $5$ და $11$ მნიშვნელობებისგან შემდგარი ნიმუში იმ მოსახლეობიდან, რომლის საშუალო არის $4$, მაგრამ სავარაუდოდ უცნობია. 

  ჩვენ შევაფასებთ, რომ $\mu$ უნდა იყოს $\overline{x} = 6$, გამოვიყენოთ ნიმუშის საშუალო, ანუ $\widetilde{x}= 5$, გამოვიყენოთ მედიანური ნიმუში, როგორც ჩვენი შეფასება. 
  
  ამ შემთხვევაში, $\widetilde{X}$ შემფასებელი აწარმოებს შეფასებას, რომელიც უფრო ახლოს არის ნამდვილ მნიშვნელობასთან, ვიდრე $\overline{X}$ შემფასებელი. 
  
  მეორეს მხრივ, თუ ჩვენი შემთხვევითი ნიმუში შეიცავს მნიშვნელობებს $2$, $6$ და $7$, მაშინ $\overline{x} = 5$ და $\widetilde{x}= 6$, ამიტომ $\overline{X}$ არის უკეთესი შემფასებელი. 
  
- ჩვენ არ ვიცით $\mu$-ს ნამდვილი მნიშვნელობა, წინასწარ უნდა გადავწყვიტოთ, გამოვიყენოთ $\overline{X}$ ან $\widetilde{X}$, როგორც შემფასებელი.

- <h3>განმარტება 12.2:</h3>

  $\hat{\Theta}$ სტატისტიკას უწოდებენ $\theta$ პარამეტრის მიუკერძოებელი შემფასებელი, თუ:
  
  $$\mu_{\hat{\Theta}} = E(\hat{\Theta}) = \theta.$$

- <h3>მაგალითი 12.1:</h3>
  
  აჩვენეთ, რომ $S^2$ არის $\sigma^2$-ის პარამეტრის მიუკერძოებელი შემფასებელი. 

<h3 align="center">წერტილოვანი შეფასების ვარიაცია</h3>

- თუ $\hat{\Theta_1}$ და $\hat{\Theta_2}$ არის ერთი და იგივე პოპულაციური $θ$ პარამეტრის ორი მიუკერძოებელი შემფასებელი, ჩვენ გვინდა ავირჩიოთ შემფასებელი, რომლის სინჯების გაფანტვას უფრო მცირე სხვაობა აქვს.

- შესაბამიოსად,  თუ $\sigma^2_{\hat\theta_2} < \sigma^2_{\hat\theta_1}$, ჩვენ ვამბობთ, რომ $\hat{\Theta_1}$ უფრო ეფექტური შემფასებელია, ვიდრე $\hat{\Theta_2}$.

- <h3>განმარტება 12.2:</h3>

  თუ განვიხილავთ პოპულაციის $\theta$ პარამეტრის ყველა შესაძლო მიუკერძოებელ შემფასებელს, ყველაზე მცირე ვარიაციის მქონე შემფასებელს **ყველაზე ეფექტურ შეფასებას** უწოდებენ.

<h3 align="center">ინტერვალური შეფასება</h3>

- ყველაზე ეფექტური მიუკერძოებელი შემფასებელიც კი ნაკლებად სავარაუდოა, რომ ზუსტად შეაფასებს პოპულაციის პარამეტრს.

  მართალია, შეფასების სიზუსტე იზრდება დიდი ნიმუშებით, მაგრამ მაინც არ არსებობს მიზეზი, რომ მოცემული ნიმუშიდან წერტილის შეფასება ზუსტად ტოლი უნდა იყოს პოპულაციის პარამეტრისა, რომლის სავარაუდოდაც უნდა შეფასდეს. 
  
  
- არსებობს მრავალი სიტუაცია, როდესაც სასურველია განისაზღვროს ინტერვალი, რომლის დროსაც ველოდებით პარამეტრის მნიშვნელობას. 


- ასეთ ინტერვალს **ინტერვალის შეფასება** ეწოდება.

- <h3>განმარტება 12.3:</h3>

  $\theta$ პოპულაციის პარამეტრის ინტერვალის შეფასება არის $\hat{\theta_L} < \theta < \hat{\theta_U}$ ფორმის ინტერვალი, სადაც $\hat{\theta_L}$ და $\hat{\theta_U}$ დამოკიდებულია კონკრეტული ნიმუშის $\hat{\Theta}$ სტატისტიკის მნიშვნელობაზე და აგრეთვე $\hat{\Theta}$ ამ ნიმუშიშ ალბათურ განაწილებაზე.

- ვინაიდან სხვადასხვა ნიმუშები ზოგადად მოგვცემენ $\hat{\Theta}$-ს განსხვავებულ მნიშვნელობებს და, შესაბამისად, სხვადასხვა მნიშვნელობებს $\hat{\theta_L}$ და $\hat{\theta_U}$-სთვის, ინტერვალის ეს ბოლო წერტილები შესაბამისი **შემთხვევითი სიდიდეების** $\hat{\theta_L}$ და $\hat{\theta_U}$ მნიშვნელობებია. 


- $\hat{\Theta}$-ს ნიმუშის განაწილებიდან ჩვენ შეგვიძლია განვსაზღვროთ $\hat{\theta_L}$ და $\hat{\theta_U}$ ისე, რომ $P(\hat{\theta_L} < \theta < \hat{\theta_U})$ ტოლი იყოს ნებისმიერი დადებითი მნიშვნელობის, რომლის დაზუსტებაც ჩვენი ამოცანაა. 

- ჩვენ შეგვიძლია ეს ალბათობა ჩავწეროთ როგორც:

  $$P(\hat{\theta_L} < \theta < \hat{\theta_U}) = 1- \alpha,$$
  
  სადაც $0 < \alpha < 1$.
  
  მაშინ ჩვენ გვაქვს შემთხვევითი ნიმუშის არჩევის ალბათობა $1 − \alpha$, რომელიც წარმოქმნის $\theta$-ს ინტერვალს. 
  
  

- <h3>განმარტება 12.4:</h3>

  ინტერვალს $\hat{\theta_L} < \theta < \hat{\theta_U}$, გამოანგარიშებული შერჩეული ნიმუშიდან, ეწოდება $100(1 - \alpha)\%$ **ნდობის ინტერვალი**.
  
  $(1 - \alpha)$-ს ეწოდება **ნდობის კოეფიციენტი** ან **ნდობის ხარისხი**.
  
  $\hat{\theta_L}$ და $\hat{\theta_U}$ წერტილების ეწოდება **ქვედა და ზედა ნდობის საზღვრები**.

<h3 align="center">ცალკეული ნიმუში: საშუალო მნიშვნელობის შეფასება</h3>

- $\overline{X}$-ის ნიმუშის განაწილება ორიენტირებულია $\mu$-ზე და უმეტეს შემთხვევაში ვარიაცია უფრო მცირეა, ვიდრე $\mu$-ს ნებისმიერი სხვა შემფასებელი. 


- ამრიგად, საშუალო $\overline{x}$ ნიმუში გამოიყენბა როგორც პოპულაციის საშუალო $\mu$.

- ცენტრალური სასაზღვრო თეორემისგან გამომდინარე, ჩვენ შეგვიძლია $\overline{X}$-ის ნიმუში ჩავთვალოთ ნორმალურად განაწილებული საშუალოთი $\mu_{\overline{X}} = \mu$ და სტანდარტული გადახრით $\sigma_{\overline{X}} = \sigma/\sqrt{n}.$

  შესაბამისად, თუ ჩვენ შემოვიტანთ $Z$ სიდიდეს:
  
  $$Z = \frac{\overline{X} - \mu}{\sigma/\sqrt{n}},$$
  
  ჩვენ შეგვიძლია გამოვთვალოთ ნდობის ინტერვალი შემდეგნაირეად:
  
  $$P \left ( -Z_{\alpha/2} < Z < Z_{\alpha/2} \right ) = 1-\alpha,$$
  
  ან
  
  $$P \left (\overline{X} - Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} < Z < \overline{X} + Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \right) = 1-\alpha,$$

- <h3>თეორემა 12.1 ($\mu$ ნდობის ინტერვალი, თუ $\sigma^2$ ცნობილია):</h3>
  
  მოცემულია $n$ ზომის შემთხვევითი ნიმუში პოპულაციიდან, რომლისთვისაც ცნობილია $\sigma^2$ ვარიაცია.
  
  თუ $\overline{x}$ არის ამ ნიმუშის საშუალო, მაშინ $100 (1 - \alpha)\%$ ნდობის ინტერვალით გამოითვლება შემდეგი ფორმულით:
    
  $$\overline{x} - Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} < \mu < \overline{x} + Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} ,$$

  სადაც $z_{\alpha/2}$ არის $z$-მნიშვნელობა, რომელიც ტოვებს $\alpha/2$ ფართობს გრაფიკის მარჯვნივ.

- <h3>მაგალითი 12.2:</h3>
  
  თუთიის საშუალო კონცენტრაცია, რომელიც აღებულია მდინარეში 36 სხვადასხვა ადგილას ჩატარებული გაზომვების ნიმუშიდან, არის $2.6$ გრამი/მილილიტრში. 
  
  იპოვეთ $95\%$ და $99\%$ ნდობის ინტერვალი მდინარეში თუთიის საშუალო კონცენტრაციისთვის. 
  
  ჩათვალეთ, რომ მოსახლეობის სტანდარტული გადახრა არის $0,3$ გრამი/მილილიტრში.

- <h3>თეორემა 12.2:</h3>
  
  თუ $\overline{x}$ გამოყენებულია $\mu$-ს შეფასებისთვის, შეიძლება $100 (1 - \alpha)\%$ პროცენტით დარწმუნებული ვიყოთ, რომ შეცდომა არ აღემატება $z_{\alpha/2} \frac{\sigma}{\sqrt{n}}$.

- <h3>თეორემა 12.3:</h3>
  
  თუ $\overline{x}$ გამოყენებულია $\mu$-ს შეფასებისთვის, შეიძლება $100 (1 - \alpha)\%$ პროცენტით დარწმუნებული ვიყოთ, რომ რომ შეცდომა არ გადააჭარბებს მითითებულ რაოდენობას $e$, როდესაც ნიმუშის ზომა იქნება:
  
  $$n = \left ( \frac{z_{\alpha/2}\sigma}{e} \right )^2.$$

- <h3>მაგალითი 12.3:</h3>
  
  რა ზომის ნიმუშია საჭირო, თუ გვსურს $95\%$ პროცენტით დარწმუნებული ვიყოთ, რომ **12.2 მაგალითში** მოყვანილი $mu$-ს შეფასება $0.05$-ზე ნაკლებია?

<h3 align="center">ცალმხრივი ნდობის საზღვრები</h3>

- <h3>განმარტება 12.5:</h3>

  მოცემულია $n$ ზომის შემთხვევითი ნიმუში პოპულაციიდან, რომლისთვისაც ცნობილია $\sigma^2$ ვარიაცია.
  
  თუ $\overline{x}$ არის ამ ნიმუშის საშუალო, მაშინ ცალხმრივი $100(1-\alpha)\%$ ნდობის ინტერვალი მოცემულია შემდეგი ფორმულ:
  
  $$\overline{x} + Z_\alpha \sigma/\sqrt{n}, \text{ თუ ზედა ცალმხრივი საზღვრისთვის,}$$
  $$\overline{x} - Z_\alpha \sigma/\sqrt{n}, \text{ თუ ქვედა ცალმხრივი საზღვრისთვის.}$$
  


- <h3>მაგალითი 12.4:</h3>
  
  ფსიქოლოგიური ტესტირების ექსპერიმენტში $25$ სუბიექტი შეირჩევა შემთხვევითად და იზომება მათი რეაქციის დრო წამებში, კონკრეტულ სტიმულზე. 
  
  წარსული გამოცდილების თანახმად, ამ ტიპის სტიმულებზე რეაქციის დროების **ვარიაცია $4$ წამია** და რეაქციის დროის **განაწილება დაახლოებით ნორმალურია**. 
  
  სუბიექტების საშუალო დროა $6,2$ წამი. 
  
  გამოთვალეთ რეაქციის $95\%$ ნდობის ინტერვალი ზედა ცალმხრივი საზღვორსთივს.

<h3 align="center">შემთხვევა როდესაც $\sigma$ უცნობია</h3>

- ხშირად, ჩვენ გვექნება შემთხვევები, როდესაც საჭიროა შევაფასოთ მოსახლეობის საშუალო მნიშვნელობა, როდესაც ვარიაცია უცნობია. 


- თუ გვაქვს შემთხვევითი ნიმუში ჩვეულებრივი განაწილებიდან, მაშინ შემთხვევით სიდიდეს:

  $$T = \frac{\overline{X} - \mu}{S / \sqrt{n}}$$
  
  ექნება $t$-განაწილება (სტიუდენტის განაწილება) $v = n-1$ თავისუფლების ხარისხით.

- <h3>თეორემა 12.4 ($\mu$ ნდობის ინტერვალი, თუ $\sigma^2$ უცნობია) :</h3>
  
  მოცემულია $n$ ზომის შემთხვევითი ნიმუში პოპულაციიდან, რომლისთვისაც $\sigma^2$ ვარიაცია უცნობია.
  
  თუ $\overline{x}$ არის ამ ნიმუშის საშუალო, მაშინ $100 (1 - \alpha)\%$ ნდობის ინტერვალით გამოითვლება შემდეგი ფორმულით:
    
  $$\overline{x} - t_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} < \mu < \overline{x} + t_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} ,$$

  სადაც $t_{\alpha/2}$ არის $t$-განაწილების მნიშვნელობა $v = n-1$ თავისუფლების ხარისხით, რომელიც ტოვებს $\alpha/2$ ფართობს გრაფიკის მარჯვნივ.

- <h3>მაგალითი 12.5:</h3>
  
  გოგირდმჟავას $7$ მსგავსი კონტეინერის შემცველობაა $9.8, 10.2, 10.4, 9.8, 10.0, 10.2$ და $9.6$ ლიტრი. 
  
  იპოვნეთ $95\%$ **ნდობის ინტერვალი** ყველა ასეთი კონტეინერისთვის, თუ ჩავთვლით რომ ისინი არიან **ნორმალურად განაწილებული**.

- <h3>მაგალითი 12.6:</h3>
  
  გროვდება სქოლასტიკური შესაძლებლობების ტესტის (SAT) მათემატიკის ქულები ტეხასის შტატში $500$ საშუალო სკოლის მოხუცების შემთხვევითი ნიმუშიდან და ნაპოვნია ნიმუშის საშუალო და სტანდარტული გადახრა, შესაბამისად, $501$ და $112$. 
  
  იპოვეთ $99\%$ ნდობის ინტერვალი SAT მათემატიკის საშუალო ქულაზე ტეხასის შტატის უფროსებისათვის.

<h3 align="center">წერტილის შეფასების სტანდარტული შეცდომა</h3>

- ჩვენ საკმაოდ მკვეთრად განვასხვავეთ წერტილის შეფასების მიზანი და ნდობის ინტერვალის შეფასების მიზანი. 


- პირველი აწვდის ექსპერიმენტული მონაცემების ნაკრებიდან მოპოვებულ ერთ რიცხვს, ხოლო მეორე უზრუნველყოფს ინტერვალს, რომელიც გონივრული იქნება პარამეტრისთვის, ექსპერიმენტული მონაცემების გათვალისწინებით.


- ანუ ასეთი გამოთვლილი ინტერვალების $100 (1 - \alpha)\%$ "ფარავს" პარამეტრს.


- შეფასების ეს ორი მიდგომა დაკავშირებულია ერთმანეთთან. საერთო თემაა წერტილის შემფასებლის შერჩევის განაწილება. 
  
- მაგალითად, განვიხილოთ შემთხვევა როდესაც $\overline{X}$ შეფასებულია $\mu$-თი და $\sigma$ ცნობილია. 

  ადრე აღვნიშნეთ, რომ მიუკერძოებელი შემფასებლის ხარისხის საზომი არის მისი ვარიაცია. $\overline{X}$-ს ვარიაციაა:
  
  $$\sigma^2_{\overline{X}} = \frac{\sigma^2}{n}$$

- <h3>განმარტება 12.6:</h3>
  
  $\overline{X}$–ის **სტანდარტული გადახრა**, ან **სტანდარტული შეცდომა** არის $\sigma / \sqrt{n}$. 

  მარტივად რომ ვთქვათ, შემფასებლის სტანდარტული შეცდომა არის მისი სტანდარტული გადახრა.
  
  $\overline{X}$-სთვის გამოთვლილი ნდობის ინტერვალი:
  
  $$ \overline{x} \pm z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \text{ ჩაიწერება როგორც: } \overline{x} \pm z_{\alpha/2} \text{ ს.შ.}(\overline{x})$$
