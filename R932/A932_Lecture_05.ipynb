<h2 align="center">შემთხვევითი სიდიდეების წრფივი კომბინაციის მათემატიკური ლოდინი და ვარიაცია</h2>


- <h3>თეორემა 5.1:</h3>
  
  თუ $X$ არის შემთხვევითი სიდიდე (**დისრკეტული** თუ **უწყვეტი**), მაშინ ნებისმიერი $a$ და $b$ სკალარებისთვის სამართლებლივია ტოლობა:
  
  $$\mathbb{E} [aX + b] = a \mathbb{E} [X] + b.$$


- <h3>შედეგი 5.1:</h3>
  
  თუ $a = 0$, მაშინ $\mathbb{E} [b] = b$.


- <h3>შედეგი 5.2:</h3>
  
  თუ $b = 0$, მაშინ $\mathbb{E} [aX] = a \mathbb{E} [X]$.


- <h3>მაგალითი 5.1:</h3>
  
  იპოვნეთ მათემატიკური ლოდინი **4.4 მაგალითში** აღწერილი $f(X) = 2X - 1$ **დისკრეტული** შემთხვევითი სიდიდესთვის.

- <h3>მაგალითი 5.2:</h3>
  
  იპოვნეთ მათემატიკური ლოდინი **4.45მაგალითში** აღწერილი $g(X) = 4X + 3$ **უწყვეტი** შემთხვევითი სიდიდესთვის..


- <h3>თეორემა 5.2:</h3>
  
  თუ $f(X)$ და $g(X)$ არის ორი შემთხვევითი ცვლადის ფუნქცია, მაშინ სამართლებლივია ტოლობა:
  
  $$\mathbb{E} [f(X) + g(X)] = \mathbb{E} [f(X)]  + \mathbb{E} [g(X)].$$


- <h3>მაგალითი 5.3:</h3>
  
   დავუშვათ, რომ $X$ არის შემთხვევითი სიდიდე შემდეგი ალბათური განაწილებით:
   
|   $x$  |       0       |       1       | 2 |       3       |
|:------:|:-------------:|:-------------:|:-:|:-------------:|
| $f(x)$ | $\frac{1}{3}$ | $\frac{1}{2}$ | 0 | $\frac{1}{6}$ |


&emsp;&emsp; იპოვნეთ $Y = (X-1)^2$ შემთხევითი სიდიდის **მათემატიკური ლოდინი**.
  

- <h3>მაგალითი 5.4:</h3>
  
  ყოველკვირეული მოთხოვნა გარკვეულ სასმელზე, ათასობით ლიტრში, წარმოდგენილია უწყვეტი შემთხვევითი სიდიდით $g (X) = X^2 + X − 2$, სადაც $X$ გააჩნია შემდეგი სიმკვრივის ფუნქცია:
  
  $$f(x)=
  \left\{\begin{matrix}
  2(x-1), & 1 < x < 2, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$
  
  იპოვნეთ სასმელზე ყოველკვირეული **მოსალოდნელი** მოთხოვნის მნიშვნელობა.

- <h3>თეორემა 5.3:</h3>
  
  თუ $f(X,Y)$ და $g(X, Y)$ არის ორი ორცვლადიანი შემთხვევითი სიდიდის ფუნქცია, მაშინ სამართლებლივია ტოლობა:
  
  $$\mathbb{E} [f(X,Y) + g(X,Y)] = \mathbb{E} [f(X, Y)]  + \mathbb{E} [g(X, Y)].$$


- <h3>შედეგი 5.3:</h3>
  
  თუ $f(X,Y) = f(X)$ და $g(X,Y) = g(Y)$, მაშინ:
  
  $$\mathbb{E}[f(X) + g(X)] = \mathbb{E}[f(X)]  + \mathbb{E}[g(X)].$$


- <h3>შედეგი 5.4:</h3>
  
  თუ $f(X,Y) = X$ და $g(X,Y) = Y$, მაშინ:
  
  $$\mathbb{E}[X + Y] = \mathbb{E}[X]  + \mathbb{E}[Y].$$


- <h3>თეორემა 5.4:</h3>
  
  თუ $X$ და $Y$ არის ორი **დამოუკიდებელი** შემთხვევითი სიდიდეები, მაშინ სამართლებლივია ტოლობა:
  
  $$\mathbb{E} [XY] = \mathbb{E} [X] \cdot  \mathbb{E} [Y].$$


- <h3>შედეგი 5.5:</h3>
  
  თუ $X$ და $Y$ არის ორი **დამოუკიდებელი** შემთხვევითი სიდიდეები, მაშინ $\sigma_{XY} = 0$.


- <h3>მაგალითი 5.5:</h3>
  
  ცნობილია, რომ გალიუმისა და არსენიდის თანაფარდობა გავლენას არ ახდენს გალიუმ-არსენიდის ვაფლების ფუნქციონირებაზე, რომლებიც მიკროჩიპების ძირითადი კომპონენტებია.
 
  თუ $X$-ით ავღნიშნავთ გალიუმის და არსენიდის თანაფარდობას, ხოლო $Y$ ავღნიშნავთ ფუნქციურ ვაფრებს, რომლებიც მიიღება 1 საათის განმავლობაში, მაშინ $$ და $$ შემთხვევითი სიდიდეების ერთობლივი განაწილების ფუნქციას ექნება შემდეგი სახე:
  
  $$f(x, y)=
  \left\{\begin{matrix}
  \frac{x(1+3y^2)}{4}&  0 < x < 2,  0 < y < 1\\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$
  
  **თეორემა 5.4 გამოყენებით**, აჩვენეთ, რომ   $\mathbb{E} [XY] = \mathbb{E} [X] \cdot  \mathbb{E} [Y]$.

- <h3>თეორემა 5.5:</h3>
  
  თუ $X$ და $Y$ არის ორი შემთხვევითი სიდიდეები ერთობლივი განაწილების ფუნქციით $f(X, Y)$, მაშინ ნებისმიერი $a$, $b$, და $c$ სკალარებისთვის სამართლებლივია ტოლობა:
  
  $$\sigma_{aX + bY + c}^2 = a^2 \sigma_{X}^2 + b^2 \sigma_{Y}^2 + 2ab \cdot \sigma_{XY}.$$


- <h3>შედეგი 5.6:</h3>
  
  თუ $b=0$, მაშინ:
  
  $$\sigma_{aX+c}^2 = a^2\sigma_{X}^2 = a^2 \sigma^2.$$


- <h3>შედეგი 5.7:</h3>
  
  თუ $b=0$ და $c = 0$, მაშინ:
  
  $$\sigma_{X+c}^2 = \sigma_{X}^2 = \sigma^2.$$


- <h3>შედეგი 5.8:</h3>
  
  თუ $b=0$ და $c = 0$, მაშინ:
  
  $$\sigma_{aX}^2 = a^2 \sigma_{X}^2 = a^2 \sigma^2.$$


- <h3>შედეგი 5.9:</h3>
  
  თუ $X$ და $Y$ არის ორი დამოუკიდებელი შემთხვევითი სიდიდეები, მაშინ:
  
  $$\sigma_{aX + bY}^2 = a^2 \sigma_{X}^2 + b^2 \sigma_{Y}^2.$$


- <h3>შედეგი 5.10:</h3>
  
  თუ $X$ და $Y$ არის ორი დამოუკიდებელი შემთხვევითი სიდიდეები, მაშინ:
  
  $$\sigma_{aX - bY}^2 = a^2 \sigma_{X}^2 + b^2 \sigma_{Y}^2.$$


- <h3>შედეგი 5.11:</h3>
  
  თუ $X_1, X_2, ..., X_n$ არის დამოუკიდებელი შემთხვევითი სიდიდეები, მაშინ:
  
  $$\sigma_{a_1 X_1 + a_2 X_2 + \cdots + a_n X_n}^2 = a_1^2 \sigma_{X_1}^2 + a_2^2 \sigma_{X_2}^2 + \cdots + a_n^2 \sigma{X_n}^2.$$


- <h3>მაგალითი 5.6:</h3>
  
  მოცემულია $X$ და $Y$ შემთხვევითი სიდიდეები შემდეგი ვარიაციის და კოვარიაციის მნიშვნელობებით:
  
  $$\sigma_{X}^2 = 2,$$
  $$\sigma_{Y}^2 = 4,$$
  $$\sigma_{XY} = -2.$$
  
  იპოვნეთ $Z = 3X - 4Y + 8$ შემთხვევითი სიდიდის **ვარიაცია**.

- <h3>მაგალითი 5.7:</h3>
  
  მოცემულია $X$ და $Y$ **დამოუკიდებელი** შემთხვევითი სიდიდეები შემდეგი ვარიაციის მნიშვნელობებით:
  
  $$\sigma_{X}^2 = 2,$$
  $$\sigma_{Y}^2 = 3.$$
  
  იპოვნეთ $Z = 3X - 2Y + 5$ შემთხვევითი სიდიდის **ვარიაცია**.

<h2 align="center">რა მოხდება, თუ ფუნქცია არაწრფივია?</h2>


- <h3>მაგალითი 5.8:</h3>
  
  მოცემულია $X$ შემთხვევითი სიდიდე $\mu_X$ მათემატიკური ლოდინით და $\sigma_{X}^2$ ვარიაციით.
  
  იპოვნეთ $\mathbb{E} \left [  e^X \right]$ მეორე რიგის სიზუსტით.


- <h3>მაგალითი 5.9:</h3>
  
  **5.8 მაგალითისთვის** იპოვნეთ $\mathbb{Var} \left [  e^X \right]$ მიახლოებითი მნიშვნელობა.


- <h3>მაგალითი 5.10:</h3>
  
  მოცემულია $X$ და $Y$ დამოუკიდებელი შემთხვევითი სიდიდეეები $\mu_X$ და $\mu_Y$ მათემატიკური ლოდინით და $\sigma_{X}^2$ და $\sigma_{Y}^2$ ვარიაციით.
  
  იპოვნეთ $\mathbb{E} \left [ \frac{X}{Y} \right ]$ და $\mathbb{Var} \left [ \frac{X}{Y} \right ]$  მიახლოებითი მნიშნველობები.

- <h3>თეორემა 5.6 (ჩებიშევის თეორემა):</h3>
  
  ალბათობა იმისა, რომ ნებისმიერი შემთხვევითი სიდიდიე $X$ მიიღებს მნიშვნელობას $k$ სტანდარტული გადახრაში, არის მინიმუმ $1 - \frac{1}{k^2}$:
  
  $$P \left ( \mu - k\sigma < X < \mu + k\sigma \right ) \geq 1 - \frac{1}{k^2}.$$


- <h3>მაგალითი 5.11:</h3>
  
  მოცემულია $X$ შემთხვევითი სიდიდე $\mu = 8$ მათემატიკური ლოდინით და $\sigma^2 = 9$ ვარიაციით. 
  
  $X$ შემთხვევითი სიდიდის განაწილების ფუნქცია უცნობია.
  
  იპოვნეთ:
  1. $P(-4 < X < 20)$.
  2. $P(|X - 8| \geq 6)$.
