- ზოგადად, ჩვენ გვაინტერესებს $x$ წარმატების ალბათობა $k$ ელემენტის არჩევით და $n - x$ წარუმატებლობების ალბათობა $N-k$ ელემენტის არჩევით, როდესაც შემთხვევითად ირჩევა $n$ ელემენტი $N$ ელემენტიან ნიმუშიდან.


- ესეთი ამოცანა აღიწერება **ჰიპერგეომეტრიული** განმაწილებით, რომელიც ფლობს შემდეგ ორ თვისებას:

  1. $n$ ზომის შემთხვევითი ნიმუში შეირჩევა $N$ ელემენტიდან **ჩანაცვლების გარეშე**. 
  
  2. $N$ ელემენტიდან $k$ შეიძლება კლასიფიცირდეს როგორც **წარმატება** და $N - k$ კლასიფიცირდება როგორც **წარუმატებლობა**.


- ჰიპერგეომეტრიული ექსპერიმენტის **წარმატების რიცხვს** $X$ ეწოდება **ჰიპერგეომეტრიულ შემთხვევითი სიდიდე**.


- ჰიპერგეომეტრიული შემთხვევითი სიდიდის ალბათურ განაწილებას **ჰიპერგეომეტრიულ განაწილებას ეწოდება** და აღნიშნავენ როგორც: 

  $$h (x; N, n, k)$$.

- <h3>განმარტება 7.1:</h3>
 
  ჰიპერგეომეტრიული შემთხვევითი სიდიდის **ალბათური განაწილების ფუნქცია** წარმოდგინდება შემდეგი სახით:

  $$h(x; N, n, k) = \frac{\binom{k}{x} \binom{N - k}{n-x}}{\binom{N}{n}}, \text{ } \max \{0, n-(N-k) \} \leq x \leq \min \{n, k\},$$

  სადაც $n$ არის $N$ ელემენტიდან შერჩეუი შემთხვევით ნიმუშში, რომელიდანაც $k$ წარმატებაა და $N - k$ წარუმატებლობა.

- <h3>თეორემა 7.1:</h3>
  
  ჰიპერგეომეტრიული განაწილების $h(x; N, n, k)$ **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულებით:
  
  $$\mu = \frac{nk}{N},$$
  
  $$\sigma^2 = \frac{N-n}{N-1} \cdot n \cdot \frac{k}{N} \left ( 1 - \frac{k}{N}\right ).$$

<h2 align="center">უარყოფითი ბინომიალური და გეომეტრიული განაწილებები</h2>


- თუ ექსპერიმენტის ჩატარების დროს $k$ **წარმატების მისაღწევად** საჭირო $X$ **ცდის ჩატარება**, მაშინ $X$ ეწოდება **უარყოფითი ბინომიალური შემთხვევითი სიდიდე**, ხოლო მისი ალბათურ განაწილებას **უარყოფითი ბინომიალური განაწილება**.

- <h3>განმარტება 7.2:</h3>

  თუ განმეორებითმა დამოუკიდებელმა ცდებმა შეიძლება გამოიწვიოს წარმატება $p$ ალბათობით და წარუმატებლობა ალბათობით $q = 1 - p$ ალბათობით, მაშინ $X$ შემთხვევითი სიდიდის ალბათობის განაწილება, რომელზეც ხდება **$k$-ური წარმატება**, აღიწერება შემდეგი ფორმულით:
  
  $$b^*(x; k, p) = \binom{x-1}{k-1} p^k q^{x-k}, \text{ სადაც } x = k, k+1, k+2, ... .$$

- <h3>თეორემა 7.2:</h3>
  
  გეომეტრიული განაწილების $g(x; p) = p q^{x-1}$ **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულებით:
  
  $$\mu = \frac{1}{p},$$
  
  $$\sigma^2 = \frac{1-p}{p^2}.$$

- <h3>თეორემა 7.3:</h3>
  
  პუასონის განაწილების $p(x; \lambda t) = \frac{e^{-\lambda t} (\lambda t)^x}{x!}$ **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულებით:
  
  $$\mu = \sigma^2 = \lambda t.$$

- <h3>თეორემა 7.4:</h3>
  
  დავუშვათ $X$ არის ბინომიალური შემთხვევითი სიდიდე ალბათური განაწილებით $b(x; n, p)$. 
  
  როდესაც $n \rightarrow \infty$, $p \rightarrow 0$ და $\lim_{n\rightarrow \infty} np = \mu$ რჩება მუდმივი, ბინომიალური განაწილება მიისწრაფის პუასონის განაწილებისკენ:
  
  $$\lim_{n\rightarrow \infty} b(x; n, p) = p(x; \mu).$$
