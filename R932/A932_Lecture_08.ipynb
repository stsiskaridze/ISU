<h2 align="center">უწყვეტი თანაბარი განაწილება</h2>

- სტატისტიკის ერთ-ერთი ყველაზე უმარტივესი უწყვეტი განაწილება არის **უწყვეტი თანაბარი განაწილება**. 


- ამ განაწილებას ახასიათებს **სიმკვრივის ფუნქცია**, რომელიც არის **თანაბარი** რაღაც დახურულ $[A,B]$ ინტერვალზე.


- მიუხედავად იმისა, რომ უწყვეტი ერთგვაროვანი განაწილების **გამოყენება არც ისე ბევრია**, მიზანშეწონილია ამ განაწილების დავიწყოთ.

- <h3>განმარტება 8.1:</h3>
 
  უწყვეტი თანაბარი შემთხვევითი სიდიდის სიმკვრივის ფუნქცია განიმარტება დახურულ $[A, B]$ ინტერვალზე შემდეგნაირად:
  
  $$f(xl A, B) = 
  \left\{\begin{matrix}
  \frac{1}{B-A}, &\text{ თუ } A \leq x \leq B,\\
  0,& \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$

- <h3>მაგალითი 8.1:</h3>
  
  დავუშვათ, რომ გარკვეული კომპანიის დიდი საკონფერენციო დარბაზი შეიძლება დაჯავშნდეს **არა უმეტეს 4 საათის** განმავლობაში.
  
  **ხანგრძლივი** და **მოკლე კონფერენციები** საკმაოდ ხშირად ტარდება. 
  
  ჩვენ შეიძლება ვივარაუდოთ, რომ **კონფერენციის ხანგრძლივობა** $X$ აღიწერება **უწყვეტი თანაბარი განაწილებით** $[0, 4]$ ინტერვალზე.
  
  ა) რისი ტოლია $X$ შემთხვევითი სიდიდის **სიმკვრივის ფუნქცია**?

  ბ) რა არის ალბათობა, რომ კონფერენცია გაგრძელდეს **მინიმუმ 3 საათი**?

- <h3>თეორემა 8.1:</h3>
  
  **თანაბრად განაწილებული** უწყვეთი შემთხვევითი სიდიდის **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულით:
  
  $$\mu = \frac{A+B}{2} \text{  და  } \sigma^2 = \frac{(B-A)^2}{12}.$$

<h2 align="center">ნორმალური განაწილება</h2>

- ალბათ, სტატისტიკის **ყველაზე მნიშვნელოვანი** უწყვეტი ალბათობის განაწილება არის **ნორმალური განაწილება**.


- ამ განაწილებით აღიწერება ბუნებაში, ინდუსტრიაში და კვლევაში არსებული მრავალი ფენომენი.


- **ნორმალურ განაწილებას** ხშირად **გაუსის განაწილებას** უწოდებენ, **კარლ ფრიდრიხ გაუსის** (1777–1855) საპატივცემულოდ.

- <h3>განმარტება 8.2:</h3>
 
  **ნორმალურად განაწილებული** შემთხვევითი სიდიდის **სიმკვრივის ფუნქცია** შემდეგი ფორმულით:
  
  $$n(x; \mu, \sigma) = \frac{1}{\sqrt{2 \pi} \sigma} e^{-\frac{1}{2\sigma^2}(x-\mu)^2}, \text{ }  -\infty < x < \infty,$$
  
  სადაც $\mu$ და $\sigma^2$ არის ამ ფუნქციის **პარამეტრები** და აღიშნავენ შესამაბისად **მათემატიკურ ლოდინს** და **ვარიაციას**.
  
  <br>

- $\mu$ და $\sigma$ დაზუსტებისთანავე, მთლიანად განისაზღვრება **ნორმალური განაწილების მრუდი**. 


- მაგალითად, თუ $\mu = 50$ და $\sigma = 5$, მაშინ $n(x; 50, 5)$ შეიძლება გამოითვალოს $x$ სხვადასხვა მნიშვნელობებისა და აიგოს გრაფიკი.

- <h3>თვისებები:</h3>
  
  $n(x; \mu, \sigma)$ ფუნქციის პირველი და მეორე წარმოებულების შესწავლით, ჩვენ შეგვიძლია ნორმალური განაწილების  შემდეგი თვისებები ჩამოვთვალოთ:

  1. მრუდის **მაქსიმალური მნიშნველობა** მიიღწევა როდესაც $x = \mu$.
   
  2. მრუდი **სიმეტრიულია** $x = \mu$ წერტილზე გამავალი ვერტიკალური ღერძის მიმართ.
  
  3. მრუდს აქვს **მოქცევის წერტილები** $x = \mu \pm \sigma$, ანუ მრუდი ჩაზნექილია $[\mu  - \sigma, \mu  + \sigma]$ ინტერვალზე და ამოზნექილი ამ ინტერვალის გარეთ.
  
  4. მრუდი **ასიმპტოტიკურად უახლოვდება ჰორიზონტალურ ღერძს**.
  
  5. მრუდის ქვეშ და ჰორიზონტალური ღერძის შორის **მთლიანი ფართობი 1-ის ტოლია**.

- <h3>თეორემა 8.2:</h3>
  
  **ნორმალურად განაწილებული** უწყვეთი შემთხვევითი სიდიდის **მათემატიკური ლოდინი** და **ვარიაცია** შესაბამისად $\mu$ და $\sigma^2$ ტოლია.

<h2 align="center">ნორმალური მრუდის ქვეშ ფართობის გამოთვლა</h2>

- **ნებისმიერი უწყვეტი** ალბათური განაწილების ან სიმკვრივის ფუნქციის მრუდი აგებულია ისე, რომ **მრუდის ქვეშ არსებული ფართობი**, შემოსაზღვრული $x = x_1$ და $x = x_2$ წერტილებით, **უდრის ალბათობას**, რომ შემთხვევითი ცვლადი $X$ იღებს მნიშვნელობას $x = x_1$ და $x = x_2$ შორის.


- შესაბამისად, ნორმალური **მრუდისთვის ქვეშ არსებული ფართბი** გამოითვლება შემდეგი ფორმულით:

  $$P(x_1 < X < x_2) = \int_{x_1}^{x_2} n(x; \mu, \sigma) dx = \frac{1}{\sqrt{2 \pi} \sigma} \int_{x_1}^{x_2} e^{-\frac{1}{2\sigma^2} (x - \mu) ^2}dx$$
