<h2 align="center">რა განსხვავებაა ალბათობასა და სტატისტიკას შრის?</h2>

- **ალბათობა** და **სტატისტიკა** ორი ერთმანეთთან მჭიდროდ დაკავშირებული საგანია.


- ძირითადი განსხვავება ალბათობასა და სტატისტიკას შორის არის:
  - **ალბათობა** არის რიცხვითი აღწერა იმისა, თუ რამდენად სავარაუდოა მოვლენა.
  - **სტატისტიკა** არის დისციპლინა, რომელიც ეხება მონაცემთა შეგროვებას, ორგანიზებას, ანალიზს, ინტერპრეტაციას და პრეზენტაციას.
  

- ჩვენ დავიწყებთ ალბათობის საფუძვლების გახსენებით და მერე გადავალთ სტატისტიკის შესწავლაზე.

<h2 align="center">1.1 ელემენტარულ ხდომილებათა სივრცე</h2>

- **განმარტება 1.1**: სტატისტიკური ექსპერიმენტის ყველა შესაძლო შედეგების ერთობლიობას ეწოდება **ხდომილობათა სივრცე** და წარმოდგენილია $S$ სიმბოლოით.

  ხდომილებათა სივრცის თითოეულ **შედეგს** ეწოდება ხდომილებათა სივრცის **ელემენტები**, ან **წევრები**, ან **სივრცის წერტილები**.

  თუ ნხდომილებათა სივრცეს აქვს ელემენტების **სასრული რაოდენობა**, მაშინ ჩვენ შეგვიძლია ჩამოვწეროთ მისი წევრები **ფრჩხილებში** და გამოვყოთ **მძიმეებით**.

  მაგალითად, მონეტის ერთჯერადი აგდების შედეგების ხდომილებათა სივრცე შეგვიძლია ჩავწეროთ როგორც:
  
  $$S = \{H, T\},$$
  
  სადაც $H$ (**Head**) აღნიშნავს **საფასურს** ხოლო $T$ (**Tail**) აღნიშნავს **გერბს**.

- **მაგალითი 1.1**: 
  
  განვიხილოთ **კამათელის გაგორების** ექსპერიმენტი. 
  
  ა) თუ ჩვენ გვაინტერესებს კამათელის გაგორებით მიღებული რიცხვი (კამათელის ზედა მხარე), მაშინ **ხდომილებათა სივრცე** იქნება:

  $$S_1 = \{1, 2, 3, 4, 5, 6\}.$$
  
  ბ) თუ ჩვენ დაინტერესებული ვართ მხოლოდ იმით, რომ მიღებული რიცხვი ლუწია თუ კენტია, მაშინ **ხდომილებათა სივრცე** იქნება:
  
    $$S_2 = \{კენტი, ლუწი\} = \{ \{1, 2, 3\}, \{2, 4, 6\}\}.$$ 

- **მაგალითი 1.2**: 
  
  ექსპერიმენტი შედგება **მონეტის აგდებისგან**. თუ მონეტის აგდების შედეგი **საფასურია**, მაშინ **მონეტას ხელმეორედ აგდებენ**.
  ხოლო, თუ მონეტის აგდების შედეგი **გერბია**, მაშინ **აგორებენ კამათელს**.
  
  ექსპერიმენტის დიაგრამა ქვემოთ არის მოყვანილი:

  <center><img src="images/L1_Ex_1_2.jpg" width="400" alt="Example" /></center>

  როგორც დიაგრამიდან ჩანს, ამ ექსპერიმენტის **ხდომილებათა სივრცე** იქნება:
  
  $$S = \{HH,  TH, T1, T2, T3, T4, T5, T6\}.$$ 

- **მაგალითი 1.3**: 
  
  ექსპერიმენტი შედგება წარმოების შედეგად მიღებული სამი დეტალის შერჩევით. 
  
  თითოეული დეტალი მოწმდება და კლასიფიცირდება როგორც დეფექტიანი ($D$) ან არადეფექტიანი ($N$). 
  
  ხდომილებათა სივრცის ელემენტების ჩამოსათვლელად ჩვენ ვაგებთ შესაბამის ხის დიაგრამას. 

  <center><img src="images/L1_Ex_1_3.jpg" width="400" alt="Example" /></center>

  ახლა, ხის ტოტების სხვადასხვა ბილიკი იძლევა ხდომილებათა სიცრცის მკაფიო წერტილებს. 
  
  პირველი გზიდან დაწყებული, მივიღებთ $DDD$ წერტილის ნიმუშს, რაც მიუთითებს იმის შესაძლებლობაზე, რომ შემოწმებული სამივე დეტალი არის დეფექტიანი.
  
  თუ ჩვენ გავივლით ყველა სხვა ბილიკების გასწვრივ, ჩვენ მივიღებთ, რომ **ხდომილებათა სივრცე** წარმოადგენს:
  
  $$S = \{DDD, DDN, DND, DNN, NDD, NDN, NND, NNN\}.$$ 

<h2 align="center">1.2 მოვლენები</h2>

- ნებისმიერი მოცემული ექსპერიმენტისთვის, ჩვენთვის შეიძლება საინტერესო იყოს გარკვეული **მოვლენები** და არა ხდომილებათა სივრცის კონკრეტული ელემენტები.


- მაგალითად, ჩვენ შეიძლება გვაინტერესებდეს **მოვლენა** $A$, როდესაც კამათელის გაგორების შედეგად მიღებული რიცხვი არის 3-ის ჯერადი.

  ეს **მოვლენა** მოხდება, თუ კამათელის გაგორების შედეგი არის (1.1) მაგალითში მოყვანილი ხდომილებათა სივრცის **ქვესიმრალვე** $ A = \{3, 6 \}$.
  
  შესაბამისად, **მოვლენის განმარტება** შემოგვაქვს შემდეგნაირად:
  
  
- **განმარტება 1.2**: ხდომილებათა სივრცის ქვესიმრალვე წარმოადგენს **მოვლენას**.



- **მაგალითი 1.4**: 
  
  მოცემულია ხდომილებათა სივრცე $S = \{t | 0 \leq t\}$, სადაც $t$ არის გარკვეული ელექტრონული კომპონენტის ცხოვრების პერიოდი.
  
  მაშინ **მოვლენა**, როდესაც კომპონენტი ხუთი წლის განმავლობაში გამოდის წყობიდან, არის $S$ სივრცის ქვესიმრალვე $A = \{t | 0 \leq t < 5\}$.


- მოვლენა შეიძლება იყოს ქვესიმრავლე, რომელიც მოიცავს მთელ ხდომილებათა სივრცეს $S$ ან ქვესიმრავლე, რომელსაც საერთოდ არ მოიცავს აცერთ ელემენტს, ანუ ცარიელი ქვესიმრავლე, და რომლეიც აღინიშნება $\{\varnothing \}$.


- მაგალითად, თუ $A$ არის ბიოლოგიურ ექსპერიმენტში შეუიარაღებელი თვალით მიკროსკოპული ორგანიზმის აღმოჩენის მოვლენა, მაშინ $A = \{ \varnothing \}$.


- ასევე, თუ $B = \{ x | x \text { არის 7-ის ლუწი გამყოფი}\}$, მაშინ $B$ უნდა იყოს ნულოვანი სიმრავლე, რადგან $7$–ის ერთადერთი შესაძლო გამყოფიბია კენტი რიცხვები $1$ და $7$.


- განვიხილოთ ექსპერიმენტი, სადაც აღირიცხება მწარმოებელი ფირმის თანამშრომლების **მოწევის ჩვევები**. 

  ხდომილებათა სივრცე შეიძლება კლასიფიცირდეს როგორც **არამწეველი**, **მსუბუქი მწეველი**, **ზომიერი მწეველი** ან **მძიმე მწეველი** ინდივიდი. 
  
  ავღნიშნოთ **მწეველთა ქვესიმრავლე** როგორც $A$ **მოველნა**, მაშინ ყველა **არამწეველი** მოხვდევბა $S$ სივრცის სხვა მოვლენაში, რომელსაც $A$ მოვლენის **დამატება** ჰქვია.

- **განმარტება 1.3**: $A$ მოვლენის **დამატება** $S$–სთან მიმართებაში არის $S$– ის ყველა ელემენტის ქვესიმრავლე, რომლებიც არ იმყოფება $A$ ქვესიმრავლეში. ჩვენ $A$–ს დანამატს აღვნიშნავთ სიმბოლო $\overline{A}$–ით.

- **მაგალითი 1.5**: 
  
  დავუშვათ ხდომილებათა სივრცე $S$ არის ჩვეულებრივი **52 ბანქოს ბარათების გემბნიდან** პირველი ამოსული ბარათი.
  
  მოდით $A$-ით ავღნიშნოთ **მოვლენა**, როდესაც პირველი ამოსული ბარათი არის **წითელი ფერის ბარათი** . 
  
  მაშინ $\overline{A}$ არის ის მოვნლენა, როდესაც გემბანიდან არჩეული ბარათი არ არის წითელი ფერის ბარათი, ანუ არის **შავი ფერის ბარათი**.

- **მაგალითი 1.6**: 
  
  განვიხილოთ შემდეგი ხდომილებათა სივცრე:
  
  $$S = \{წიგნი, რვეული, კალამი, ფანქარი, საშლელი\}.$$
  
  თუ მოვლენა $A = \{წიგნი, ფანქარი, კალამი \}$, მაშინ მოვლენა $\overline{A} = \{რვეული, საშლელი\}$
 

- **განმარტება 1.4**: ორი $A$ და $B$ მოვლენის **თანაკვეთა**, რომლებიც აღინიშნება სიმბოლოთი $A \cap B$, არის მოვლენა, რომელიც შეიცავს ყველა ელემენტს, რომლებიც საერთოა $A$ და $B$ ქვესიმრავლეებისთვის.

- **მაგალითი 1.7**: 
  
  დავუშვათ $A$ არის **მოვლენა** როდესაც კლასში შემთხვევით შერჩეული ადამიანი სპეციალობით ინჟინერია:

  $$A = \{ \text{კლასში შემთხვევით შერჩეული ადამიანი სპეციალობით ინჟინერია}\}.$$
  
  ხოლო $B$ არის **მოვლენა**, როდესაც კლასში შემთხვევით შერჩეული ადამიანი ქალია:
  
  $$B = \{ \text{კლასში შემთხვევით შერჩეული ადამიანი ქალია }\}.$$
  
  მაშინ, $A \cap B$ არის მოვლენა, როდესაც კლასში შემთხვევით შერჩეული ადამიანი არის ქალი ინჟინერი:
  
  $$A \cap B = \{ \text{კლასში შემთხვევით შერჩეული ადამიანი არის ქალი ინჟინერი}\}.$$
 

- **მაგალითი 1.8**: 
  
  დავუშვათ $U = \{ა, ე, ი, ო, უ\}$ და $V = \{ჭ, ღ, რ\}$, მაშინ $U \cap V = \{\varnothing\}$.

  ანუ ეს ორი მოვლენა ერთდროულად ვერ მოხდება, ანუ ისინი **ურთიერთგამომრიცხავი** მოვლენებია.

- **განმარტება 1.5**: ორი $A$ და $B$ მოვლენა **ურთიერთგამომრიცხავია** ან **ერთმანეთისაგან გამიჯნული**, თუ $A \cap B = \{\varnothing\}$, ანუ თუ $A$ და $B$ არ აქვთ საერთო ელემენტები.

- **მაგალითი 1.9**: 
  
  საკაბელო ტელევიზიის კომპანია სთავაზობს პროგრამებს რვა სხვადასხვა სატელევიზიო არხზე, რომელთაგან სამი დაკავშირებულია $ABC$–სთან, ორი $NBC$–სთან და ერთი $CBS$–სთან, ხოლო დანარჩენი ორი არის საგანმანათლებლო არხი და $ESPN$ სპორტული არხი. 
  
  დავუშვათ, რომ ამ მომსახურების აბონენტი რთავს სატელევიზიო არხს არჩევის გარეშე.
  
  მოდით $A$ იყოს მოვლენა, რომ პროგრამა ეკუთვნის $NBC$ ქსელს და $B$, თუ ის არის $CBS$ ქსელი. 
  
  ვინაიდან სატელევიზიო პროგრამა არ შეიძლება ერთზე მეტ ქსელს მიეკუთვნებოდეს, $A$ და $B$ მოვლენებს საერთო პროგრამები არ აქვთ. 
  
  ამიტომ, $A \cap B$ გადაკვეთა არ შეიცავს პროგრამებს ჰქონდეს და, შესაბამისად, $A$ და $B$ მოვლენები ურთიერთგამომრიცხავია.

- **განმარტება 1.6**: ორი მოვლენის $A$ და $B$ **გაერთიანება**, რომელიც აღინიშნება სიმბოლოთი $A\cup B$, არის მოვლენა, რომელიც შეიცავს ყველა იმ ელემენტს, რომლებიც ეკუთვნის ან $A$-ს ან $B$-ს, ან ორივეს.

- **მაგალითი 1.10**: 
  
  დავუშვათ $A = \{a, b, c\}$ და $B = \{b, c, d\}$, მაშინ $A \cup B = \{a, b, c, d\}$.

- **მაგალითი 1.11**: 
  
  დავუშვათ $P$ არის **მოვლენა**, როდესაც ნავთობის საბურღი კომპანიიდან შემთხვევით შერჩეული **თანამშრომელი სიგარეტს ეწევა**.
  
  მოდით $Q$ იყოს **მოვლენა**, როდესაც **თანამშრომელი** ირჩევს ალკოჰოლურ სასმელებს.
  
  მოვლენა $P \cup Q$ არის ყველა ის თანამშრომელი, რომლებიც ან სვამენ ან ეწევიან ან ორივეს ერთად აკეთებენ.

- **მაგალითი 1.12**: 
  
  დავუშვათ $M = \{x | 3 < x < 9\}$ და $N = \{y | 5 < y < 12\}$, მაშინ :
  
  $$M \cup N = \{z | 3 < z < 12\}.$$

- დამოკიდებულება **მოვლენებსა** და შესაბამის **ხდომილებათა სივრცეს** შორის შეიძლება ილუსტრირებული იყოს **ვენის დიაგრამების** საშუალებით.


- **ვენის დიაგრამაზე** ჩვენ ვგულისხმობთ, რომ **ხდომილებათა სივრცე** არის **მართკუთხედი** და წარმოვადგენთ **მოვლენებს** მართკუთხედის შიგნით დახატული **წრეებით**:

  <center><img src="images/L1_Vienn_Diagram.png" width="600" alt="Example" /></center>

- სურათიდან გამომდინარე:

  $$A \cap B = \text{ რეგიონი 1 და 2}$$
  $$A \cup B = \text{ რეგიონი 1, 2, 3, 4, 6 და 7}$$
  $$\overline{A} = \text{ რეგიონი 3, 5 და 6}$$
  
  $$B \cap C = \text{ რეგიონი ?}$$
  $$C \cap A = \text{ რეგიონი ?}$$
  

  $$B \cup C = \text{ რეგიონი ?}$$
  $$C \cup A = \text{ რეგიონი ?}$$
  

  $$\overline{B} = \text{ რეგიონი ?}$$
  $$\overline{C} = \text{ რეგიონი ?}$$
 

- რამდენიმე შედეგი, რომელიც გამომდინარეობს ზემოხსენებული განმარტებებიდან 
  1. $A \cap \varnothing = \varnothing$.
  2. $A \cup \varnothing = A$.
  3. $A \cap \overline{A} = \varnothing$.
  4. $A \cup \overline{A} = S$.
  5. $\overline{S} = \varnothing$.
  6. $\overline{\varnothing} = S$.
  7. $\overline{A \cup B} = \overline{A} \cap \overline{B}$.
  8. $\overline{A \cap B} = \overline{A} \cup \overline{B}$.
  9. $\overline{\overline{A}} = A.$
  
  ეს შედეგები ადვილად შეიძლება გადამოწმდეს ვენის დიაგრამების საშუალებით.

<h2 align="center">1.3 ხდომილებათა სივრცის ელემენტების დათვლა</h2>

- ერთერთი პრობლემა, რომელიც სტატისტიკოსმა უნდა გაითვალისწინოს და შეეცადოს შეაფასოს არის შემთხვევითობის ელემენტი, რომელიც დაკავშირებულია ექსპერიმენის ჩატარების დროს გარკვეული მოვლენების წარმოშობაზე.


- ხშირ შემთხვევაში, ჩვენ შეგვიძლია ამოვხსნათ ალბათობის პრობლემა, ხდომილებათა სივრცეში ელემენტების რაოდენობის დათვლით, თითოეული ელემენტის რეალურად ჩამოთვლის გარეშე.


- დათვლის ფუნდამენტური პრინციპი, რომელსაც ხშირად უწოდებენ **გამრავლების წესს**, აღიწერება შემდეგ წესში:

 - **წესი 1.1 (გამრავლების წესი)**: თუ ოპერაცია შეიძლება შესრულდეს $n_1$ გზით, და თუ თითოეული ამ მეთოდისთვის შეიძლება შესრულდეს მეორე ოპერაცია $n_2$ გზით, მაშინ ეს ორი ოპერაცია შეიძლება შესრულდეს ერთად $n_1 \cdot n_2$ გზით.

- **მაგალითი 1.13**: 
  
  რამდენი ელემენტია ხდომილებათა სივრცეში, როდესაც ერთხელ გავაგორებთ წყვილ კამათელს?
  
  **პასუხი**: პირველი კამათელი შეიძლიბა დაეშვას $n_1 = 6$ განსხვავებული გზით. თითოეული ამ $6$ შემთხვევისთვის, მეორე კამათელი ასევე შეიძლება დაეშვას $n_2 = 6$ გზით. ამიტომ,  წყვილი კამათლი შეიძლება დაეშვას $n = n_1 \cdot n_2 = 6 \cdot 6 = 36$ შესაძლო გზით.

- **მაგალითი 1.14**: 
  
  ამხანაგობას, რომელშიც გაწევრიანებული 57 ადამიანი, სჭირდება **თავმჯდომარისა** და **ხაზინადარის** არჩევა. 
  **რამდენი განსხვავებული გზით** შეიძლება ამ ორის ადამიანის არჩევა?

 - **წესი 1.2 (განზოგადებული გამრავლების წესი)**: თუ ოპერაცია შეიძლება შესრულდეს $n_1$ გზით, და თუ თითოეული მათგანისთვის შეიძლება შესრულდეს მეორე ოპერაცია $n_2$ გზით, და პირველი ორიდან თითოეული მესამე ოპერაცია შეიძლება შესრულდეს $n_3$ გზით და ა.შ., მაშინ $k$ თანმიმდევრობის ოპერაციების შესრულება შეიძლება $n_1 \cdot n_2 \cdot \cdots \cdot  n_k$ გზით.

- **განმარტება 1.7**: ობიექტების მთლიანი ან ნაწილობრივი გადალაგებას ეწოდება **პერმუტაცია**.

- **განმარტება 1.8**: ნებისმიერი არა-უარყოფითი მთელი $n$ რიცხვისთვის, ამ რიცხვის **ფაქტორიალი** განისაზღვრება შემდეგნაირად: 

  $$n! = 1 \cdot 2 \cdot \cdots \cdot (n - 1) \cdot n,$$
  
  განსაკუთრებული შემთხვევით $0! = 1$.

- **თეორემა 1.1**: $n$ ობიქეტის ყველანაირი გადალაგების რაოდენობა არის $n!$.

- **თეორემა 1.2**: $n$ ობიქეტიდან $k$ დალაგებული ობიექტის არჩევის რაოდენობა ტოლია:

  $$A_{n}^{k} = 
\begin{pmatrix}
n \\
k
\end{pmatrix}
=\frac{n!}{(n-k)!}$$

- **თეორემა 1.3**: $n$ ობიექტის არჩევის რაოდენობა, რომელთა $n_1$ არის პირველი ტიპის, $n_2$ არის მეორე ტიპის. ... ... , $n_k$ არის  $k$-ი ტიპის, არის:

  $$A_{n}^{n_1, n_2, ... , n_k} = 
\begin{pmatrix}
n \\
n_1, n_2, ..., n_k
\end{pmatrix}= 
\frac{n!}{n_1! n_2! \cdots n_k!},$$

  სადაც $n_1 + n_2 + \cdots + n_k = n$.

- **თეორემა 1.4**: $n$ ობიექტიდან $k$ დაულაგებელი ობიექტის არჩევის რაოდენობა ტოლია:

  $$C_{n}^{k} = 
\begin{pmatrix}
n \\
k, n-k
\end{pmatrix}= 
\frac{n!}{k! (n-k)!}.$$

<h2 align="center">1.4 მოვლენის ალბათობა</h2>

- ალბათ ეს იყო კაცობრიობის აზარტული თამაშების დაუოკებელი წყურვილი, რამაც გამოიწვია ალბათობის თეორიის ადრეული განვითარება. 


- მოგების გაზრდის მიზნით, აზარტული მოთამაშეები მოუწოდებდნენ მათემატიკოსებს მოეფიქრებინათ სტრატეგიები აზარტული თამაშებისთვის. 


- ზოგიერთი მათემატიკოსი, რომლებიც ამ სტრატეგიას გვთავაზობდნენ, იყვნენ **პასკალი**, **ლაიბნიცი**, **ფერმა** და **ჯეიმს ბერნული**.

- **განმარტება 1.9**: $A$ მოვლენის **ალბათობა** არის $A$-ში მყოფი ყველა ხდომილების სივრცის ელემენტების წონის ჯამი:

  $$0 \leq P(A) \leq 1.$$
  $$P(\varnothing) = 0.$$
  $$P(S) = 1.$$
  
  ამის გარდა, თუ $A_1, A_2, A_3, ...$ არის ურთიერთგამომრიცხავი მოვლენების თანმიმდევრობაა, მაშინ:
  
  $$P(A_1 \cup A_2 \cup A_3 \cup \cdots ) = P(A_1) + P(A_2) + P(A_3) + \cdots$$

 - **წესი 1.3**: თუ ექსპერიმენტის ჩატარების შედეგად შესაძლებელია $N$ განსხვავებული შედეგის მიღება თანაბარი სიხშირით, და თუ ამ შედეგებიდან ზუსტად $n$ შედეგი შეესაბამება $A$ მოვლენას, მაშინ $A$ მოვლენის ალბათობაა:
 
 $$P(A) = \frac{n}{N}.$$

<h2 align="center">1.5 ალბათობის დამატების წესი</h2>

- **თეორემა 1.5**:  თუ $A$ და $B$ არის ორი მოვლენა, ხოლო $P(A)$ და $P(B)$ არის ამ მოვლენათა ალბათობები, მაშინ:
  
  $$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$
  
  
  <center><img src="images/L1_Additive_Rule.png" width="400" alt="Example" /></center>


**თეორემიდან გამომდინარე შედეგები**:


- თუ $A$ და $B$ მოვლენები არიან ურთიერთგამომრიცხავია, მაშინ:

  $$P(A \cup B) = P(A) + P(B).$$
  
  
- თუ $A_1, A_2, ..., A_n$ ურთიერთგამომრიცხი სიმრავლეებია, მაშინ:

  $$P(A_1 \cup A_2 \cup \cdots \cup A_n) = P(A_1) + P(A_2) + \cdots + P(A_n).$$
  
  
- თუ $A_1, A_2, ..., A_n$ სიმრავლეები წარმოადგენენ $S$ ხდომილებათა სივრცის დაყოფას, მაშინ:

  $$P(A_1 \cup A_2 \cup \cdots \cup A_n) = P(A_1) + P(A_2) + \cdots + P(A_n) = P(S) = 1.$$

- **თეორემა 1.6**:  თუ $A$, $B$ და $C$ მოვლენების ალბათობებია, შესაბამისად, $P(A)$, $P(B)$, და $P(C)$ , მაშინ:

  $$P(A \cup B \cup C) = P(A) + P(B) + P(C) - P(A \cap B) - P(B \cap C)  - P( C \cap A) + P(A \cap B \cap C).$$

  <center><img src="images/L1_ABC_Diagram.png" width="600" alt="Example" /></center>

- **თეორემა 1.7**: თუ $\overline{A}$ არუს $A$ მოვლენის დამარტება, მაშინ:

  $$P(A) + P(\overline{A}) = 1.$$

<h2 align="center">1.6 პირობითი ალბათობა, მოვლენების დამოუკიდებლობა და გამრავლეიბს წესი</h2>

- $B$ მოვლენის ალბათობა, როდესაც ცნობილია, რომ რაღაც $A$ მოვლენა მოხდა, **პირობით ალბათობას** უწოდებენ და აღინიშნება $P (B | A)$.


- სიმბოლო $P (B | A)$ ჩვეულებრივ იკითხება: $B$**-ს ალბათობა როცა მოხდა $A$**.

- **განმარტება 1.10**: თუ $A$ **მოვლენის ალბათობა არანულოვანია**, მაშინ **პირობითი ალბათობა** $P(B | A)$ განისაზღვრება როგორც:

  $$P(B | A) = \frac{P(A \cap B)}{ P(A)}.$$

- **განმარტება 1.11**: ორი $A$ და $B$ მოვლენა **დამოუკიდებელია**, თუ:

  $$P(B|A) = P(B) \text{  ან  } P(A | B) = P(A),$$ 
  
  პირობითი ალბათობების არსებობის ვარაუდით. 
  
  **წინააღმდეგ** შემთხვევაში, $A$ და $B$ მოვლენები **დამოკიდებულნი** არიან.

- **თეორემა 1.8**: თუ ორივე $A$ და $B$ მოვლენების **ალბათობები არანულოვანია**, მაშინ:

  $$P(A \cap B) = P(A) P(B |A).$$

  

- **თეორემა 1.9**: ორი $A$ და $B$ მოვლენა დამოუკიდებელია მაშინ და მხოლოდ მაშინ, თუ:

  $$P(A \cup B) = P(A)P(B)$$
  
  ანუ, იმის გასაგებად არის თუ არა ორი მოვლენა დამოუკიდებელი, ჩვენ უბრალოდ ვპოულობთ მათ ინდივიდუალური ალბათობის ნამრავლს.
  

- **თეორემა 1.10**: თუ $A_1, A_2, ..., A_n$ მოვლენების ალბათობები არანულოვანია, მაშინ:

  $$P(A_1 \cap A_2 \cap \cdots \cap A_n) = P(A_1)P(A_2|A_1)P(A_3| A_2 \cap A_1) \cdots P(A_n | A_1 \cap A_2 \cap \cdots A_{n-1}).$$
  
  თუ $A_1, A_2, ..., A_n$ მოვლენები დამოუკიდებლები არიან, მაშინ:
  
  $$P(A_1 \cap A_2 \cap \cdots \cap A_k) = P(A_1)P(A_2) \cdots P(A_n).$$

- **განმარტება 1.12**: ორი $A$ და $B$ მოვლენა პირობითად დამოუკიდებელია $C$ მოვლენის მიმართ, თუ:

  $$P(A\cap B |C) = P(A|C) P (B|C).$$
  
  თუ $A$ და $B$ მოვლენები დამოუკიდებელია, ისინი პირობითად დამოუკიდებელია.
  
  პირიქით არ არის სწორი.
