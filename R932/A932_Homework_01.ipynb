<h3 align="center">ამოცანა №1</h3>

დავუშვათ $(\Omega, \Sigma, P)$ არის ხდომილებათა სივრცე, სადაც: 

$\Omega = \{1, 2, 3, 4\}$;

$\Sigma = 2^\Omega$;

$P(i) = \frac{1}{4}$ ყოველი $i = 1, 2, 3, 4$.


დავუშვათ $A = \{1,2\}$, $B = \{1,2\}$ და $C = \{1,2\}$ არის მოვლენები ამ ხდომილებათა სივრცეზე.


- აჩვენეთ, რომ $A$, $B$ და $C$ არის **წყვილ-წვყილად დამოუკიდებელი** მოვლენები, მაგრამ არა **დამოუკიდებელი მოვლენები**.

<h3 align="center">ამოცანა №2</h3>

**სამი მოთხილამურე** დამოუკიდებლად ეშვება მთიდან. 

თითოეული $\frac{1}{3}$ ალბათობით შეიძლება დავარდეს და, **ორჯერ დავარდნის** შემდეგ ის ტოვებს ასპარეზობას. 

- იპოვნეთ **ალბათობა** იმისა, რომ **ერთი მოთხილამურე მაინც მივა ფინიშამდე**.

<h3 align="center">ამოცანა №3</h3>

მოცემულია $X$ და $Y$ შემთხვეითი სიდიდეები, შესაძლო მნიშვნელობებით $x = 1, 2, 3$ და $y = 1, 2, 3$.

- იპოვეთ $c$-ს მნიშვნელობა, რომლისთვისაც ფუნქციით $f(x,y) = cxy$ იქნება $X$ და $Y$ შემთხვეითი სიდიდეების **ერთობლივი ალბათური განაწილების** ფუნქცია.

<h3 align="center">ამოცანა №4</h3>

  მარათონში მონაწილე კაცების რაოდენობა $X$ და ქალი რაოდენობა $Y$ აღწერილია **ერთობლივი სიმკვრივის ფუნქციით**:
  
  $$f(x,y)=
  \left\{\begin{matrix}
  8xy, & 0 \leq y \leq x \leq 1, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$
  
  იპოვნეთ $X$ და $Y$ შემთხვევითი სიდიდეების **კოვარიაცია**.

  სადაც $X$ ღებულობს მნიშნველობებს $x = \{-2, 0, 1, 7\}$, ხოლო $Y$ ღებულობს მნიშნველობებს $y = \{-6, 0, 3\}$.
  
  
(ა) იპოვეთ $\alpha$-ს **მნიშვნელობა**;

(ბ) გამოთვალეთ $X$-ის **ზღვრული განაწილება**, ანუ $\operatorname{P}(X = x_i)$;

(გ) გამოთვალეთ $X$-ის **მათემატიკური ლოდინი** და **დისპერსია**, ანუ $\mu_X$ და $\sigma^2_X$;

(დ) გამოთვალეთ $Y$-ის **ზღვრული განაწილება**, ანუ $\operatorname{P}(Y = y_i)$;

(ე) გამოთვალეთ $Y$-ის **მათემატიკური ლოდინი** და **დისპერსია**, ანუ $\mu_Y$ და $\sigma^2_Y$;

(ვ) გამოთვალეთ $X$-ის და $Y$-ის **კოვარიაცია**, ანუ $\sigma_{XY}$;

(ზ) გამოთვალეთ $X$-ის და $Y$-ის **კორელაციის კოეფიციენტი**, ანუ $\rho_{XY}$.
