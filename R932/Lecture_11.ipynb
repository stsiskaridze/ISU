<h2 align="center">ნიმუშების ფუნდამენტალური განაწილებები</h2>


<h3 align="center">შემთხვევითი ნიმუში</h3>

- ამ ლექციაში ჩვენ ყურადღებას გავამახვილებთ **პოპულაციებზე** და **ნიმუშებზე**, და შევისწავლით ისეთ მნიშვნელოვან სიდიდეებს, როგორიცაა **ნიმუშის საშუალო** და **ნიმუშის ვარიაცია**, რომელსაც სასიცოცხლო მნიშვნელობა ექნება მომავალ მასალის ასახსნელად.

- <h3>განმარტება 11.1:</h3>

  **პოპულაცია (population)** წარმოადგენს დაკვირვებად სიდიდეების ერთობლიობას, რომელსაც ჩვენ ვაკვირდებით.

-  დაკვირვებების რაოდენობა პოპულაციაში განისაზღვრება როგორც **პოპულაციის ზომა**.
  
   მაგალითად, თუ სკოლაში 600-ი მოსწავლეა, რომლებიც დაჯგუფებულია სისხლის ჯგუფის მიხედვით, ჩვენ ვამბობთ, რომ პოპულაციის ზომა 600-ის ტოლია. 

- პოპულაციაში თითოეული დაკვირვება არის $X$ შემთხვევითი სიდიდის მნიშვნელობა, რომელსაც აქვს გარკვეული ალბათობის განაწილება $f(x)$.


- მაგალითად, თუ ადამიანი ამოწმებს არის თუ არა დეტალი დეფექტიანი, მაშინ პოპულაციაში თითოეული დაკვირვება იქნება ბერნულის $X$ შემთხვევითი სიდიდის $0$ ან $1$ მნიშვნელობა ალბათობის განაწილებით:

  $$b(x; 1, p) = p^x q ^{1-x},$$

  სადაც $x = 1$ მიუთითებს წუნდებულ დეტალზე ხოლო 0 მიუთითებს არაწუნდებულ დეტალზე.
 
 
- როდესაც ლაპარაკი იქნება **ბინომიალურ პოპულაციზე**, **ნორმალურ პოპულაციზე** ან, ზოგადად, **პოპულაციზე $f(x)$**, ჩვენ ვიგულისხმებთ პოპულაციას, რომლის დაკვირვებაც არის შემთხვევითი სიდიდე, რომელსაც აქვს **ბინომიალური განაწილება**, **ნორმალური განაწილება**, ან **ალბათური განაწილება** $f (x)$. 


- მაშასადამე, **შემთხვევითი სიდიდის საშუალო** და **ვარიაცია** ასევე მოიხსენიება როგორც შესაბამისი **პოპულაციის საშუალო** და **ვარიაცია**.

- ხშირად, საჭიროა პოპულაციზე დასკვნების გაკეთება იმ შემთხვევაში, როდესაც შეუძლებელია ან არაპრაქტიკულია დაკვირვებათა სრული ჯგუფის შესწავლა. 


- მაგალითად, გარკვეული ბრენდის ნათურის სიცოცხლის საშუალო ხანგრძლივობის დადგენისას შეუძლებელია ყველა ასეთი ნათურის შემოწმება. 


- გადაჭარბებული ხარჯები შეიძლება ასევე იყოს მთლიანი პოპულაციის შესწავლის შემზღუდავი ფაქტორი. 


- ამიტომ, ჩვენ უნდა დავერყდნოთ პოპულაციის დაკვირვების ქვესიმრავლეზე, რათა დასკვნა გავაკეთოთ თავდაპირევლ პოპულაციაზე.


- ამას მივყავართ **ნიმუშის ცნებასთან**.

- <h3>განმარტება 11.2:</h3>

  **ნიმუში (sample)** წარმოადგენს პოპულაციის ქვესიმრავლეს.

- თუ ჩვენი პოპულაციაზე გაკეთებული დასკვნები რაღაც ნიმუშის შესწავლის საფუძველზე მართებული იქნება, ჩვენ უნდა მივიღოთ ისეთი ნიმუშები, რომლებიც პოპულაციას წარმოადგენენ.


- ძალიან ხშირად ნიმუშად ჩვენ ვცდილობთ ავირჩიოთ პოპულაციის ყველაზე მოსახერხებელი წევრები, რამაც შეიძლება პოპულაციაზე  არასწორი დასკვნების გაკეთება გამოიწვიოს.


- შერჩევის ნებისმიერი **პროცედურა**, რომელიც წარმოშობს დასკვნებს, რომლებიც მუდმივად ან ნაკლებად გადაჭარბებულად აფასებს პოპულაციის გარკვეულ მახასიათებლებს, ეწოდება **დაძრული (biased)**.


- მოდით $f(x)$ პოპულაციიდან $n$ ზომის შემთხვევითი ნიმუშის არჩევისას განვსაზღვროთ შემთხვევითი სიდიდეები $X_1, X_2, ..., X_n$, სადაც $i$-ი სიდიდე წარმოადგენს $i$-რ ნიმუშის მნიშვნელობას, რომელსაც ჩვენ ვაკვირდებით. 

  შემთხვევითი ცვლადები $X_1, X_2,..., X_n$ შემდეგ წარმოადგენს შემთხვევით ნიმუშს $f (x)$ პოპულაციიდან რიცხვითი ნიშვნელობებით $x_1, x_2,. . . , x_n$ თუ გაზომვები მიიღება $n$ დამოუკიდებელი ექსპერიმენტის  ჩატარებით არსებითად იგივე პირობებში.

- <h3>განმარტება 11.3:</h3>

  დავუშვათ $X_1, X_2,..., X_n$ წარმოადგენს $n$ დამოუკიდებელ შემთხვევით სიდიდეებს ერთი და იგივე ალბათური განაწილების ფუნქციით $f(x)$.
  
  ჩვენ შეგვიძლია განსვსაზღვროთ $X_1, X_2,..., X_n$ როგორც $n$ ზომის პოპულაციის **შემთხვევითი ნიმუში** და ჩავწეროთ მისი ერთობლივი ალბათური განაწილება როგორც:
  
  $$f(x_1, x_2, ..., x_n) = f(x_1)f(x_2)\cdots f(x_n).$$

- ჩვენი მთავარი მიზანი **შემთხვევითი ნიმუშის შერჩევასას** არის ინფორმაციის ამოღება **პოპულაციის უცნობი პარამეტრების** შესახებ.


- დავუშვათ, მაგალითად, რომ ჩვენ გვინდა დასკვნა გავაკეთოთ იმის შესახებ, თუ საქართველოში მოსახლეობის რა ნაწილს ურჩევნია ღვინის გარკვეული ბრენდი. 

  სამწუხაროდ, შეუძლებელია ყველას (მოსახლეობის ყველა წარმომადგენლის) გამოკითხვა რათა გამოვთვალოთ ამ პარამეტრის მნიშვნელობა. 

  ამის ნაცვლად, ჩვენ ვირჩევთ **დიდ შემთხვევით ნიმუშს** (მოსახლეობის დიდ რაოდენობას) და გამოვთვლით ამ ნიმუშის მაგალითზე მოცემული ღვინის ბრენდის მომხრე ადამიანების წილს. 

  ამისთვის გამოიყენება $\hat{p}$ სიდიდე, რომელიც $p$-ს ნამდვილ მნიშვნელობაზე დასკვნის გაკეთების საშალებას მოგვცემს.

- <h3>განმარტება 11.4:</h3>

  შემთხვევითი სიდიდის ნებისმიერი ფუნქციას, რომელიც წარმოადგენს შემთხვევით ნიმუშს, ეწოდება **სტატისტიკა**.

<h3 align="center">ნიმუშების ადგილმდებარეობა და ცვალებადობა</h3>

- წინა ლექციებზე ჩვენ შემოვიტანეთ ორი პარამეტრი $\mu$ და $\sigma^2$, რომლებიც ზომავს ალბათური განაწილების **ადგილმდებარეობის ცენტრს** და ალბათური განაწილების **ცვალებადობას**.


- ეს სიდიდეები პოპულაციის **მუდმივი პარამეტრებია** და შემთხვევითი ნიმუშის არჩევა არანაირ გავლენას არ ახდენს მათზე.


- ამასთან, ჩვენ განვსაზღვრავთ რამდენიმე მნიშვნელოვან სტატისტიკას, რომელიც აღწერს შემთხვევითი ნიმუშის შესაბამის პარამეტრებს:


- **ნიმუშის საშუალო**:
  
  $$\overline{x} = \frac{1}{n} \sum_{i=1}^{n} x_i.$$
  

- **ნიმუშის მედიანა**:
  
  $$\widetilde{x} = 
  \left\{\begin{matrix}
  x_{(n+1)/2},&  \text{ თუ } n \text{ კენტია,} \\
  \frac{1}{2}(x_{n/2} + x_{n/2 + 1}),& \text{ თუ } n \text{ ლუწია.}
  \end{matrix}\right.$$
  
  
- **ნიმუშის მოდი**: ყველაზე ხშირი მნიშვნელობა რომელიც ნიმუშში გვხვდება.

- <h3>მაგალითი 11.1:</h3>
  
  დავუშვათ ნიმუში შედგება შემდეგი დაკვირვებებისგან:
  
  $$0.32 \text{ } 0.53 \text{ } 0.28 \text{ } 0.37 \text{ } 0.47 \text{ } 0.43 \text{ } 0.36 \text{ } 0.42 \text{ } 0.38 \text{ } 0.43.$$
  
  გამოთვალეთ ნიმუშის საშუალო, მედიანა და მოდი. 

- განვიხილოთ შემდეგი გაზომვები (ლიტრებში) $A$ და $B$ კომპანიების მიერ ჩამოსხმული ღვინის ორი ნიმუშისთვის:

|           |      |      |      |      |      |
|-----------|------|------|------|------|------|
| ნიმუში $A$| 0.97 | 1.00 | 0.94 | 1.03 | 1.06 |
| ნიმუში $B$| 1.06 | 1.01 | 0.88 | 0.91 | 1.14 |

- ორივე ნიმუშს აქვს ერთი და იგივე საშუალო, 1,00 ლიტრი. 

  აშკარაა, რომ კომპანია $A$ ბოთლებში ღვინოს უფრო ერთგვაროვანი შემცველობით ავსებს ვიდრე კომპანია $B$. 
  
  ჩვენ ვამბობთ, რომ საშუალოზე დაკვირვების **ცვალებადობა** ან **დისპერსია** $A$ ნიმუშზე ნაკლებია ვიდრე $B$ ნიმუშზე.
  

- **ნიმუშის ვარიაცია**:

  $$S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \overline{X})^2$$
  

- **ნიმუშის სტანდარტული გადახრა**:

  $$S = \sqrt{S^2}.$$
 
- **ნიმუშის დიაპაზონი**:

  $$R = X_{\max} - X_{\min}.$$

- <h3>თეორემა 11.1:</h3>
  
  თუ $S^2$ არის $n$ ზომის შემთხვევითი ნიმუშის ვარიაცია, მაშინ:
  
  $$S^2 = \frac{1}{n(n-1)}\left [ n\sum_{i=1}^{n} X_i^2 - \left ( \sum_{i=1}^{n} X_i \right )^2 \right].$$

- <h3>მაგალითი 11.2:</h3>
  
  ბათუმის $4$ შემთხვევით შერჩეულ სასურსათო მაღაზიაში ფორთოხლის ფასების შედარებამ აჩვენა, რომ წინა თვესთან შედარებით $1$ კილოგრამის ფასი გაიზარდა $12$, $15$, $17$ და $20$ თეთრით. 
  
  იპოვნეთ ამ ფასების ზრდის **შემთხვევითი ნიმუშის ვარიაცია**.

- <h3>მაგალითი 11.3:</h3>
  
   მოცემულია შემთხვევითი ნიმუში  $\{3, 4, 5, 6, 6, 7\}$, რომელიც წარმოადგენს კალმახის რაოდენობას, რომელიც დაიჭირეს $6$ მეთევზემ ლისის ტბაზე.
   
   იპოვნეთ **შემთხვევითი ნიმუშის ვარიაცია**.

- <h3>მაგალითი 11.4:</h3>

  დავუშვათ $X_1, X_2, ..., X_n$ არის დამოუკიდებელი და ერთნაირად განაწილებული შემთხვევითი სიდიდე **მათემატიკური ლოდინით** $\mu$ და **ვარიაციით** $\sigma^2$.

  **ნიმუშის საშუალო** და **შეუსწორებელი ნიმუშის ვარიაცია** განმარტებულია როგორც:

$$\overline{X} = \frac{1}{n}\sum_{i=1}^{n} X_i \text{ და } S^2 = \frac{1}{n}\sum_{i=1}^{n} \left ( X_i - \overline{X} \right )^2.$$

  ა) გამოთვალეთ **ნიმუშის საშუალო** და შეადარეთ ის **პოპულაციის საშუალოს** $\mu$.

  ბ) გამოთვალეთ **შეუსწორებელი ნიმუშის ვარიაციის მოსალოდნელი მნიშვნელობა** $\mathbf{E}[S^2]$, და შეადარეთ ის **პოპულაციის ვარიაციას** $\sigma^2$.

  გ) როგორ უნდა **შეიცვალოს** ნიმუშის ვარიაცია რომ მივიღოთ **პოპულაციის ვარიაციის მიუკერძოებელი შემფასებლის** $\sigma^2$.


- <h3>ამოხსნა 11.4:</h3>

  ა) **ნიმუის საშუალო** მნიშვნელობა გამოითვლება შემდეგნაირად: 
  
  $$\overline{X}=\frac{1}{n}\sum_{i=1}^{n}X_{i}.$$
  
  ხოლო **პოპულაციის საშუალო** მოცემულია როგორც: 
  
  $$\mu=\frac{1}{N}\sum_{i=1}^{N}X_{i},$$
  
  სადაც $N$ არის პოპულაციის ზომა. 
  
  იოლი საჩვენებელია, რომ ნიმუშის საშუალოს მოსალოდნელი მნიშვნელობა ტოლია პოპულაციის საშუალოს მოსალოდნელ მნიშვნელობის:
  
  $$\operatorname {E}(\overline{X}) = \operatorname {E} \left ( \frac{1}{n}\sum_{i=1}^{n} X_i \right )=\frac{1}{n}\sum_{i=1}^{n} \operatorname {E}(X_i) =  \frac{1}{n}\sum_{i=1}^{n} \mu = \mu.$$
   
  ბ) $$\begin{aligned}\operatorname {E} [S^2] &=\operatorname {E} \left[{\frac {1}{n}}\sum _{i=1}^{n}\left(X_{i}-\overline{X}\right)\right]\\[5pt]&=\operatorname{E} \left[ {\frac {1}{n}}\sum _{i=1}^{n}\left(X_{i}-{\frac {1}{n}}\sum _{j=1}^{n}X_{j}\right)^{2}\right]={\frac {1}{n}}\sum _{i=1}^{n}\operatorname {E} \left[X_{i}^{2}-{\frac {2}{n}}X_{i}\sum _{j=1}^{n}X_{j}+{\frac {1}{n^{2}}}\sum _{j=1}^{n}X_{j}\sum _{k=1}^{n}X_{k}\right]\\[5pt]&={\frac {1}{n}}\sum _{i=1}^{n}\left[{\frac {n-2}{n}}\operatorname {E} [X_{i}^{2}]-{\frac {2}{n}}\sum _{j\neq i}^{n}\operatorname {E} [X_{i}X_{j}]+{\frac {1}{n^{2}}}\sum _{j=1}^{n}\sum _{k\neq j}^{n}\operatorname {E} [X_{j}X_{k}]+{\frac {1}{n^{2}}}\sum _{k=1}^{n}\operatorname {E} [X_{k}^{2}]\right]\\[5pt]&={\frac {1}{n}}\sum _{i=1}^{n}\left[{\frac {n-2}{n}}(\sigma ^{2}+\mu ^{2})-{\frac {2}{n}}(n-1)\mu ^{2}+{\frac {1}{n^{2}}}n(n-1)\mu ^{2}+{\frac {1}{n}}(\sigma ^{2}+\mu ^{2})\right]\\[5pt]&=\frac{1}{n}\sum_{i=1}^n\left[{\frac {n-1}{n}}(\sigma ^{2}+\mu^{2})-\frac{n}{n-1}\mu^{2}\right]=\frac{1}{n}\sum_{i=1}^n\left[{\frac {n-1}{n}}(\sigma ^{2})\right]={\frac {n-1}{n}}\sigma ^{2}.\end{aligned}$$
  
  გ) იმისთვის, რომ პოპულაციის ვარიაცია მიუკერძოებელი შემფასებელი გახდეს, საჭიროა ნიმუშის საშუალო გამრავლდეს $\frac{n}{n-1}$-ზე: 
  
  $$n\operatorname{E}=(n-1)\sigma^{2}\text{ და } \sigma^{2}=\frac{n}{n-1}\operatorname{E}$$

- <h3>განმარტება 11.5:</h3>

  სტატისტიკის ალბათობის განაწილებას ეწოდება **ნიმუშის განაწილება**.

- სტატისტიკის ნიმუშის განაწილება დამოკიდებულია **პოპულაციის განაწილებაზე**, **ნიმუშების ზომაზე** და **ნიმუშების არჩევის მეთოდზე**!

- პირველი **მნიშვნელოვანი ნიმუშის განაწილება**, რომელსაც ჩვენ განვიხილავთ არის **საშუალო** $\overline{X}$.

  დავუშვათ, რომ შემთხვევითი ნიმუში შედგება **ნორმალური პოპულაციის** $n$ დაკვირვებისგან ლოდინით $\mu$ და ვარიაციით $\sigma^2$.
  
  შესაბამისად, ჩვენ გვაქვს, რომ:

  $$\overline{X} = \frac{1}{n}(X_1 + X_2 + \cdots  + X_n)$$
  
  არის **ნორმალური განაწილება** შემდეგი პარამეტრებით:
  
  $$\mu_{\overline{X}}=\frac{1}{n}(\mu + \mu + \cdots + \mu) = \mu \text{ და } \sigma^2_{\overline{X}} = \frac{1}{n^2}(\sigma^2 + \sigma^2 + \cdots + \sigma^2) = \frac{\sigma^2}{n}.$$

- თუ ჩვენ გვაინტერესებს **უცნობი განაწილების მქონე პოპულაცია**, სასრული ან უსასრულო, $\overline{X}$-ის ნიმუშის განაწილება მაინც იქნება **ნორმალური განაწილება** ლოდინით $\mu$ და ვარიაციით $\sigma^2/n$, იმ პირობით, **რომ ნიმუშის ზომა დიდია**. ეს საოცარი შედეგი არის შემდეგი თეორემის უშუალო შედეგი, რომელსაც ეწოდება **ცენტრალური სასაზღვრო თეორემა**:

- <h3>თეორემა 11.2 (ცენტრალური სასაზღვრო თეორემა):</h3>
  
  თუ მოცემულია პოპულაცია ლოდინით $\mu$ და ვარიაციით $\sigma^2$, და თუ $\overline{X}$ არის ამ პოპულაციიდან მიღებული $n$ ზომის შემთხვევითი ნიმუშის საშუალო, მაშინ ალბათური განაწილება $Z$:
  
  $$Z = \frac{\overline{X}-\mu}{\sigma/\sqrt{n}}$$
  
  **მიისწრაფის ნორმალურ განაწილებისკენ** $\mathcal{N}(z; 0, 1)$, როცა $n \rightarrow \infty$.

- ცენტრალური სასაზღრო თეორემის გამოსაყენებლად ხელმძღვანელობენ $n \geq 30$ ნიმუშის ზომით:

- მეორე **მნიშვნელოვანი ნიმუშის განაწილება**, რომელსაც ჩვენ განვიხილავთ არის **სტატისტიკა** $S^2$.

- <h3>თეორემა 11.3:</h3>
  
  თუ მოცემულია პოპულაცია ლოდინით $\mu$ და ვარიაციით $\sigma^2$, და თუ $S^2$ არის ამ პოპულაციიდან მიღებული $n$ ზომის შემთხვევითი ნიმუშის ვარიაცია, მაშინ სატისტიკა:
  
  $$\chi^2 = \frac{(n-1)S^2}{\sigma^2} = \sum_{i=1}^{n} \frac{(X_i - \overline{X})^2}{\sigma^2}$$
  
  არის განაწილებული $\chi^2$-განაწილებით $v = n-1$ თავისუფლების ხარისხით.
