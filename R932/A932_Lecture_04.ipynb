<h1 align="center">მათემატიკური ლოდინი</h1>


<h3 align="center">შემთვევითი სიდიდის მათემატიკური ლოდონი</h3>

- განვიხილიოთ ორი მონეტის აგდების მაგალითი.


- თუ მონეტები ააგდეს 16-ჯერ და თუ $X$ არის თითოეული აგდების შედეგად მიღებული საფასურის რაოდენობა, მაშინ $X$ არის შემთხვევითი სიდიდე რომელიც ღებულობს მნიშვნელობებს $0$, $1$ და $2$.


- დავუშვათ, ორი მონეტიდან ორივე საფასური ამოვიდა 5-ჯერ, ერთი გერბი და ერთი საფასური ამოვიდა 7-ჯერ და ორივე გერბი ამოვიდა 4-ჯერ, მაშინ, $X$ **შემთხვევითი სიდიდის საშუალო მნიშნველობა** იქნება:


  $$\frac{(0)\cdot(4) + (1)\cdot(7) + (2)\cdot(5)}{16} = 1.06.$$


- როგორც ხედავთ, ის არ ემთხვევა $X$ შემთხვევითი სიდიდის შესაძლო შედეგებს $\{0, 1, 2\}$.


- მოდით, მოვახდინოთ ჩვენი გამოანგარიშების ფორმულის შემდეგი ეკვივალენტური ფორმით ჩაწერა:

  $$(0) \cdot \left ( \frac{4}{16} \right ) + (1) \cdot \left ( \frac{7}{16} \right ) + (2) \cdot \left ( \frac{5}{16} \right ) = 1.06.$$
  
  
- ასეთ ჩანაწერში, რიცხვები $\frac{4}{16}$, $\frac{7}{16}$ და $\frac{5}{16}$ წარმოადგენენ **ფართობით სიხშირეებს**. 


- **ფარდობითი სიხშირეების** ეს მეთოდი **გამოიყენება** ორი მონეტის აგდების შედეგად მიღებული საფასურის **საშუალო რაოდენობის გამოთვლისთვის**, რომელსაც უნდა **ველოდოთ გრძელვადიან პერსპექტივაში**.


- ეს საშუალო მნიშვნელობას განიმარტება როგორც $X$ **შემთხვევითი ცვლადის საშუალო მნიშვნელობა** და ხშირად აღინიშნება როგორც $\mu_{X}$, ან უბრალოდ$\mu$, როდესაც ცხადია რომელ შემთხვევით სიდიდეზეა ლაპარაკი.

  სტატისტიკოსები ამ მნიშვნელობას ხშირად მოიხსენიებენ როგორც **მათემატიკურ ლოდის**, და იყენებენ აღნიშვნას $\mathbb{E} [X]$.

- მოდით განვიხილოთ ნამდვილი მონეტის ორჯერ აგდების მაგალითი.

  როგორც ჩვენ წინა ლექციებში ვნახეთ, ამ შემთხვევაში ხდომილებათა სივრცე იქნება:
  
  $$S = \{HH, HT, TH, TT\}$$.

  რადგან მონეტა ნამდვილია, ყველა შემთხვევა თანაბარად ალბათურია და შეადგენს $\frac{1}{4}$, შესაბამისად:
  
  $ P(X = 0) = P(TT) = \frac{1}{4},$
  
  $ P(X=1) = P(TH) + P(HT) = \frac{1}{2},$
  
  $ P(X = 2) = \frac{1}{4}.$
  
  ეს ალბათობები არის **ფართობით სიხშირეები** და შესაბამისად, ჩვენ შეგვიძლია გამოვთვალოთ **მათემატიკური ლოდინი**:
  
  $$\mu = \mathbb{E}(X) = (0) \cdot \left ( \frac{1}{4} \right ) + (1) \cdot \left ( \frac{1}{2} \right ) + (2) \cdot \left ( \frac{1}{2} \right ) = 1.$$
  
  ეს შედეგი ნიშნავს, რომ ადამიანი, ვინც ააგდებს მონეტას ორჯერ ზედიზედ, საშუალოდ, მიიღებს 1 საფასურს ყოველ აგდებაზე.

- <h3>განმარტება 4.1:</h3>

  დავუშვათ $X$ არის შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$. 
  
  თუ $X$ არის **დისრკეტული** შემთხვევითი სიდიდე, მაშინ ამ სიდიდის **მათემატიკური ლოდინი** ეწოდება სიდიდეს:
  
  $$\mu = \mathbb{E}[X] = \sum_{x} x f(x).$$
 
   თუ $X$ არის **უწყვეტი** შემთხვევითი სიდიდე, მაშინ ამ სიდიდის **მათემატიკური ლოდინი** ეწოდება სიდიდეს:
  
  $$\mu = \mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x).$$


- <h3>მაგალითი 4.1:</h3>
  
  ინსპექტორი ამოწმებს **7 კომპონენტიდან** შემდგა პარტიას.
  
  პარტიაში არის **4 გამართული** და **3 გაუმართავი** კომპონენტი.
  
  ინსპექტორი შემთხვევითად ირჩევს **3 კომპონენტს** შესამოწმებლად.
  
  **იპოვნეთ გამართული კომპონენტების რაოდენობისთვის მოსალოდნელი მნიშვნელობა**. 


- <h3>მაგალითი 4.2:</h3>
  
  სამედიცინო აპარატების კომპანიის **გამყიდველს** მოცემულ დღეს **ორი შეხვედრა**აქვს. 
  
  **პირველი შეხვედრაზე** მას აქვს  $70\%$ **შანსი** გარიგების დადების, საიდანაც წარმატების შემთხვევაში $1000$ **ლარის** საკომისიოს შოვნა შეუძლია. 
  
  **მეორეს შეხვედრაზე** მას აქვს მხოლოდ $40\%$ **შანსი** გარიგების დადებაის, საიდანაც, წარმატების შემთხვევაში, $1500$ **ლარს** გამოიმუშავებს. 
  
  იმის დაშვებით, რომ **შეხვედრების შედეგი არ არის ერთმანეთზე დამოკიდებული**, რა **ჯამური საკომისიოს** უნდა **ელოდებოდეს** გამყიდველი?
 

- <h3>მაგალითი 4.3:</h3>
  
  დავუშვათ $X$ არის შემთხვევითი სიდიდე, რომელიც აღნიშნავს გარკვეული ელექტრონული მოწყობილობის ცხოვრებას დროს საათებში. 
  
  $X$ შემთხვევითი სიდიდის  ალბათობის სიმკვრივის ფუნქცია მოცემულია შემდეგი გამოსახულებით:
  
  $$f(x)=
  \left\{\begin{matrix}
  \frac{20,000}{x^3}, & x > 100, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$
  
  იპოვნეთ ამ მოწყობილობის **მოსალოდნელი** ცხოვრების დრო.

- <h3>თეორემა 4.1:</h3>
  
  თუ $X$ არის **დისრკეტული** შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$, მაშინ $g(X)$ შემთხვევითი სიდიდის **მათემატიკური ლოდინი** გამოითვლება გამოითვლება შემდეგი ფორმულით:
  
  $$\mu_{g(x)} = \mathbb{E} [ g(X)] = \sum_{x} g(x) f(x).$$
  
  თუ $X$ არის **უწყვეტი** შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$, მაშინ $g(X)$ შემთხვევითი სიდიდის **მათემატიკური ლოდინი** გამოითვლება გამოითვლება შემდეგი ფორმულით:
    
  $$\mu_{g(x)} = \mathbb{E} [ g(X)] = \int_{-\infty}^{\infty} g(x) f(x) dx.$$
  


- <h3>მაგალითი 4.4:</h3>
  
  დავუშვათ, რომ $X$ მანქანების რაოდენობა, რომლებიც გადიან ავტოსამრეცხაოს 16:00 საათიდან 17:00 საათამდე ნებისმიერ მზიან პარასკევს, წარმოდგენილია შემდეგი განაწილების ფუნქციით:


  |     $x$    |        4       |        5       |       6       |       7       |       8       |       9       |
  |:----------:|:--------------:|:--------------:|:-------------:|:-------------:|:-------------:|:-------------:|
  | $P(X = x)$ | $\frac{1}{12}$ | $\frac{1}{12}$ | $\frac{1}{4}$ | $\frac{1}{4}$ | $\frac{1}{6}$ | $\frac{1}{6}$ |
  
    დავუშვათ, g(X) = 2X - 1 წარმოადგენს თანხის ოდენობას (ლარებში) რომელიც გადახდილია მოსამსახურე პერსონალისთვის მენეჯერის მიერ. 
  
    იპოვნეთ მომსახურე პერსონალის მოსალოდნელი მოგება ამ კონკრეტული პერიოდისთვის.
  

- <h3>მაგალითი 4.5:</h3>
  
  დავუშვათ, რომ $X$ არის შემთხვევითი სიდიდე შემდეგი განაწილების ფუნქციით:

  $$f(x)=
  \left\{\begin{matrix}
  \frac{x^2}{3}, & -1 < x < 2, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$
  
  იპოვნეთ მოსალოდნელი მნიშვნელობა $g(X) = 4X + 3$ ფუნქციისთვის.

- <h3>განმარტება 4.2:</h3>

  თუ $X$ და $Y$ არის **დისკრეტული** შემთხვევითი სიდიდეები ერთობლივი განაწილების ფუნქციით $f(x,y)$, მაშინ $g(X, Y)$ შემთხვევითი სიდიდის **მათემატიკური ლოდინი** გამოითვლება ფორმულით:
  
  $$\mu_{g(X,Y)} = \mathbb{E}[g(X,Y)] = \sum_{x}\sum_{y}g(x,y) f(x,y).$$
 
  თუ $X$ და $Y$ არის **უწყვეტი** შემთხვევითი სიდიდეები ერთობლივი განაწილების ფუნქციით $f(x,y)$, მაშინ $g(X, Y)$ შემთხვევითი სიდიდის **მათემატიკური ლოდინი** გამოითვლება ფორმულით:
  
  $$\mu_{g(X,Y)} = \mathbb{E}[g(X,Y)] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} g(x,y) f(x,y) dx dy.$$

- <h3>მაგალითი 4.7:</h3>
  
  დავუშვათ $X$ და $Y$ არის შემთხვევითი სიდიდეები შემდეგი ერთობლივი განაწილების ფუნქციით $f(x,y)$:

  \end{matrix}\right.$$
  
  იპოვნეთ მოსალოდნელი მნიშვნელობა $g(X) = Y/X$ ფუნქციისთვის.

<h3 align="center">შემთვევითი სიდიდის ვარიაცია და კოვარიაცია</h3>

- $X$ შემთხვევითი სიდიდის **ცვალებადობის** ყველაზე მნიშვნელოვანი **საზომი** მიიღება **4.1 თეორემიში** $g (X) = (X - \mu)^2$ ფუნქციის გამოყენებით.


- <h3>განმარტება 4.3:</h3>

  დავუშვათ $X$ არის შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$. 

  თუ $X$ არის **დისრკეტული** შემთხვევითი სიდიდე, მაშინ ამ სიდიდის **ვარიაცია (ან დისპერსია)** ეწოდება სიდიდეს:
  
  $$\sigma^2 = \mathbb{E}[(X-\mu)^2] = \sum_{x} (x-\mu)^2 f(x).$$
 
   თუ $X$ არის **უწყვეტი** შემთხვევითი სიდიდე, მაშინ ამ სიდიდის **ვარიაცია (ან დისპერსია)** ეწოდება სიდიდეს:
  
  $$\sigma^2 = \mathbb{E}[(X-\mu)^2] = \int_{-\infty}^{\infty} (x-\mu)^2 f(x) dx.$$


- <h3>მაგალითი 4.8:</h3>
  
  დავუშვათ $X$ არის **ავტომობილების რაოდენობას**, რომლებიც გამოიყენება ოფიციალური საქმიანი მიზნებისათვის მოცემულ სამუშაო დღეს.
  
  ალბათური განაწილება $A$ კომპანიისთვის მოცემულია ცხრილით:
  
  
  |   $x$  |  0  |  1  |  2  |
  |:------:|:---:|:---:|:---:|
  | $f(x)$ | 0.3 | 0.4 | 0.3 |
  
    ხოლო, ალბათური განაწილება $B$ კომპანიისთვის მოცემულია ცხრილით:
    
  |   $x$  |  0  |  1  |  2  |  3  |  4  |
  |:------:|:---:|:---:|:---:|:---:|:---:|
  | $f(x)$ | 0.2 | 0.1 | 0.3 | 0.3 | 0.1 |
  
  
    აჩვენეთ, რომ ალბათური განაწილების ვარიაცია B კომპანიის უფრო დიდია ვიდრე A კომპანიისთვის.
  

- <h3>თეორემა 4.2:</h3>
  
  თუ $X$ შემთხვევითი სიდიდის **ვარიაცია** გამოითვლება ფორმულით:
  
  $$\sigma^2 = \mathbb{E}[X^2] - \mu^2.$$


- <h3>მაგალითი 4.9:</h3>
  
  დავუშვათ $X$ შემთხვევითი სიდიდეები წარმოადგენდეს დანადგარის დეფექტური ნაწილების რაოდენობას, როდესაც საწარმოო ხაზიდან 3 ნაწილის შერჩევა და ტესტირება ხდება. 
  
  ქვემოთ მოცემულია $X$ შემთხვევითი სიდიდეები ალბათური განაწილების ფუნქცია $f(x,y)$:

|   $x$  |  0  |  1  |  2  |  3  |
|:------:|:---:|:---:|:---:|:---:|
| $f(x)$ | 0.51 | 0.38 | 0.10 | 0.01 |
  
&emsp;&emsp; იპოვნეთ $X$ შემთხევითი სიდიდის ვარიაცია **4.2 თეორემის გამოყენებით**.

- <h3>მაგალითი 4.10:</h3>
  
  ყოველკვირეული მოთხოვნა სასმელ წყლზე (ათასობით ლიტრში), **gwp** კომპანიის ადგილობრივი ქსელიდან არის უწყვეტი შემთხვევითი ცვლადი $X$, რომელსაც აქვს შემდეგი ალბათური განაწილების სიმკვრივე:
  
  $$f(x)=
  \left\{\begin{matrix}
  2(x-1), & 1 < x < 2,\\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$
  
  იპოვნეთ $X$ შემთხვევითი სიდიდის **ლოდინი** და **ვარიაცია**.

- <h3>თეორემა 4.2:</h3>
  
  თუ $X$ არის **დისკრეტუყლი** შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$, მაშინ $g(X)$ შემთხვევითი სიდიდის ვარიაცია გამოითვლება ფორმულით:
  
  $$\sigma^2_{g(X)} = \mathbb{E}  \left [ g(X) - \mu_{g(X)}^2  \right [ = \sum_{x} \left [ g(x) - \mu_{g(X)} \right ]^2 f(x).$$
  
  თუ $X$ არის **უწყვეტი** შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$, მაშინ $g(X)$ შემთხვევითი სიდიდის ვარიაცია გამოითვლება ფორმულით:
  
  $$\sigma^2_{g(X)} = \mathbb{E} \left [ g(X) - \mu_{g(X)}^2  \right [ = \int_{-\infty}^{\infty} \left [ g(x) - \mu_{g(X)} \right ]^2 f(x) dx.$$


- <h3>მაგალითი 4.11:</h3>
  
  დავუშვათ $X$ არის შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$:
  
  
  |   $x$  |  0  |  1  |  2  |  3  |
  |:------:|:---:|:---:|:---:|:---:|
  | $f(x)$ | $\frac{1}{4}$ | $\frac{1}{8}$ | $\frac{1}{2}$ | $\frac{1}{8}$ | 
  

&emsp;&emsp; იპოვნეთ $g(X) = 2X + 3$ შემთხევითი სიდიდის **ვარიაცია**.

- <h3>მაგალითი 4.12:</h3>
  
  დავუშვათ $X$ არის შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$ 
  
  $$f(x)=
  \left\{\begin{matrix}
  \frac{x^2}{3}, & -1 < x < 2, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right.$$
  
  იპოვნეთ $g(X) = $X + 3$ შემთხევითი სიდიდის **ვარიაცია**.

- <h3>განმარტება 4.4:</h3>

  თუ $X$ და $Y$ არის **დისკრეტული** შემთხვევითი სიდიდეები ერთობლივი განაწილების ფუნქციით $f(x,y)$, მაშინ $g(X, Y)$ შემთხვევითი სიდიდის **კოვარიაცია** ეწოდება სიდიდეს:
  
  $$\sigma_{XY} = \mathbb{E} \left [(X - \mu_X)(Y - \mu_Y) \right] = \sum_{x}\sum_{y}(X - \mu_X)(Y - \mu_Y) f(x,y).$$
 
  თუ $X$ და $Y$ არის **უწყვეტი** შემთხვევითი სიდიდეები ერთობლივი განაწილების ფუნქციით $f(x,y)$, მაშინ $g(X, Y)$ შემთხვევითი სიდიდის **კოვარიაცია** ეწოდება სიდიდეს:
  
  $$\sigma_{XY} = \mathbb{E} \left [(X - \mu_X)(Y - \mu_Y) \right] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} (X - \mu_X)(Y - \mu_Y) f(x,y) dx dy.$$


- <h3>თეორემა 4.4:</h3>
  
  თუ $X$ და $Y$ არის ორი შემთხვევითი სიდიდე $\mu_X$ და $\mu_Y$ **ლოდინით**, შესაბამისად, მაშინ **კოვარიაცია** გამოითვლება შემდეგი ფორმულით:
 
  $$\sigma_{XY} = \mathbb{E} [XY] - \mu_X \mu_Y$$

  იპოვნეთ $X$ და $Y$ შემთხვევითი სიდიდეების **კოვარიაცია**.

- <h3>მაგალითი 4.14:</h3>
  
  მარათონში მონაწილე კაცების რაოდენობა $X$ და ქალი რაოდენობა $Y$ აღწერილია ერთობლივი სიმკვრივის ფუნქციით:

  იპოვნეთ $X$ და $Y$ შემთხვევითი სიდიდეების **კოვარიაცია**.

- <h3>განმარტება 4.5:</h3>

  თუ $X$ და $Y$ არის შემთხვევითი სიდიდეები **კოვარიაციით** $\sigma{XY}$ და **ვარიაციით** $\sigma_X$ და $\sigma_Y$, შესაბამისად.
  
  $X$ და $Y$ არის შემთხვევითი სიდიდეების კორელაციის კოეფიციენტი ეწოდება სიდიდეს:
  
  $$\rho_{XY} = \frac{\sigma_{XY}}{\sigma_X \sigma_Y}.$$
