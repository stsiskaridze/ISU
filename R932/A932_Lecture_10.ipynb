<h3 align="center">შემთხვევითი სიდიდეების გარდაქმნები</h3>

- <h3>თეორემა 10.1:</h3>
  
  დავუშვათ, რომ $X$ არის დისკრეტული შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$.
    
  თუ $Y = u (X)$ წარმოადგენს ურთიერთ ცალსახა გარდაქმნას $X$ და $Y$ მნიშვნელობებს შორის, მაშინ განტოლება $y = u (x)$ შეიძლება ცალსახად ამოიხსნას $x$-ის მიმართ, ვთქვათ $x = w (y)$ და $Y$ ალბათური განაწილება ჩაიწერება შემდეგი სახით: 
  
  $$g (y) = f [w (y)].$$

- <h3>მაგალითი 10.1:</h3>
  
  მოცემულია $X$ დისკრეტული შემთხევით სიდიდე გეომეტრიული ალბათური განაწილებით:
  
  $$f(x) = \frac{3}{4} \left ( \frac{1}{4}\right )^{x-1},  \text{ } x = 1, 2, 3, ... .$$
  
  იპოვნეთ $Y = X^2$ შემთხვევითი სიდიდის ალბათური განაწილება.

- <h3>თეორემა 10.2:</h3>
  
  დავუშვათ, რომ $X_1$ და $X_2$ არის დისკრეტული შემთხვევითი სიდიდეები ერთობლივი ალბათური განაწილებით $f(x_1, x_2)$.
    
  თუ $Y_1 = u_1 (X_1, X_2)$ და $Y_2 = u_2 (X_1, X_2)$  წარმოადგენს ურთიერთ ცალსახა გარდაქმნებს $(x_1, x_2)$ და $(y_1, y_2)$ მნიშვნელობებს შორის, მაშინ განტოლებათა სისტემა:
  
  $$\left\{\begin{matrix}
  y_1 = u_1(x_1, x_2)\\ 
  y_2 = u_2(x_1, x_2) 
  \end{matrix}\right.$$
  
  შეიძლება ცალსახად ამოიხსნას $x_1$-ის და $x_2$-ის მიმართ, ვთქვათ $x_1 = w_1 (y_1, y_2)$ და $x_2 = w_2 (y_1, y_2)$, ხოლო $Y_1$ და $Y_2$ შემთხვევითი სიდიდეების ერთობლივი ალბათური განაწილება ჩაიწერება შემდეგი სახით: 
  
  $$g (y_1, y_2) = f [w_1 (y_1, y_2), w_2 (y_1, y_2)].$$

- <h3>მაგალითი 10.2:</h3>
  
  მოცემულია $X_1$ და $X_2$ ორი დამოუკიდებელი შემთხვევითი სიდიდე **პუასონის** განაწილებით $\mu_1$ და $\mu_2$ პარამეტრებით, შესაბამისად. 
  
  იპოვნეთ $Y_1 = X_1 + X_2$ შემთხვევითი ცვლადის განაწილება.

- <h3>თეორემა 10.3:</h3>
  
  დავუშვათ, რომ $X$ რის უწყვეტი შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$.
  
  თუ $Y = u (X)$ წარმოადგენს ურთიერთ ცალსახა გარდაქმნას $X$ და $Y$ მნიშვნელობებს შორის, მაშინ განტოლება $y = u (x)$ შეიძლება ცალსახად ამოიხსნას $x$-ის მიმართ, ვთქვათ $x = w (y)$ და $Y$ ალბათური განაწილება ჩაიწერება შემდეგი სახით: 
  
  $$g (y) = f [w (y)] |J|,$$
  
  სადაც $J$ წარმოადგენს ამ გარდაქმნის **იაკობიანს** (**Jacobian**):
  
  $$J = \frac{dx}{dy}.$$

- <h3>მაგალითი 10.3:</h3>
  
  მოცემულია $X$ უწყვეტი შემთხევით სიდიდე გეომეტრიული ალბათური განაწილებით:
  
  $$f(x) = 
    \left\{\begin{matrix}
   \frac{x}{12}, & 1 < x < 5,  \\ 
   0,& \text{ საწინააღმდეგო შემთხვევაში.}  
   \end{matrix}\right.
   $$
  
  იპოვნეთ $Y = 2X - 3$ შემთხვევითი სიდიდის ალბათური განაწილება.

- <h3>თეორემა 10.4:</h3>
  
  დავუშვათ, რომ $X_1$ და $X_2$ არის უწყვეტი შემთხვევითი სიდიდეები ერთობლივი ალბათური განაწილებით $f(x_1, x_2)$.
  
  თუ $Y_1 = u_1 (X_1, X_2)$ და $Y_2 = u_2 (X_1, X_2)$  წარმოადგენს ურთიერთ ცალსახა გარდაქმნებს $(x_1, x_2)$ და $(y_1, y_2)$ მნიშვნელობებს შორის, მაშინ განტოლებათა სისტემა:
  
  $$\left\{\begin{matrix}
  y_1 = u_1(x_1, x_2)\\ 
  y_2 = u_2(x_1, x_2) 
  \end{matrix}\right.$$
  
  შეიძლება ცალსახად ამოიხსნას $x_1$-ის და $x_2$-ის მიმართ, ვთქვათ $x_1 = w_1 (y_1, y_2)$ და $x_2 = w_2 (y_1, y_2)$, ხოლო $Y_1$ და $Y_2$ შემთხვევითი სიდიდეების ერთობლივი ალბათური განაწილება ჩაიწერება შემდეგი სახით:
  
  $$g (y_1, y_2) = f [w_1 (y_1, y_2), w_2 (y_1, y_2)] |J|,$$
  
  სადაც $J$ წარმოადგენს ამ გარდაქმნის **იაკობიანს** (**Jacobian**):
  
  $$J = 
  \begin{vmatrix}
  \frac{\partial x_1}{\partial y_1} & \frac{\partial x_1}{\partial y_2} \\ 
  \frac{\partial x_2}{\partial y_1} & \frac{\partial x_2}{\partial y_2} \\ 
  \end{vmatrix}.
  $$


- <h3>მაგალითი 10.4:</h3>
  
  მოცემულია $X_1$ და $X_2$ უწყვეტი შემთხევით სიდიდეები შემდეგი ერთობლივი ალბათური განაწილებით:
  
  $$f(x_1, x_2) = 
    \left\{\begin{matrix}
   \frac{4 x_1 x_2}{12}, & 0 < x_1 < 1 , 0 < x_2 < 1,  \\ 
   0,& \text{ საწინააღმდეგო შემთხვევაში.}  
   \end{matrix}\right.
   $$
  
  იპოვნეთ $Y_1 = X_1^2$ და $Y_1 = X_1 X_2$ შემთხვევითი სიდიდეების ერთობლივი ალბათური განაწილება.

- <h3>თეორემა 10.5:</h3>
  
  დავუშვათ, რომ $X$ რის უწყვეტი შემთხვევითი სიდიდე ალბათური განაწილებით $f(x)$.
  
  ხოლო $Y = u (X)$ წარმოადგენს გარდაქმნას $X$ და $Y$ შორის რომელიც **არ არის ურთიერთ ცალსახა გარდაქმნა**.
  
  თუ $X$-ის განსაზღვირს ინტერვალი შეიძლება დაიყოს $k$-ცალ ურთიერთ უკვეთ იმრავლეებად ისე, რომ თითოეულ ინტერვალზე:
  
  $$x_1 = w_1(y), \text{ } x_2 = w_2(y), ... , \text{ } x_k = w_k(y), \text{ },$$
  
  $y  = u(x)$ არის ურთიერთცალსახა, მაშინ $Y$ შემთხვევითი სიდიდის ალბათური განაწილება ჩაიწერება შემდეგი სახით:
  
  $$g(y) = \sum_{i=1}^{k} f[w_i(y)] |J_i|$$
  
  სადაც $J_i$ წარმოადგენს ამ გარდაქმნების **იაკობიანს** (**Jacobian**):
  
  $$J_i = \frac{dx_i}{dy}, \text{ } i = 1, 2, ..., k.$$

- <h3>მაგალითი 10.5:</h3>
  
  თუ $X$ არის შემთხვევითი სიდიდე **ნორმალური** ალბათური განაწილებით $\mu$ საშუალოთი და ვარიაციით $\sigma^2$, მაშინ $Y = (X − \mu)^2 / \sigma^2$-ს აქვს $\chi^2$-ის განაწილება **ერთეულოვანი თავისუფლებით ხარისხით**.

- <h3>მაგალითი 10.6:</h3>
  
  იპოვნეთ $X$ **ბინომიალური** შემთხვევითი სიდიდის **მომენტის გენერაციის ფუნქცია** და შემდეგ მის გამოიყენეთ აჩვენეთ, რომ $\mu = np$ და $\sigma^2 = npq$.

- <h3>თეორემა 10.10:</h3>
  
  თუ $X_1, X_2, ..., X_n$ არიან დამოუკიდებელი შემთხვევითი სიდიდეები, შესაბამისად, $M_{X_1}(t), M_{X_2}(t), ..., M_{X_n}(t)$ **მომენტის გენერაციის ფუნქციებით**, და $Y = X_1 + X_2 + \cdots + X_n$, მაშინ:
  
  $$M_Y(t) = \prod_{i=1}^{n} M_{X_i}(t).$$

<h3 align="center">შემთხვევითი სიდიდეების წრფივი კომბინაცია</h3>

- <h3>თეორემა 10.11:</h3>
  
  თუ $X_1, X_2, ..., X_n$ არიან დამოუკიდებელი შემთხვევითი სიდიდეები **ნორმალური** განაწილებით, შესაბამისად, $\mu_1, \mu_2, ..., \mu_n$ საშუალოებით და $\sigma_1^2, \sigma_2^2, ..., \sigma_n^2$ ვარიაცეიბით, მაშინ
  
  $$Y = a_1 X_1 + a_2 X_2 + \cdots + a_n X_n$$
  
  შემთხვევითი სიდიდე იქნება **ნორმალურად** განაწილებული **საშუალოთი**:
  
  $$\mu_Y = a_1 \mu_1 + a_2 \mu_2 + \cdots + a_n \mu_n.$$
  
  და **ვარიაციით**:
  
  $$\sigma_Y^2 = a_1^2 \sigma_1^2 + a_2^2 \sigma_2^2 + \cdots + a_n^2 \sigma_n^2.$$
  

- <h3>თეორემა 10.12:</h3>
  
  თუ $X_1, X_2, ..., X_n$ არიან ურთიერთ დამოუკიდებელი შემთხვევითი სიდიდეები $\chi^2$ **განაწილებით**, შესაბამისად, $v_1, v_2, ..., v_n$ თავისუფლების ხარისხით, მაშინ $Y = X_1 + X_2 + \cdots + X_n$ შემთხვევითი სიდიდეს ექნება $\chi^2$ **განაწილება** თავისუფლების ხარისხით $v = v_1 + v_2 + \cdots + v_n$.

- <h3>დასკვნა 10.1:</h3>
  
  თუ $X_1, X_2, ..., X_n$ არიან დამოუკიდებელი შემთხვევითი სიდიდეები **ნორმალური** განაწილებით, შესაბამისად, $\mu$ საშუალოთი და $\sigma^2$ ვარიაციით, მაშინ შემთხვევითი სიდიდეს:
  
  $$Y = \sum_{i=1}^{n} \left ( \frac{X_i - \mu}{\sigma}\right )^2$$
  
  ექმება $\chi^2$ **განაწილება** თავისუფლების ხარისხით $v = n$.

- <h3>დასკვნა 10.2:</h3>
  
  თუ $X_1, X_2, ..., X_n$ არიან დამოუკიდებელი შემთხვევითი სიდიდეები **ნორმალური** განაწილებით, შესაბამისად, $\mu_1, \mu_2, ..., \mu_n$ საშუალოებით და $\sigma_1^2, \sigma_2^2, ..., \sigma_n^2$ ვარიაცეიბით, მაშინ შემთხვევითი სიდიდეს:
  
  $$Y = \sum_{i=1}^{n} \left ( \frac{X_i - \mu_i}{\sigma_i}\right )^2$$
  
  ექმება $\chi^2$ **განაწილება** თავისუფლების ხარისხით $v = n$.
