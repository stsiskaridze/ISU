<h3 align="center">ერთობლივი ალბათური განაწილება</h3>

- აქამდე შემთხვევითი სიდიდეები და მათი ალბათური განაწილები შემოფარგლული იყო **ერთგანზომილებიანი** ხდომილებათა სივრცეებით.


- ჩვენ ახლა გადავალთ **მრავალგანზიმილებიან** ხდომილებათა სივრცეების შესწავლაზე.


- **განმარტება 3.1**: 
  
  ორი ცვლადის $f(x, y)$ ფუნქცის ეწოდება $X$ და $Y$ **დისკრეტული** შემთხვევითი სიდიდეების **ერთობლივი ალბათური განაწილება** ან **ალბათური მასის ფუნქცია**, თუ:
  
  1. $f(x,y) \geq 0$ ყველა $(x,y)$ წყვილისთვის,
  2. $\sum_{x}\sum_{y}f(x,y) = 1$,
  3. $P(X = x, Y = y)=f(x,y)$,
  4. $P[(X,Y) \in A] = \sum_{(x,y)\in A}f(x,y)$ ნებისმიერი $A$ სიმრავლისთვის $xy$-სიბრტყეზე.

- **მაგალითი 3.1**: 
  
  ყუთიდან, რომელშც დევს 3 წითელი, 2 ყვითელი და 3 მწვანე ბურთები, ჩაუხედავად ვიღებთ 2 ბურთს.
  
  თუ $X$ არის ამოღებული წითელ ბურთთა რაოდენობა, ხოლო $Y$ არის ამოღებულიწვითელ ბურთთა რაოდენობა, მაშინ იმპოვნეთ:
  
  1. ერთობლივი ალბათური განაწილება $f(x,y)$.
  2. $P[(X,Y) \in A]$, სადაც $A = \{(x,y) | x+y \leq 1\}$.

- **განმარტება 3.2**: 
  
  ორი ცვლადის $f(x, y)$ ფუნქცის ეწოდება $X$ და $Y$ უწყვეტი შემთხვევითი სიდიდეების **ერთობლივი ალბათური განაწილება** ან **ალბათური მასის ფუნქცია**, თუ:
  
  1. $f(x,y)\geq 0$ ყველა $(x,y)$ წყვილისთვის,
  2. $\int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} f(x,y) dx dy = 1$,
  3. $P[(X,Y) \in A] = \int_{(x,y)\in A}f(x,y)dx dy$ ნებისმიერი $A$ სიმრავლისთვის $xy$-სიბრტყეზე.
  

- **მაგალითი 3.2**: 
  
  დავუშვათ მოცემულია ორი შემთხვევითი სიდიდე $X$ და $Y$, და მათი ერთობლივი ალბათური განაწილების ფუნქცია:
  
  $$f(x,y) = 
  \left\{\begin{matrix}
    \frac{2}{5}(2x+3y), &  \text{ ყველა } 0 \leq x \leq 1 \text{ და} 0 \leq y \leq 1, \\
    0, & \text{ საწინააღმდეგო შემთხვევაში.}
  \end{matrix}\right.$$
  1. შეამოწმეთ ერთობლივი ალბათური განაწილების **მეორე პირობა**.
  2. იპოვეთ $P[(X,Y) \in A]$, სადაც $A = \left \{(x,y) | 0 < x < \frac{1}{2}, \frac{1}{4} < y < \frac{1}{2} \right \}$.

- **განმარტება 3.3**: 
  
  $X$ და $Y$ **დისკრეტული** შემთხვევითი სიდიდეების **ზღვრული განაწილება** ეწოდება ფუნქციებს:
  
  $$g(x) = \sum_{y} f(x,y) \text{ და } h(y)=\sum_{x}f(x,y).$$
    
  $X$ და $Y$ **უწყვეტი** შემთხვევითი სიდიდეების **ზღვრული განაწილება** ეწოდება ფუნქციებს:
  
  $$g(x) = \int_{-\infty}^{+\infty} f(x,y) dy \text{ და } h(y)=\int_{-\infty}^{+\infty}f(x,y) dx.$$

- **მაგალითი 3.3**: 
  
  **3.1 ამოცანაში** აგებული ცხრილისთვის აჩვენეთ, რომ **ჯამური სტრიქონი** და **ჯამური სვეტი** წარმოადგენენ $X$ და $Y$ დისკრეტული შემთხვევითი სიდიდეების ზღვრულ განაწილებებს.

- **მაგალითი 3.4**: 
  
  **3.2 ამოცანაში** მოყვანილი ერტობლივი ალბათური ფუნქციისთვის იპოვეთ ზღვრული განაწილებებები.

- **განმარტება 3.4**: 
  
  დავუშვათ $X$ და $Y$ არის ორი, **დისკრეტული** ან **უწყვეტი**,  შემთხვევითი სიდიდეები.
  
  $Y$ შემთხვევითი სიდიდის **პირობითი განაწილება** იმის გათვალისწინებით, რომ $X = x$, არის:
  
  $$f(y | x) = \frac{f(x,y)}{g(x)}, \text{ სადაც იხულისხმება, რომ } g(x) > 0.$$
  
  შესაბამისად,  $X$ შემთხვევითი სიდიდის **პირობითი განაწილება** იმის გათვალისწინებით, რომ $Y = y$, არის:
  
  $$f(x | y) = \frac{f(x,y)}{h(y)}, \text{  სადაც იხულისხმება, რომ } h(y) > 0.$$
 

- მაგალითად, თუ ჩვენ გვსურს ვიპოვოთ ალბათობა, რომ დისკრეტული შემთხვევითი ცვლადი $X$ მოხვდება $[a,b]$ ინტერვალში, იმის გათვალისწინებით, რომ დისკრეტული ცვლადი $Y = y$, ჩვენ გამოვიყენებთ შესაბამის პირობით განაწილებას:

  $$P(a < X < b | Y = y) = \int_{a}^{b} f(x|y) dx.$$

- **მაგალითი 3.5**: 
  
  **3.1 ამოცანაში** მოყვანილი $X$ შემთხვევითი სიდიდესთვის იპოვეთ პირობითი განაწილება, იმის გათვალისწინებით, რომ $Y = 1$, და შედეგის გამოყენებით,  იპოვეთ $P(X = 0| Y = 1)$.

- **მაგალითი 3.6**: 
  
  მოცემულია ერთობლივი სიმკვრივის ფუნქციის:
  
  $$f(x,y) = 
  \left\{\begin{matrix}
  \frac{x(1+3y^2)}{0}, & 0 < x < 2, 0 < y < 1.\\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right. $$
  
  იპოვეთ $g(x)$, $h(x,y)$ და $f(x | y)$, და გამოთვალეთ:
  
  $$P \left (\frac{1}{4} < X < \frac{1}{2} | Y = \frac{1}{3} \right ).$$

<h3 align="center">სტატისტიკური დამოუკიდებლობა</h3>


- თუ $f (x | y)$ **არ არის დამოკიდებული** $y$- ზე, როგორც ეს იყო **3.6 მაგალითში**, მაშინ $f (x | y) = g (x)$
და $f (x, y) = g (x) h (y)$.

- **განმარტება 3.5**: 
  
  დავუშვათ $X$ და $Y$ არის ორი, **დისკრეტული** ან **უწყვეტი**,  შემთხვევითი სიდიდეები $f (x, y)$ ერთობლივი ალბათური  განაწილებით და $g (x)$ და $h (y)$ ზღვრული განაწილებებით.
  
 ამბობენ, რომ $X$ და $Y$ შემთხვევითი სიდიდეები **სტატისტიკურად დამოუკიდებლები** არიან, მაშინ და მხოლოდ მაშინ, თუ:
  
  $$f(x, y ) = g(x) h(y)$$
  
  ყველა $(x,y)$ წყვილისთვის.

- **მაგალითი 3.7**: 
  
  აჩვენეთ, რომ **3.1 ამოცანაში** მოყვანილი $X$ და $Y$ შემთხვევითი სიდიდეები **არ არიან** სტატისტიკურად დამოუკიდებლები.

- ყველაფერი, რაც აქამდე განვიხილეთ, ეხებოდა მხოლოდ ორ შემთხვევით სიდიდეს.

- იგივე შედეგები შეიძლება იქნას მიღებული $n$ შემთხვევით სიდიდეებისთვის.

- **განმარტება 3.8**: 
  
  დავუშვათ $X_1, X_2, ..., X_n$ არის $n$, **დისკრეტული** ან **უწყვეტი**, შემთხვევითი სიდიდეები $f (x_1, x_2, ..., x_n)$ ერთობლივი ალბათური განაწილებით და $f_1 (x), f_2(x), ..., f_n(x)$ ზღვრული განაწილებით.
  
  ამბობენ, რომ $X_1, X_2, ..., X_m$ შემთხვევითი სიდიდეები **სტატისტიკურად დამოუკიდებლები** არიან, მაშინ და მხოლოდ მაშინ, თუ:
   
  $$f(x_1, x_2, ..., x_n) = f_1(x) f_2(y) \cdots f_n(x)$$
  
  ყველა $(x_1, x_2, ..., x_n)$ ერთობლიობისთვის.

- **მაგალითი 3.9**: 
  
  დავუშვათ, რომ მუყაოს ყუთებში შეფუთული მალფუჭებადი საკვები პროდუქტის შენახვის ვადა არის შემთხვევითი სიდიდე, რომლის ალბათური სიმკვრივის ფუნქცია მოცემულია შემდეგი სახით:
  
  $$f(x) = 
  \left\{\begin{matrix}
  e^{-x},  & x > 0,\\
  0, & \text{ საწინააღმდეგო შემთხვევაში}.
  \end{matrix}\right. $$
  
  დავუშვათ,  $X_1$, $X_2$ და $X_3$ შემთხვევითი სიდიდეები წარმოადგენენ სამი დამოუკიდებლად შერჩეული  კონტეინერების შენახვის ვადებს. 
  
  იპოვნეთ $P (X_1<2, 1 <X_2<3, X_3> 2)$.

```python
import numpy as np

P = (1 - np.exp(-2)) * (np.exp(-1) - np.exp(-3)) * np.exp(-2)
round(P, 4)
```
