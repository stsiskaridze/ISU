<h3 align="center">ამოცანა №1: დისკრეტული შემთხვევითი სიდიდეების გარდაქმნა</h3>

მოცემულია $X_1$ და $X_2$ **დისკრეტული** შემთხევითი სიდიდეები ერთობლივი ალბათური განაწილებით:
  
  $$f(x_1, x_2) = 
    \left\{\begin{matrix}
   \frac{x_1 x_2}{18}, & x_1 = \{1, 2\} \text{ და } x_2 = \{1, 2, 3\},  \\ 
   0,& \text{ საწინააღმდეგო შემთხვევაში.}  
   \end{matrix}\right.
   $$
  
იპოვნეთ $Y = X_1 X_2$ შემთხვევითი სიდიდის ალბათური განაწილება.

<h3 align="center">ამოცანა №2: მომენტის გენერაციის ფუნქცია</h3>

მოცემულია $X$ დისკრეტული შემთხევით სიდიდე შემდეგი ალბათური განაწილებით:

  $$f(x) = 
   \left\{\begin{matrix}
   \frac{1}{4} & x = -2,  \\ 
   \frac{1}{2} & x = 0,  \\ 
   \frac{1}{4} & x = +2,  \\ 
   0,& \text{ სხვა შემთხვევაში.}  
   \end{matrix}\right.
  $$

იპოვნეთ $X$ შემთხვევითი სიდიდის **მომენტის გენერაციის ფუნქცია** და მის საშუალებით გამოთვალეთ **მათემატიკური ლოდინი** და **ვარიაცია**.

<h3 align="center">ამოცანა №3: ნიმუშის ადგილმდებარეობა და ცვალებადობა</h3>

მოცემულია ნიმუში რომელიც შედგება **30** დაკვირვებებისგან:
  
  $$5, 7, 7, 9, 10, 5, 9, 4, 8, 8, 5, 6, 9, 7, 8, 6, 8, 8, 8, 9, 6, 10, 5, 3, 6, 7, 6, 7, 8, 7$$

<h3 align="center">ამოცანა №4: ნიმუშის საშუალოს შეფასება, როდესაც $\sigma^2$ ცნობილია</h3>

უნივერსიტეტის კურსდამთავრებულთაგან აირჩიეს შემთხვევით 12 სტუდენტი და შეისწავლეს მათი სიტყვების აკრეფის სიჩქარე.

ცნობილია, რომ სტუდენტები საშუალოდ წუთში კრეფენ $79.3$ სიტყვას, სტანდარტული გადახრით $7.8$ სიტყვა წუთში.

იპოვნეთ $95\%$ ნდობის ინტერვალი ამ უნივერსიტეტის ყველა კურსდამთავრებულის მიერ აკრეფილი სიტყვების საშუალო რაოდენობაზე.

ჩათვალეთ, რომ წუთში აკრეფილი სიტყვების რაოდენობა არის ნორმალურად განაწილებული.


<h3 align="center">ამოცანა №5: ნიმუშის ზომის შეფასება</h3>

დანადგარის ექსპერტს სურს დაადგინოს გარკვეული ლითონის ნაჭერის გასაბურღად საჭირო **საშუალო დრო**. 

რა ზომის ნიმუში დასჭირდება მას, რომ $95\%$ დარწმუნებული იყოს, რომ მის მიერ გამოთვლილი საშუალო მაჩვენებელი ნამდვილი საშუალო მაჩვენებელისგან განსხვავდებოდეს არაუმეტეს $15$ წამისა? 

ჩათვალეთ, რომ წინა კვლევებიდან ცნობილია, რომ $\sigma = 40$ წამი.

<h3 align="center">ამოცანა №6: ნიმუშის საშუალოს შეფასება, როდესაც $\sigma^2$ უცნობია</h3>

გოგირდმჟავას $9$ მსგავსი კონტეინერის შემცველობაა $9.8$, $10.0$, $10.2$, $10.4$, $10.2$, $9.8$, $10.0$, $10.2$ და $9.6$ ლიტრი.
