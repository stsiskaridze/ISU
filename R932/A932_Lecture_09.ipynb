- მიუხედავად იმისა, რომ **ნორმალური განაწილება** შეიძლება გამოყენებულ იქნას ინჟინერიასა და მეცნიერებაში მრავალი პრობლემის გადასაჭრელად, არსებობს კიდევ  უამრავი სიტუაცია, რომლებიც საჭიროებს სხვა ტიპის სიმკვრივის ფუნქციებს. 


- ორი ასეთი სიმკვრივის ფუნქციაა **გამა** და **ექსპონენციალური** განაწილების ფუნქციები.


- ექსპონენციალური და გამა განაწილება მნიშვნელოვან როლს ასრულებს როგორც **რიგების თეორიაში**, ასევე **საიმედოობის პრობლემებში**. 


- მაგალითად, **დრო სამსახურში მისვლამდე**, და **კომპონენტის ნაწილების და ელექტრო სისტემების წყობიდან გამოსვლის დრო** აღიწერება ექსპონენციალური განაწილებით.


- როგორც ჩვენ ვნახავთ, **ექსპონენციალური** განაწილება წარმოადგენს **გამა** განაწილების **განსაკუთრებული შემთხვევას**.

- <h3>განმარტება 9.1:</h3>
 
  **გამა ფუნქცია** ეწოდება შემდეგ ფუნქციას:
  
  $$\Gamma(\alpha) = \int_{0}^{\infty} x^{\alpha-1} e^{-x} dx, \text{ სადაც } \alpha > 0.$$

- <h3>თვისებები:</h3>

  გამა ფუნქციის თვისებებია:
  
  ა) $\Gamma(n) = (n-1)(n-2) \cdots (1) \Gamma(1)$ ყველა დადებითი მთელი $n$ რიცხვისთვის;
  
  ბ) $\Gamma(n) = (n-1)!$ ყველა დადებითი მთელი $n$ რიცხვისთვის;
  
  გ) $\Gamma(1) = 1;$
  
  დ) $\Gamma(1/2) = \sqrt{\pi}.$

- <h3>განმარტება 9.2:</h3>

  უწყვეტი $X$ შემთხვევითი სიდიდეს გააჩნია **გამა განაწილება**, $\alpha$ და $\beta$ პარამეტრებით, თუ მისი **სიმკვრივის ფუნქცია** მოცემულია შემდეგი სახით:
  
  $$f(x; \alpha, \beta) = 
  \left\{\begin{matrix}
  \left [\beta^{\alpha} \Gamma(\alpha) \right ]^{-1} x^{\alpha -1}e^{-x/\beta}, & x>0, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში,}
  \end{matrix}\right.$$
  
  სადაც $\alpha > 0$ და $\beta > 0$.

- <h3>განმარტება 9.3:</h3>

  უწყვეტი $X$ შემთხვევითი სიდიდეს გააჩნია **ექპონენციალური განაწილება**, $\beta$ პარამეტრით, თუ მისი **სიმკვრივის ფუნქცია** მოცემულია შემდეგი სახით:
  
  $$f(x; \alpha, \beta) = 
  \left\{\begin{matrix}
  \frac{1}{\beta} e^{-x/\beta}, & x>0, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში,}
  \end{matrix}\right.$$
  
  სადაც $\beta > 0$.

- <h3>თეორემა 9.1:</h3>
  
  **გამა** განაწილების **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულით:
  
  $$\mu = \alpha \beta \text{  და  } \sigma^2 = \alpha \beta^2.$$

- <h3>დასკვნა 9.1:</h3>
  
  **ექსპონენციალური** განაწილების **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულით:
  
  $$\mu = \beta \text{  და  } \sigma^2 = \beta^2.$$

<h2 align="center">ექსპონენციალური განაწილების კავშირი პუასონის პროცესთან</h2>

- **ექსპონენციალური** განაწილების **ყველაზე მნიშვნელოვანი გამოყენება** არის იმ შემთხვევებში, როდესაც გამოიყენება **პუასონის პროცესი**.


- კავშირი **ექსპონენციალურ** განაწილებას და **პუასონის პროცესს** შორის საკმაოდ მარტივია. 


- **მე-7 ლექციაზე** ჩვენ განვმარტეთ **პუასონის განაწილება** როგორც **ერთპარამეტრიანი განაწილება** $\lambda$ პარამეტრით, სადაც $\lambda$ შეიძლება ინტერპრეტირებული იყოს როგორც **მოვლენათა საშუალო რაოდენობა დროის ერთეულში**. 


- ახლა განვიხილოთ **შემთხვევითი სიდიდე** $X$, რომელიც აღწერს **დროს პირველი მოვლენის აღმოჩენამდე**. 


- **პუასონის განაწილების თანახმად**, ჩვენ შეგვიძლია დავწეროთ, რომ ალბათობა იმისა, რომ $t$ დრომდე **არ ხდება არცერთი მოვნელა**, არის:

  $$p(0; \lambda t) = \frac{e^{-\lambda t}(\lambda t)^0}{0!} = e^{-\lambda t}.$$
  
  
- შესაბამისად, იმის ალბათობა, რომ დროის ხანგრძლივობა პირველ მოვლენამდე გადააჭარბებს $x$-ს, იგივეა, რაც $x$-ში პუასონის მოვლენები არ მოხდება. ეს უკანასკნელი, რა თქმა უნდა, მოცემულია $e^{− λx$-ით. 


- Როგორც შედეგი, ვღებულობთ:

  $$P(X > x) = e^{-\lambda x}.$$
  
  
- შესაბამისად, უმულაციური განაწილების ფუნქცია $X$-თვის იქნება:

  $$P(0 \leq X \leq x) = 1 - e^{-\lambda x}.$$
  
  
- ახლა, იმისათვის, რომ ჩვენ დავინახოთ **ექსპონენციალური** განაწილების არსებობა, ჩვენ უნდა გავაწარმოოთ კუმულაციური განაწილების ფუნქცია:

  $$f(x) = \lambda e^{-\lambda x},$$
  
  რაც არის **ექსპონენციალური** განაწილების სიმკვრივის ფუნქცია $\beta = \frac{1}{\lambda}$ პარემეტრით.

<h2 align="center">ხი-კვადრატის განაწილება</h2>

- **გამა** განაწილების კიდევ ერთი **ძალიან მნიშვნელოვანი** განსაკუთრებული შემთხვევაა **ხი-კვადრატის** განაწილება. 


- ის **მიიღება გამა განაწილებიდან** $\alpha = v /2$ და $\beta = 2$-ის შემთხვევაში, სადაც $v$ არის დადებითი მთელი რიცხვი. 


- **ხი-კვადრატის** განაწილებას აქვს **ერთი პარამეტრი**, $v$, რომელსაც **თავისუფლების ხარისხები** ეწოდება.


- **ხი-კვადრატის** განაწილება **სტატისტიკური ჰიპოთეზის ტესტირებისა** და **შეფასების** მნიშვნელოვანი კომპონენტია.

- <h3>განმარტება 9.4:</h3>

  უწყვეტი $X$ შემთხვევითი სიდიდეს გააჩნია **ხი-კვადრატის განაწილება**, $v$ **თავისუფლების ხარისხით**, თუ მისი **სიმკვრივის ფუნქცია** მოცემულია შემდეგი სახით:
  
  $$f(x; v) = 
  \left\{\begin{matrix}
  \frac{1}{2^{v/2} \Gamma(v/2)} x^{v/2 - 1} e^{-x/2}, & x>0, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში,}
  \end{matrix}\right.$$
  
  სადაც $v$ არის დადებითი მთელი რიცხვი.

- <h3>თეორემა 9.2:</h3>
  
  **ხი-კვადრატის** განაწილების **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულით:
  
  $$\mu = v \text{  და  } \sigma^2 = 2v.$$

<h2 align="center">ბეტა განაწილება</h2>

- **ბეტა** განაწილება წარმოადგენს **თანაბარი** განაწილების გაფართოებულ შემთხვევას.

- <h3>განმარტება 9.5:</h3>

  **ბეტა ფუნქცია** ეწოდება შემდეგ ფუნქციას:
  
  $$B(\alpha, \beta) = \int_{0}^{1} x^{\alpha-1} (1-x)^{\beta - 1} dx = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)},$$
   
  სადაც $\alpha , \beta > 0$ და $\Gamma( \alpha )$ არის გამა ფუნქცია.

- <h3>განმარტება 9.6:</h3>

  უწყვეტი $X$ შემთხვევითი სიდიდეს გააჩნია **ბეტა განაწილება**, $\alpha$ და $\beta$ პარამეტრებით, თუ მისი **სიმკვრივის ფუნქცია** მოცემულია შემდეგი სახით:
  
  $$f(x; \alpha, \beta) = 
  \left\{\begin{matrix}
  \frac{1}{B(\alpha, \beta)} x^{\alpha - 1} (1 - x)^{\beta-1}, & 0 < x < 1, \\
  0, & \text{ საწინააღმდეგო შემთხვევაში.}
  \end{matrix}\right.$$
  


  <center><img src="images/L9_Beta.svg" width="800"  alt="Example" /></center>


-   მარტივი დასანახია, რომ **თანაბარი განაწილება** $(0,1)$ ინტერვალზე წარმოადგენს **ბეტა განაწილებას** პარამეტრებით $\alpha = \beta = 1$.


- <h3>თეორემა 9.3:</h3>
  
  **ბეტა** განაწილების **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულით:
  
  $$\mu = \frac{\alpha}{\alpha + \beta} \text{  და  } \sigma^2 = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}.$$

<h2 align="center">ლოგნორმალური განაწილება</h2>

- **ლოგნორმალური განაწილებას** მრავალფეროვანი გამოყენება აქვს. 


- ეს განაწილებას გამოიყენება იმ შემთხვევებში, როდესაც **ნატურალური ლოგარითმის ტრანსფორმაცია** იწვევს **ნორმალურ განაწილებას**.

- <h3>განმარტება 9.7:</h3>

  უწყვეტი $X$ შემთხვევითი სიდიდეს გააჩნია **ლოგნორმალური განაწილება** თუ შემთხვევით სიდიდეს $Y = \ln (X)$ გააჩნია **ნორმალური განაწილება** მათემატიკური ლოდინით $\mu$ და ვარიაციით $\sigma$.
  
  $X$-ის სიმკრვივის ფუნქცია აღიწერება შემდეგი ფორმულით:
  
  $$f(x; \mu, \sigma) = 
  \left\{\begin{matrix}
  \frac{1}{\sqrt{2\pi} \sigma x} e^{-\frac{1}{2\sigma^2} \left [ \ln(x) - \mu\right]^2}, & x \geq 0, \\
  0, & x < 0.
  \end{matrix}\right.$$

- <h3>თეორემა 9.4:</h3>
  
  **ლოგნორმალური** განაწილების **მათემატიკური ლოდინი** და **ვარიაცია** გამოითვლება შემდეგი ფორმულით:
  
  $$\mu = e^{\mu + \sigma^2/2} \text{  და  } \sigma^2 = e^{2\mu + \sigma^2}(e^{\sigma^2} - 1).$$
