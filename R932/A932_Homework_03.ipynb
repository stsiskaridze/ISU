<h3 align="center">ამოცანა №1: დისკრეტული შემთხვევითი სიდიდის გარდაქმნა</h3>


მოცემულია $X$ შემთხევით სიდიდე **ბინომიალური** განაწილებით:
  
  $$f(x) = 
    \left\{\begin{matrix}
   \binom{3}{x} \left ( \frac{2}{5} \right )^x \left ( \frac{3}{5} \right )^{3-x} & x = 0, 1, 2, 3,  \\ 
   0,& \text{ საწინააღმდეგო შემთხვევაში.}  
   \end{matrix}\right.
  $$
  
  იპოვნეთ $Y = X^2 - 1$ შემთხვევითი სიდიდის ალბათური განაწილება.

<h3 align="center">ამოცანა №2: უწყვეტი შემთხვევითი სიდიდის გარდაქმნა</h3>

  მოცემულია $X$ უწყვეტი შემთხევითი სიდიდე ალბათური განაწილებით:
  
  $$f(x) = 
    \left\{\begin{matrix}
   2x & 0 < x < 1,  \\ 
   0,& \text{ საწინააღმდეგო შემთხვევაში.}  
   \end{matrix}\right.
   $$
  
  იპოვნეთ $Y = 8 X^3$ შემთხვევითი სიდიდის ალბათური განაწილება.

<h3 align="center">ამოცანა №3: მომენტის გენერაციის ფუნქცია</h3>

მოცემულია $X$ დისკრეტული შემთხევით სიდიდე **თანაბარი** განაწილებით:

  $$f(x) = 
   \left\{\begin{matrix}
   \frac{1}{k} & x = 0, 1, 2, ..., k,  \\ 
   0,& \text{ საწინააღმდეგო შემთხვევაში.}  
   \end{matrix}\right.
  $$

აჩვენეთ, რომ $X$ მომენტის გენერაციის ფუნქცია ჩაიწერება შემდეგი სახით:

  $$M_X (t) = E(e^{tX}) = \sum_{x} e^{tx} f(x) = \frac{e^t(1-e^{kt})}{k(1-e^t)}$$

<h3 align="center">ამოცანა №6: ცენტრალური სასაზღვრო თეორემის გამოყენება</h3>

გარკვეული დანადგარი აწარმოებს ელექტრულ **რეზისტორებს**, რომელთა **საშუალო წინაღობა $40$ ომია** და **სტანდარტული გადახრა $2$ ომი**.

რა არის ალბათობა, რომ ასეთი ტიპის $36$ რეზისტორებიდან შემდგარ შემთხვევით ნიმუშს ჰქონდეს $1458$ ომზე მეტი ჯამური წინაღობა?

<h3 align="center">ამოცანა №7: ნიმუშის საშუალოს შეფასება, როდესაც $\sigma^2$ ცნობილია</h3>

ელექტრო ფირმა აწარმოებს ნათურებს, რომლის სიცოცხლის ხანგრძლივობაა აღიწერება ნორმალური განაწილებით ვარიაციით $\sigma^2 = 40$ საათი. 

თუ $30$ ნათურიან ნიმუშს აქვს საშუალო ხანგრძლივობა $780$ საათი, იპოვნეთ $96\%$ ნდობის ინტერვალი ამ ფირმის მიერ წარმოებული ყველა ნათურის პოპულაციის $\mu$ საშუალო მაჩვენებლისთვის.

<h3 align="center">ამოცანა №9: ნიმუშის საშუალოს შეფასება, როდესაც $\sigma^2$ უცნობია</h3>

გარკვეული დანადგარი აწარმოებს ლითონის ნაჭრებს, რომლებიც ცილინდრული ფორმისაა. 

აღებულია $9$ ნაჭრიან ნიმუშისთვის აღმოჩნდა, რომ ნაჭრების დიამეტრები შეადგენს შესაბამისად $1.01$, $0.97$, $1.03$, $1.04$, $0.99$, $0.98$, $0.99$, $0.96$, $1.03$ და  სანტიმეტრს. 

იპოვნეთ $99\%$-იანი ნდობის ინტერვალი ამ დანადგარის ნაჭრების საშუალო დიამეტრისთვის, იმის ვარაუდით, რომ დიამეტრი აღიწერება ნორმალური განაწილებით.
