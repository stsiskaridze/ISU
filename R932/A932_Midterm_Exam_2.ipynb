<h3 align="center">ამოცანა №1: ბინომიალური განაწილება</h3>

დავუშვათ, რომ თვითმფრინავის ძრავები **დამოუკიდებლად** მუშაობენ და ძრავის **მწყობრიდან გამოსვლის ალბათობა** $0.4$ ტოლია.

ჩვენ ვთვლით, რომ თვითმფრინავი **უსაფრთხო ფრენას ახდენს**, თუ მისი ძრავების **ნახევარი მაინც** მუშაობს.

რომელ თვითმფრინავს აქვს წარმატებული ფრენის უფრო მაღალი ალბათობა , **4-ძრავიანს** თუ **2-ძრავიანს**?

<h3 align="center">ამოცანა №2: მულტინომიალური განაწილება</h3>

ალბათობა იმისა, რომ გარკვეულ კრებაზე ქვეყნების წარმომადგენლები ჩამოვა **თვითმფრინავით**, **მატარებლით**, **ავტობუსით** თუ **კერძო ავტომობილით**, არის შესაბამისად, $0.4$, $0.3$, $0.2$ და $0.1$.
რისი ტოლია **ალბათობაა**, რომ ამ კრებაზე შემთხვევით შერჩეულ $9$ **წარმომადგენელიდან** $3$ ჩამოვიდა **თვითმრფინავით**, $3$ ჩამოვიდა **მატარებლით**, $2$ ჩამოვიდა **ავტობუსით** და $1$ ჩამოვიდა **კერძო ავტომობილით**?

<h3 align="center">ამოცანა №3: ჰიპერგეომეტრიული განაწილება</h3>

მჭიდში $30$ ვაზნიდან $3$ არის უვარგისი.

რა არის ალბათობა იმისა, რომ $4$ გასროლიდან ოთხივე ვაზნა გაისვრის?

<h3 align="center">ამოცანა №4: უარყოფითი ბინომიალური განაწილება</h3>

  NBA-ს (კალათბურთის ეროვნული ასოციაცია) ჩემპიონატის სერიაში გამარჯვებულია გუნდი, რომელიც **შვიდიდან ოთხ თამაშს მოიგებს**.
  
  დავუშვათ, რომ $A$ და $B$ გუნდები დაუპირისპირდნენ ერთმანეთს ჩემპიონატის თამაშებში და რომ $A$ გუნდს აქვს $0.51$ მოგების ალბათობა $B$ გუნდთან შედარებით.
  
  ა) რა არის იმის ალბათობა, რომ $A$ **გუნდი მოიგებს სერიას $6$ თამაშში**?

  ბ) რა არის იმის ალბათობაა, რომ $B$ **გუნდი მოიგებს სერიას**?

<h3 align="center">ამოცანა №5: გეომეტრიული განაწილება</h3>

  
  პანდემიის დროს ცხელი ხაზი სამსახური ძალზე გადატვირთულია, ამიტომ აბონენტებს უჭირთ ოპერატორთან დაკავშირება. 
  
  **ალბათობა** იმისა, რომ აბონენტი ერთი დარეკვით დაუკავშირდებით ცხელ ხაზს $p = 0.10$ ტოლია.
  
  რა არის იმის ალბათობა, რომ კავშირის დასამყარებლად აუცილებელია **$5$ ზარის განხორციელება**?


<h3 align="center">ამოცანა №6: პუასონის განაწილება</h3>

სამრეცხაოში საათში მოდის მომხმარებელთა გარკვეული რაოდენობა, რომელიც აღიწერება **პუასონის განაწილებით** საშუალო მნიშვნელობით $\lambda = 7$.

ა) რისი ტოლია **ალბათობა**, რომ $2$ საათში მოვა $10$-ზე მეტი მომხმარებელი?

ბ) **საშუალოდ** რამდენი მომხმარებელი მოვა $2$ საათის განმავლობაში?

<h3 align="center">ამოცანა №7: ნორმალური განაწილება</h3>

  მოცემულია $X$ **ნორმალურად განაწილებული** შემთხვევითი სიდიდე **ლოდინით** $\mu = 20$ და **ვარიაციით** $\sigma = 10$.
  
  იპოვნეთ ალბათობა, რომ $X$ **ღებულობს მნიშვნელობას** $30$-დან $50$-მდე.

<h3 align="center">ამოცანა №8: გამა განაწილება</h3>

გარკვეულ ქალაქში წყლის ყოველდღიური მოხმარება (მილიონობით ლიტრში) მიჰყვება დაახლოებით გამა განაწილებას პარამეტრებით $\alpha  = 1$ და $\beta = 3$. 

თუ ამ ქალაქის დღიური წყლის მაგარის მოცულობაა $9$ მილიონი ლიტრი წყალი, რა არის ალბათობა იმისა ქალაქი დარჩება წლის გარეშე?

<h3 align="center">ამოცანა №9: ექსპონენციალური განაწილება</h3>

გარკვეული კომპიუტერული სისტემის რეაგირების დროს აღიწერება **ექსპონენციალური** განაწილებით **საშუალო მნიშვნელობით** $3$ წამში.

ა) რა არის იმის ალბათობა, რომ რეაგირების დრო **არ აღემატება $5$ წამს**?

ბ) რა არის იმის ალბათობა, რომ რეაგირების დრო **აღემატება $10$ წამს**?

<h3 align="center">ამოცანა №10: ლოგნორმალური განაწილება</h3>

ელმავლების გარკვეული ტიპის ელექტრონულ კონტროლის **გამძლეობა** (**ათას კილომეტრებში**) აღიწერბა დაახლოებით **ლოგნორმალური განაწილებით** $\mu = 5$ და $\sigma = 1$. იპოვნეთ ასეთი ელექტრონული კონტროლის **გამძლეობის $10$ პროცენტი**.
