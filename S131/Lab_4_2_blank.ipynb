#### ნაბიჯი 6. `countplot`-მეთოდის საშუალებით მიიღეთ ქვევით მოყვანილი გრაფიკები

- `boxplot` და `violinplot` მეთოდები გამოიყენება კატეგორიული მონაცემების განაწილების საჩვენებლად. 


- `boxplot` (ან `box-and-whisker` გრაფიკი) გვიჩვენებს რაოდენობრივი მონაცემების განაწილებას ისე, რომ ხელი შეუწყოს შედარებას ცვლადებს შორის ან კატეგორიული ცვლადის დონეზე. 


- `box`-ში მოცემულია მონაცემთა ნაკრების კვარტლები, ხოლო ***ულვაშები*** ვრცელდება განაწილების დანარჩენ ნაწილზე, გარდა იმ წერტილებისა, რომლებიც განისაზღვრება ***გარეთ*** არის მეთოდის გამოყენებით, რომელიც არის მეოთხედი დიაპაზონის ფუნქცია.

#### ნაბიჯი 7. `boxplot`-მეთოდის საშუალებით მიიღეთ ქვევით მოყვანილი გრაფიკები¶

- `violinplot`-მეთოდი ანალოგიურ როლს ასრულებს, როგორც `boxplot`-მეთოდი. 


- ის გვიჩვენებს რაოდენობრივი მონაცემების განაწილებას ერთი (ან მეტი) კატეგორიული ცვლადის რამდენიმე დონეზე, რომ ამ განაწილების შედარება იყოს შესაძლებელი. 


- `boxplot`-მეთოდისგან განსხვავებით, რომელშიც გრაფიკის ყველა კომპონენტი შეესაბამება რეალურ მონაცემთა წერტილებს, `violinplot`-მეთოდი წარმოადგენს ძირითადი განაწილების ბირთვის სიმკვრივის შეფასებას (`KDE`).

- `rugplots`-მეთოდი წარმოადგენს ძალიან მარტივი კონცეფციას, ის უბრალოდ ხაზს უსვამენ ტიპების ნიშანს უნივარირებული განაწილების ყველა წერტილისთვის.


- ის გამოიყენება ბირთვული სიმკვრივის შეფასება (KDE) მისაღებად

#### ნაბიჯი 8. `violinplot`-მეთოდის საშუალებით მიიღეთ ქვევით მოყვანილი გრაფიკები¶

- `stripplot`-მეთოდი ხატავს `scatterplot`-ს, სადაც ერთი ცვლადი კატეგორიულია. 


- `stripplot`-მეთდი შეიძლება დამოუკიდებლად იყოს გამოყენებული, მაგრამ ის ასევე კარგი დამატებაა `boxplot` ან `violinplot` მეთოდებისთვის, როდესაც გსურთ წარმოაჩინოთ ყველა დაკვირვება და ძირითადი განაწილების გარკვეული წარმოდგენა.


- `swarmplot`-მეთოდი მსგავსია `stripplot`-მეთდისა, მაგრამ წერტილები მორგებულია (მხოლოდ კატეგორიული ღერძის გასწვრივ) ისე, რომ ისინი არ გადაფარონ.
