<h3 align="center">ამოცანა 2: თამაში "სიცოცხლე" (5 ქულა)</h3>

- დაწერეთ თამაში **სიცოცხლე** და მოახდინეთ მისი ვიზუალიზაცია:

  https://en.wikipedia.org/wiki/The_Game_of_Life


- **თამაშის აღწერა**:
  - თამაში **სიცოცხლე** ხდება **უჯრედებიან ფურცელზე**, რომელსაც ტრადიციულად **სამყარო** ჰქვია;
  - **თითოეული უჯრედი** შეიძლება იყოს ან **ცოცხალი** ან **მკვდარი**;
  - თაობები სინქრონულად იცვლება მარტივი წესების მიხედვით:
    - **მკვდარ უჯრედში**, რომლის გვერდით **ზუსტად სამი ცოცხალი უჯრედია**, იბადება **სიცოცხლე**;
    - თუ **ცოცხალ უჯრედს** ჰყავს **ორი ან სამი ცოცხალი მეზობელი**, მაშინ ეს **უჯრედი განაგრძობს ცხოვრებას**;
    - თუ **ცოცხალ უჯრედს** ჰყავს **ორზე ნაკლები** ან **სამზე მეტი ცოცხალი მეზობელი**, მაშინ ეს **უჯრედი კვდება** (**მარტოობის** ან **ჭარბი პოპულაცის გამო**).
    - **მეზობლებად ითვლება უჯრედები**, რომლებსაც აქვთ ერთი საერთო წერტილი მაინც, ანუ **თითო უჯრედს ჰყავს 8 მეზობელი**.
    

- **დავალება**:
  - აირჩიეთ **სამყაროს** ზომა.
  - **დააგენერირეთ** თავდაპირველი ცოცხალი უჯრედების კონფიგურაცია;
  - **შექმენით შემდეგი თაობის (ნაბიჯის) ფუნქცია**;
  - **ცვალეთ თაობები** სანამ არ მიიღებთ უცვლელ სურათს ან სანამ არ მოხდება განმეორება;
  - მოახდინეთ **ვიზუალიზაცია**.
