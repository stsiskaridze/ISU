- ამ სავარჯიშოში ჩვენ იფორმაციის **შეყვანის** კიდევ ერთ მეთოდს ვისწავლით, რომელიც სკრიპტისთვის ცვლადების გადაცემის საშუალებას იძლევა.


- **სკრიპტი** არის ბრძანება, რომელიც უშვებს თვენქს ფაილებს: `.python ex_13.py`.


- როდესაც თქვენ უშვებთ სკრიპტს, თქვენ წერთ `python ex_13.py`. ამ გამოსახულებაში `ex_13.py`-ს ეწოდება **არგუმენტი**. 


- ახლა კი ჩვენ დავწერთ *სკრიპტს*, რომელიც ასევე იღებს *არგუმენტებს*.


- გადმოწერეთ ეს პროგრამა, და ქვემოთ უფრო დეტალურად აგიხსნით:

```python
from sys import argv
# ქვემოთ ნახავთ, სწორად როგორ გაუშვათ ეს პროგრამა
script, first, second, third = argv

print("ამ სკრიპტის სახელწოდებაა:", script)
print("თქვენი პირველი ცვლადია:", first)
print("თქვენი მეორე ცვლადია:", second)
print("თქვენი მესამე ცვლადია:", third)
```

- პირველ სტრიქონზე გვაქვს ე.წ. `import`-ი. 

  ამ გზით ჩვენ ვამატებთ მოდულებს **Python**-ის ბიბლიოთეკიდან. 
  
  იმის მაგივრად, რომ ყველაფერზე ერთად მოგცეთ წვდომა, **Python**-ი გთხოვთ, დააზუსტოთ, თუ რომელი ბიბლიოთეკის გამოიყენება გსურთ. 
  
  ამ გზით თქვენი კოდი იყენებს მხოლოდ **Python**-ის მხოლოდ იმ საშუალებებს, რაც მას უშუალოდ სჭირდება გასაშვებად.

  ამიერიდან ჩვენ ვიტყვით, რომ **ვაიმპორტებთ მოდულებს**.


- `argv` მოკლე სახელია **"argument variable"**-ისთვის (ქართ., "**არგუმენტის ცვლადი**"). 

  აღნიშნულ არგუმენტში ჩაწერილი იქნება ის ცვლადები, რომელიც შემდგომ სკრიპტს უნდა გადაეცეს.
   

- მესამე სტრიქონი ხსნის (eng. "unpacks") `argv`-ს და მათ მნიშვნელობებს მათთვის გამოყოფილ ცვლადებს ანიჭებს: `script`, `first`, `second`, `third`. 

  ამის შემდეგ კი მათ ჩვეულებრისამებრ იყენებს.
  

- ახლა კი გავუშვათ ჩვენი პროგრამა. ტერმინალში უნდა აკრიფოთ შემდეგი ბრძანება:

  `python ex_13.py პირველი მეორე მესამე`

<h2 align="center">ხშირად დასმული კითხვები</h2>

- ქვევით მოყვანილია კითხვები, რომლებიც გაუჩნდათ რეალურ სტუდენტებს ამ სავარჯიშოს შესრულებისას:


1. **სკრიპტის გაშვებისას მიჩვენებს შეცდომას `ValueError: need more than 1 value to unpack.**

   ყურადღებით წაიკითხეთ ამ ლექციაში მოცემული ინფორმაცია და დააკვირდით დეტალებს. 
   
   თქვენი სკრიპტის გაშვებისას, ტერმინალში უნდა შეიყანოთ იმდენი ცვლადი, რამდენიც მითითებული გაქვს კოდში.
   
   
2. **რა განსხვავებაა `argv`-სა და `input()`-ს შორის?**

   განსხვავება მდგომარეობს იმაში, თუ ვისგან მოითხოვება ინფორმაცია. 
   
   თუ თქვენ არგუმენტებს აწვდით სკრიპტს, მაშინ გამოიყენეთ `argv`-ი.
   
   ხოლო, თუ მომხმარებლისგან დაგჭირდათ ინფორმაცია სკრიპტის მუშაობის დროს, მაშინ გამოიყენეთ `input()`-ი. 
   
   
3. **`command line`-ის არგუმენტები სტრიქონებია?**

    კი, ისინი აღიქმება სტრიქონება, თუნდაც მხოლოდ რიცხვები გექნებათ შეყვანილი. გამოიყენეთ `int()` ფუნქცია, რათა რიცხვებად აქციოთ.
    
    
4. **როგორ გამოვიყენოთ `command line`?**

    სიმართლე რომ გითხრათ, აქამდე უნდა გაგაჩნდათ ეს ცოდნა. თუმცა, ჩვენ მიერ გამოყენებული წიგნის **Appendix A**-ში მოიპოვება ეს ინფორმაცია. 
    
    
5. **ვერ ვიყენებთ `argv`-სა და `input()` ერთ კოდში.**

    ზედმეტი ფიქრი არ გინდათ. ერთ ხაზზე გამოიყენეთ `argv` , ხოლო შემდეგზე - `input()`, რომ რაიმე ინფორმაცია ჰკითხოთ მომხმარებელს. 
    
    შემდეგ უბრალოდ ამობეჭდეთ შედეგი.
