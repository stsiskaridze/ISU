* გახსოვთ, `while` ციკლი რომ ვისწავლეთ, ჩვენ სიას დავურთეთ (`append`) ახალი ელემენტები და ამოვბეჭდეთ ისინი:

```python
elements = []

for i in range (0, 6):
    print(f"სიას ვამატებთ {i}-ს.")
    elements.append(i)

for i in elements:
    print(f"ეს ელემენტია: {i}")
```

* **რეალურად რას აკეთებს ფუნქცია `elements.append()`**?


1. Python ხედავს, რომ ვახსენეთ `elements`, და დაიწყებს ამ ცვლადის ძებნას.

2. როგორც კი ის აღიქვამს წერტილს `.`, ის `elements`-ის წაკითხვას დაიწყებს. რადგან `elements` სიაა, Python-მა იცის, რომ მასში უამრავი ცვლადი არსებობს.

3. შემდეგ Python დაძებნის სიტყვას "append" `elements` სიის ცვლადებში და, რა თქმა უნდა, იპოვის.

4. შემდეგ Python ხედავს გახსნილ მრგვალ ფრჩხილს და ხვდება, რომ მის წინ ფუნქციაა; და ამ ფუნქციას სჭირდება გაშვება. Python უშვებს ამ ფუნქციას *დამატებითი* არგუმენტით.

5. დიახ. ეს დამატებით არგუმენტია `elements`! უცნაურია, ხო? მაგრამ Python ხედავს მას, როგორც `append(elements, i)`, ჩვენ კი ვკითხულობთ `elements.append(i)`. 

* უმეტეს შემთხვევაში, ჩვენ არ დაგვჭირდება ვიცოდეთ, როგორ მუშაობს Python. მაგრამ ეს ცოდნა Python-ისგან მიღებული **შეცდომების აღქმაში** დაგვეხმარება.


```python
>>> class Thing(object):
...     def test(message):
...       print(message)
...
>>> a = Thing()
>>> a.test("hello")
```

- **რა ეს არის?** ეს არის **powershell**-ში აკრეფილი მცირე კოდი.


- თქვენ ჯერ არ იცით, რა არის `class`, ამაზე შემდეგ ლექციაზე ვილაპარაკებთ. 


- ჯერ მოდით დავაკვირდეთ, რომ Python-მა გვითხრა `test() takes exactly 1 argument (2 given)`. 


- ამას თუ დაინახავთ, ესე იგი Python-მა **შეცვალა** ჩვენი კოდი `a.test("hello")` და **აღიქვა** როგორც `test(a, "hello")`, და რომ ვიღაცამ დაუშვა შეცდომა და `a`-სთვის არ დაუმატებია არგუმენტი.

* მოდით ცოტა განვიტვირთოთ და ვითამაშოთ სკრიპტით:

```python
ten_things = "ვაშლი ფორთოხალი ყვავი ტელეფონი შუქი შაქარი"

print("მოიცადეთ, ამ სიაში 10 ნივთი არაა. მოდით გამოვაწოროთ.")

stuff = ten_things.split(' ')
more_stuff = ["დღე", "ღამე", "სიმღერა", "თამაში", 
              "სიმინდი", "ბანანი", "გოგო", "ბიჭი"]

while len(stuff) != 10:
    next_one = more_stuff.pop()
    print("დამატეთ: ", next_one)
    stuff.append(next_one)
    print(f"ახლა სიაში გვაქვს {len(stuff)} ნივთი.")
    
print("აი ახლა ვნახოთ ჩვენი სია: ", stuff)

print("მოდით კიდევ რამე გავაკეთოთ.")

print(stuff[1])
print(stuff[-1]) 
print(stuff.pop())
print(' '.join(stuff))
print('#'.join(stuff[3:5]))
```

<h2 align="center">რა შეუძლიათ სიებს</h2>

- წარმოიდგინეთ, რომ სია არის *ბანქოს დასტა*. პროგრამისტები ასეთ სტრუქტურას **მონაცემთა სტრუქტურას** უწოდებენ.


- რა არის **მონაცემთა სტრუქტურა**? ეს არის რაღაც ფაქტების (მონაცემების) სტრუქტურირების (დალაგების, ორგანიზების) გზა. ეს მართლაც ასეთი მარტივია. 


- სხვა სიტყვებით რომ ვთქვათ, ეს სტუქტურა საშუალებას გვაძლევს, კოდში მონაცემები სტრუქტურირებულად შევინახოთ და სხვადასხვა გზით მათზე წვდომა გვქონდეს.


- მონაცემთა სტრუქტურის ყველაზე ხშირი სახეობაა სია. სია არის დანომრილი ფაქტების ჩამონათვალი, შეგიძლია შეინახო თითქმის ნებისმიერი ტიპის ინფორმაცია და მიიღო მათზე წვდომა **შემთხვევითად** (randomly), **სათითაოდ** (linearly) ან **ინდექსის** მიხედვით.


- ახლა კი დავუბრუნდეთ ბანქოს დასტას და განიხილოთ ის როგორც სია:

  1. ჩვენ გვაქვს ბევრი ბანქო სხვადასხვა მნიშვნელობით.
  2. ეს ბანქოები დასტაშია, სიაში, ზედა ბანქოდან დაწყებული ქვედათი დამთავრებული.
  3. შენ შეგიძლია აიღო ბანქო ზევიდან, ქვევიდან და შემთხვევითად დასტის შუა ნაწილიდან.
  4. კონრეტული ბანქო რომ დაგჭრდეს, მთელი დასტა უნდა აიღო და სათითაოდ ჩამოიარო.


- ეს არის და ეს, რასაც **სია** აკეთებს. 


- პროგრამირების ყველა კონცეპტი, როგორც წესი, გარკვეულ კავშირშია რეალურ სამყაროსთან. ყოველ შემთხვევაში სასარგებლო კონცეპციები არის. თუ იპოვით მათ ანალოგს რეალურ სამყაროში, მათი ათვისებაც გაგიადვილდებათ.
