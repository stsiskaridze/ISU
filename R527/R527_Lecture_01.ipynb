- კურსი განკუთვნილია მათთვის ვისაც **არ აქვს გამოცდილება** პროგრამირებაში და არ ქონია შეხება კომპიუტერული მეცნიერებების დისციპლინებთან.


- გთხოვთ დამისვათ შეკითხვები თუ რამეს ვერ გაიგებთ, ან ჩათვლით რომ რთულად იყო ახსნილი.

<h2 align="center">რა უნდა ვიცოდეთ კურსის დაწყებამდე?</h2>

- საბაზისო მათემატიკა: რა არის ფუნქცია? არითმეტიკული და ლოგიკური ოპერაციები.


- საბაზისო სტატისტიკა: რა არის ჰისტოგრამა? რა არის განაწილება?


- ამოცანის გადაჭრის თანმიმდერვრული და ლოგიკური გზის განსაზღვრის უნარი.

<h2 align="center">რას ვისწავლით ამ კურსზე?</h2>

- ვისწავლით თუ რა არის პროგრამირება

  გამოვიყენებთ პითონს, რათა შევქმნათ კომპიუტერისთვის გასაგები მოქმედებების თანმიმდევრობა რომელიც მიგვიყვანს პრობლემის გადაწყვეტამდე 
  
  თუ გვქონია ასეთი გამოცდილება უკვე?


- გავიგებთ თუ როგორ უნდა შევიმუშავოთ კონკრეტული პრობლემის გადაჭრის ალგორითმი მსგავსად ზემოთ ხსენებული პრობლემისა.


- გავიგებთ თუ როგორ შეგვიძლია ეფექტურად დავამუშავოთ მონაცემები და მიღებული შედეგებიდან გამოვიტანოთ დასკვნები

<h2 align="center">კურსის სტრუქტურა</h2>


- **აქტივობა** (**10 ქულა**):

  დამისვით კითხვები რომელიც შეეხება კურსში გადმოცემულ თეორიულ საკითხებს და შეეცადეთ სწორად უპასუხოთ ჩემს მიერ დასმულ კითხვებს


- **ლაბორატორიული დავალება** (**20 ქულა**):

  კურსის განმავლობაში გექნებათ **4 ლაბორატორიული** დავალება **5 ამოცანით**. 
  
  **თითო ამოცანა** შეფასდება **1 ქულით**.


- **შუალედური გამოცდები** (**20 ქულა + 20 ქულა**):

  კურსის განმავლობაში იქნება **2 შუალედური გამოცდა** რომელიც შეფასდება 20 ქულით.


- **დასკვნითი გამოცდა** (**30 ქულა**):

  დასკვნითი გამოცდის შეფასდება 30 ქულით.
  

- **მინიმალური ზღვარი** კურსის გავლისთვის არის **51 ქულა**.

<h2 align="center">დამატებითი შეხვედრები</h2>

- სასურველია **დამატებითი შეხვედრების** ორგანიზება წინასწარ შევათანხმოთ.


- შეხვედრის დასანიშნად შეგიძლიათ მომწეროთ **ARGUS**-ის საშუალებით.


- შეხვედრებს **ნუ დანიშნავთ ინდივიდუალურად**! დაჯგუფდით, რომ უფრო ეფექტურად გავცე პასუხი თქვენს კითხვებს.


- თუ კითხვა **არ არის პირადი** და მასზე პასუხი სხვებსაც გამოადგება, თხოვნაა **ARGUS**-ზე **დისკუსიების** გრაფში დაწეროთ ის **კომენტარის სახით**.

<h2 align="center">ლიტერატურა და წასაკითხი მასალა</h2>

- **ქართულად ლიტერატურა არ არსებობს!**


- შევეცდები პარალელურად მოვამზადო **რიდერი** რაც დაგეხმარებათ მასალის ათვისებაში.


- თუ ფლობთ **ინგლისურს** ან **რუსულს** ენებს თქვენ გაგიადვილდებათ მასალების მოძიება. უამრავი წიგნი და სალექციო კურსია რაც ამ ენებზე.


- კურსი არის **საშუალო სირთულის**, ორიენტირებულია იმაზე რომ თქვენგან ბევრმა აიღოს ცოდნა და წარმოდგენები პროგრამირებაზე მაინც.


- სემინარების დავალებები გაყვება მითითებულ სახელმძღვანელოს: 

  **Zed Shaw, *Learn Python 3 the Hard Way*, Addison Wesley (Verlag), 2017**
  
  ყოველი **სემინარის თავში** მითითებული იქნება თუ რომელი **თავი არის გასარჩევი**.
  

- ყველა მასალა ატვირთული იქნება **ARGUS**-ზე.


- მიმდინარეობს ლექციების ჩაწერა **Zoom**-ის საშუალებით და იმავე დღეს ვიდეომასალა ატვირთული იქნება **ARGUS**-ზე.
