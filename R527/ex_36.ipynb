<h1 align="center">სავარჯიშო 36: პროგრამის დაწერის წესები</h1>

<center><img src="images/Ex_36_Intro.jpg" width="500"alt="Example" /></center>

- ამ სავარჯიშოში რამდენიმე **წესს** შემოგთავაზებთ, რომელიც გაგმიარტივებთ კოდის გამართვას.


- ასევე, მოგცემთ რჩევას **დებაგინგთან** დაკავშირებით.


- და ბოლოს, წინა სავარჯიშოს მსგავსს თამაშის სტრუქტურას დავწერთ.

<h3 align="center">If ბრძანებების წესები: </h3>

1. ყოველ `if` ბრძანებას აუცილებლად უნდა ახლდეს `else`.


2. თუ `else`-ის კოდი მოკლებულია აზრს, ჯობია მასში `die` ფუნქცია ჩაწეროთ, რომელიც ამოაგდებს შეცდომას და გააჩერებს სკრიპტს. 


3.  `if` ბრძანებების ძალიან ღრმად, რამდენიმე დონით ქვემოთ დაწერას აარიდეთ თავი; 2-ზე მეტი დონე არ არის რეკომენდირებული.


4. ყოველი `if-elif-else` გამოყავით პარაგრაფებად: გამოტოვეთ თითო ხაზი თავში და ბოლოში. 


5. თქვენი ლოგიკური ამოსახულებები მარტივი უნდა იყოს. 

   თუ დაგჭირდათ ვრცელი კალკულაცია, უმჯობესია გადაიტანოთ ის ოდნავ ზევით და მისი მნიშვნელობა ცვლადს მიანიჭოთ. 
   
   არ დაგავიწყდეთ ამ ცვლადისთვის კარგი სახელის მოფიქრება.

- **თუ ამ წესებს მიყვებით, პროგრამისტების უმრავლესობაზე უკეთეს კოდს დაწერთ.** 

   **დაუბრუნდით წინა სავარჯიშოს და გადაამოწმეთ, თუ არის ყველა წესი დაცული, თუ არა -- გამოასწორეთ კოდი.**

<h3 align="center">ციკლის წესები: </h3>

1. გამოიყენეთ `while`-ციკლი სამუდამო ციკლებისთვის, ანუ თითქმის არასდროს. 

   ეს წესი მხოლოდ Python-ისთვისაა, სხვა ენებში `while`-ციკლს სხვა დატვირთვა აქვს.


2. `for`-ციკლი გამიოყენეთ ყველა დანარჩენ შემთხვევაში, განსაკუთრებით თუ ციკლის გამეორების სასრული ან ფიქსირებული რიცხვი გაქვთ.


<h3 align="center">რჩევა დებაგინგინგთან დაკავშირებით: </h3>

1. შეეცადეთ **დებაგერი** არ გამოიყენოთ! 

   დებაგერის გამოყენება არის ტოლი იმისა, რომ ავადმყოფ ადამიანს მთელი სხეულის გამოკვლევა ჩაუტაროთ. 
   
   საჭირო ინფორმაციის მიღების მაგივრად უფრო დაიბნევით.


2. საუკეთესო დებაგინგი `print()` ფუნქციის გამოყენებაა. 

   კოდის მუშაობის სხვადასხვა ეტაპზე ამობეჭდეთ ცვლადების მნიშვნელობები და თვალი ადევნეთ მათ ცვლილებას. 
   
   შეცდომასაც აუცილებლად აღმოაჩენთ.


3. კოდის წერისას დარწმუნდით რომ კოდის ნაწილი, რომელზედაც ამ კონკრეტულ მომენტში მუშაობ, გამართულია. 

   არ შეეცადოთ დაწეროთ მასიური კოდი და არ დატოვოთ ისინი უყურადღებოდ. 
   
   პატარა კოდი დაწერე, პარატა კოდი გაუშვი, პატარა კოდი ჩაასწორე.

<h2 align="center">საშინაო დავალება</h2>


ახლა კი, დაწერეთ წინა სავარჯიშოს მსგავსი თამაში იმ ცოდნით, რაც აქამდე დაგიგროვდათ, გამოიყენეთ სიები, ფუნქციები, მოდულები.


სანამ დაიწყებთ წერას, ფურცელზე დახატეთ თქვენი თამაში, გამოყავით ოთახები, მოიფიქრეთ ამოცანები / ურჩხულები / მახეები.


თვალწინ რომ გექნებათ რუკა, დაიწყეთ კოდის წერა. შესწორებების შეტანა რომ დაგჭირდეთ, რუკაზეც მონიშნეთ, რა შეცვლეთ. 


დიდი ამოცანის შესასრულებლად ყველაზე კარგია, დაყოთ ის მცირე ეტაპებად:


1. შექმენით `to-do list` ამოცანების/კოდის ნაწილები, რისი შესრულებაც იქნება საჭირო.


2. ამოარჩიეთ ყველაზე ადვილი ამოცანა და იმით დაიწყეთ.


3. კომენტარის სახით დაწერეთ ე.წ. სახელმძღვანელო.


4. ამ კომენტარს ქვემოთ დაურთეთ შესაბამისი კოდი.


5. სწრაფად გაუშვით სკრიპტის ნაწილი, რათა დარწმუნდეთ მის სისწორეში.


6. იმუშავეთ მასეთ რეჟიმში: მცირე კოდის დაწერა, მისი გაშვება და შემოწმება.


7. ამოშალეთ შესრულებული დავალება თქვენი `to-do list`-დან, და გადადით პუნქტ მეორეზე. 

ეს საშინაო დავალება დაგეხმარებათ გამოიმუშაოთ მეთოდოლოგიური, საფუძვლიანი მიდგომა ამოცანისადმი.
