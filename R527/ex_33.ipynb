<h1 align="center">სავარჯიშო 33: While ციკლები</h1>

- ახლა საერთოდ რომ დაგაბნიოთ, შემოგაქვს **`while` ციკლი**.


- `while`-ციკლი ამუშავებს მის ქვეშ დაწერილ კოდის ბლოკს მანამ, სანამ ციკლში მოცემული ლოგიკური ბრძანება იქნება მართალი (`True`). 


- გავიხსენოთ, რომ თუ ხაზს ვამთავრებთ ორიწერტილით (:), ქვემოთ იქმნება *კოდის ბლოკი*, რომელიც შეწეულია 4 გამოტოვებით. 


- **რას შვრება `while`-ციკლი?** ის უბრალოდ ამოწმებს მასში მოცემულ `if`-ბრძანებას და, კოდის ბლოკის მუშაობის დამთავრებისას, ის *ხტება* ისევ ამ `if`-ბრძანებაზე და თავიდან ამოწმებს მის სისწორეს. როგორც კი გამოსახულება იქნება მცდარი (`False`), ციკლი შეწყდება.


- მაგრამ აი პრობლემა: ხანდახან `while`-ციკლი არ წყვეტს მუშაობს. 


- ამგვარ პრობლემას თავი რომ ავარიდოთ, მივყვეთ შემდეგ ინსტრუქციას:

  1. დარწმუნდით, რომ `while`-ციკლს ნაკლებად იყენებთ. როგორც წესი, `for`-ციკლი უკეთესია.
    
  2. გადახედეთ თქვენს ბრძანებას ციკლში და დარწმუნდით, რომ რაღაც მომენტში ის აუცილებლად გახდება მცდარი.

  3. თუ არ ხართ დარწმუნებული ციკლში, დააბეჭდინეთ ტესტური ცვლადი ამ ციკლის თავსა და ბოლოში, რათა დაინახოთ შედეგი.
    

- შემდეგ დავალებაში თქვენ ისწავლით `while`-ციკლს და თან გამოიყენებთ ზემოთ მოცემულ შემოწმების 3 ეტაპს.

```python
i = 0
numbers = []

while i < 6:
    print(f"კოდის თავში i უდრის {i}-ს")
    numbers.append(i)
    
    i = i + 1
    print("ჩვენს სიაშია ახლა: ", numbers)
    print(f"კოდის ბოლოში i უდრის {i}-ს")
    
    
print("ჩვენი სია: ")

for num in numbers:
    print(num)
```

<h2 align="center">რუბრიკა "ბოროტი მანტიორი"</h2>

- ჩვენ ხშირად შევცვლით კოდს, რომ დავადგინოთ მისი გაშვების საზღვრები.


- ქვევით მოყვანილია **კოდის მაგალითები** რომლებიც **არ კომპილირდება** და **იძლევა შეცდომას**. 


- **გაარჩიეთ** თითოეული მათგანი და **გაასწორეთ შეცდომები**!

```python
string = "აი ია"
while bool(string):
    string = string[:-1]
    print(string)
```

```python
string = "აი ია"
while string:
    string = string[:-1]
    print(string)
```

1. **რა განსხვავებაა `while` ციკლსა და `for` ციკლს შორის?**

   `for` ციკლი იციკლება მხოლოდ საგნის კოლექციაზე მაშინ, როცა `while` ციკლს შეუძლია ნებისმიერი იტერაციის განმეორება. თუმცა  `while` ციკლის სწორად დაწერა უფრო ძნელია.
   
   
2. **ციკლი ძნელია გასაგებად. როგორ ვისწავლო ისინი უკეთ?**
   
   ციკლის მთავარი სიძნელე მდგომარეობს იმაში, რომ ბევრს თავიდან არ ესმის თუ როგორ მოქმედებს ციკლი. ციკლი ბოლოში რომ ჩაირბენს, ისევ თავში ხტება და ამოწმებს მოცემულ ბრძანებას. ამის ვიზუალიზაციისთვის ჩაწერეთ ყველგან `print()` ფუნქციები და თვალი ადევნეთ, შეისწავლეთ შედეგი და გააანალიზეთ ის.
