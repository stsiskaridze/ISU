- სერიოზული თემაა, არა? დღეს ჩვენგ გავარჩევთ თუ რა არის **ფუნქცია** და როგორ გამოვიყენოთ ის! 


- ნებისმიერ პროგრამისტს რომ ჰკითხოთ, გაუჩერებლივ ილაპარაკებენ ფუნქციებზე, რას აკეთებენ ისინი, რას იყენებენ, როგორ მუშაობენ.


- ჩვენ მაგრამ მე უადვილეს ახსნას შემოგთავაზებთ. 


- **ფუნქციები** აკეთებენ **3 რამეს**:
  1. ისინი **ასახელებენ კოდის ნაწილს** ისევე, როგორც ცვლადები ასახელებენ სტრიქონებს** და რიცხვებს;
  2. ისინი **იღებენ არგუმენტებს** ისევე, როგორც სკრიპტი იღებს `argv`-ს;
  3. **პირველი** და **მეორე პუნქტის მეშვეობით** შესაძლებლობას გაძლევენ, **შექმნათ თქვენი "მინი-სკრიპტები"** ან **"მცირე ბრძანებები"**.


- **Python**-ში ფუნქციის **შესაქმნელად** გამოოიყენეთ `def` **სიტყვა**, რაც ნიშნავს **define**-ს, ქართ., **განსაზღვრა**.


- ქვედა მაგალითში მოყვანილია **4 ფუნქცია**:

```python
# ეს თქვენი სკრიპტის მსგავსია, როდესაც `argv`-ს ვიყენებთ
def print_two(*args):
    arg1, arg2 = args
    print(f"პირველი არგუმენტია: {arg1}, მეორე არგუმენტია: {arg2}")
      
# სიმართლე გითხრათ, ეს *args აზრს მოკლებულია, ჩვენ შეგვიძლია დავწეროთ
def print_two_again(arg1, arg2):
    print(f"პირველი არგუმენტია: {arg1}, მეორე არგუმენტია: {arg2}")
    
# ეს აიღებს მხოლოდ ერთ არგუმენტს
def print_one(arg1):
    print(f"პირველი არგუმენტია: {arg1}")
    
# ეს საერთოდ არ აიღებს არგუმენტს 
def print_none():
    print("სადაა ჩემი არგუმენტი?!")
    
    
print_two("კონსტანტინე","ცისკარიძე")
print_two_again("კონსტანტინე","ცისკარიძე")
print_one("პრიველი!")
print_none()
```

- მოდით ჯერ განვიხილოთ ფუნქცია `print_two`:
  1. პირველ რიგში, `def`-ის მეშვეობით (**line 2**) ჩვენ ვაცნობებთ **Python**-ს, რომ **ფუნქციის შექმნა გვინდა**.
  2. `def`-ის ხაზზე ჩვენ ჩვენს ფუნქცია **ვაძლევთ სახელს**, ამ შემთხვევაში `print_two`. იგივენაირად შეგვეძლო დაგვერქვა `პარასკევი`. **ყურადღება**: ფუნქციას დასაქმევად გამოიყენეთ მოკლე და ადვილად გასაგები სახელი. 
  3. შემდეგ ვეუბნებით, რომ ჩვენ **გვჭირდება არგუმენტები** `*args` (**ფიფქი არგს**), რომელიც ძალიან ჰგავს აქამდე გამოყენებულ `argv`-ს. 
  <br>ის **აუცილებლად უნდა მოიქცეს ფრჩხილებში**  `()` .
  4. **ხაზის ბოლოს აუცილებელია დავწეროთ`:`**, ამით ჩვენ ფუნქციის განსაზღვრას ვიწყებთ.
  5. ორწერტილის შედეგ **ყველა ხაზი**, რომელიც **4 გამოტოვებით შეწეული** იქნება (**lines 3-4**), მიეკუთვნება `print_two` ფუნქციას. 
  <br>ჩვენი ფუნქციის პირველს ხაზს გამოაქვს არგუმენტები იგივე პრინციპით, რაც `argv`.
  6. ფუნქციის მუშაობის **სადემონსტრაციოდ** ვიყენებთ `print`-ს. 


- თუმცა, `print_two` ფუნქცია **არ წარმოადგენს ფუნქციის განსაზღვრის საუკეთესო ვარიანტს**!

  **Python** საშუალებას გვაძლებს, გამოვტოვოთ **unpacking**-ის ნაბიჯები და `()` ფრჩილებში პირდაპირ ჩავწეროთ საჭირო არგუმენტები. 

  ამ ყველაფერს აკეთებს ფუნქცია `print_two_again`. 


- ამის შემდეგ მოყვანილი **ერთარგუმენტიანი ფუნქციის მაგალით** `print_one`.


- და ბოლოს, **ფუნქცია არგუმენტების გარეშე** `print_none`.
