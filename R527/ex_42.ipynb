<h1 align="center">სავარჯიშო 42: არის (is-a), აქვს (has-a), ობიექტები და კლასები</h1>

- მნიშვნელოვანი ცნება, რომელიც ახლა უნდა გავიაზროთ არის: **რა განსხვავებაა კლასსა და ობიექტს შორის?**

  პასუხი არის, რომ მათ შორის **არ არსებობს რეალური განსხვავება**.


- მოდით ასე შევხედოთ საკითხს:

  **რა განსხვავებაა ქაშაყსა და თევზს შორის?**


- ამ კითხვამ დაგაბნიათ? ანუ, თევზი და ქაშაყი განსხვავებული ცნებაა, მაგრამ მათ შორის ისეთი განსხვავება არ არსებობს. თქვენ ეს კითხვა არ დაგებადათ, რადგან ჩვენ, ადამიანებს, *გვესმის* როგორ არიან ისინი დაკავშირებული.



- დავუშვათ, თქვენ გყავთ **3 ქაშაყი** და დაარქვით მათ სახელები **ავთო**, **ბესო** და **გოჩა**. 

- მოდით უფრო გავართულოთ შეკითხვა. 

  **რა განასხვავებს გოჩას ქაშაყისგან?**


- ისევ უცნაური კითხვა, მაგრამ ჩვენ ვიცით, რომ გოჩა ქაშაყია, ამიტომაც ისინი დიდად არ განსხვავდება ერთმანეთისგან. გოჩა უბრალოდ ქაშაყის სპეციფიკური მაგალითია, ანუ მასსაც ჩვეულებრივი ქაშაყის თვისებები გააჩნია!


- ახლა შევაჯამოთ. **თევზი არის კლასი**, **ქაშაყი არის კლასი** და **გოჩა არის ობიექტი**.

- ანუ, **თევზი არის კლასი** და **არა რეალური ობიექტი**. გაგაჩნია ლაყუჩები და ფარფლები? ცხოვრობ წყალში? ესე იგი, ალბათ, თევზი ხარ.


- მერე გამოდის ზაზა დოქტორის ხარისხით და ამბობს: "არა მეგობრებო, ეს თევზი სინამდვილეში ქაშაყია, წაგრძელებული სხეულით და პაწაწუნა თვალებით". ამით **შეიქმნა კიდევ ერთი კლასი** სახელად **ქაშაყი**, რომელსაც უფრო სპეციფიკური ატრიბუტები აქვს. 


- მერე გამოდის მზარეული ზაირა და ამბობს: ამ ქაშაყს ხედავთ? მე მას **გოჩას დავარქმევ** და გემრიელ ფილეს მოვამზადებ გადასარევ სოუსში. 

- ანუ, ჩვენ ავირჩიეთ ქაშაყის (თევზის) კონკრეტული შემთხვევა სახელად გოჩა. ის გახდა ობიექტი.


- რომ შევაჯამოთ, გვაქვს ობიექტი **გოჩა** რომელიც **ქაშაყის ტიპისაა**, რომელიც **თევზის ტიპისაა**. ანუ **ობიექტი არის კლასი რომელიც არის კლასი**.

<h3 align="center">როგორ გამოიყურება ეს კოდში</h3>

- კი, ჩვენ ვისწავლეთ უცნაური ცნება, მაგრამ ამაზე ფიქრი მოგვიწევს მხოლოდ მაშინ, როცა ახალ კლასს ვქმნით და ვიყენებთ მას კოდში.


- პირველ რიგში ფრაზების **არის (is-a)** და **აქვს (has-a)** გამოყენება უნდა ვისწავლოთ. 


- **არის (is-a)** გამოიყენება, როცა ვსაუბრობთ კლასებსა და ობიქეტებზე, რომლებიც დაკავშირებულია კლასის ტიპის ურთიერთობით.


- **აქვს (has-a)** გამოიყენება კლასებსა და ობიექტებზე საუბრისას, რომლებსაც ერთმანეთზე მიმართება, **მითითება (reference)** აქვთ.


- ახლა კი, წინამდებარე კოდში **ჩასვით** გამოტოვებულ კომენტარებში **არის (is-a)** ან **აქვს (has-a)** ურთიერთობები და ახსენით მათი კავშირები.


- **გახსოვდეთ**: **არის (is-a)** არის ურთიერთობა თევზსა და ქაშაყს შორის, ხოლო **აქვს (has-a)** არის ურთიერთობა ქაშაყსა და ლაყუჩებს შორისაა.

```python
## Animal *არის* ობიექტი
class Animal(object):
    pass

## ??
class Dog(Animal):

    def __init__(self, name):
        ## ??
        self.name = name

## ??
class Cat(Animal):

    def __init__(self, name):
        ## ??
        self.name = name

## ??
class Person(object):

    def __init__(self, name):
        ## ??
        self.name = name

## Person-ს *აქვს* რაღაც ტიპის ცხოველი 
        self.pet = None

## ??
class Employee(Person):

    def __init__(self, name, salary):
        ## ?? hmm what is this strange magic?
        super(Employee, self).__init__(name)
        ## ??
        self.salary = salary

## ??
class Fish(object):
    pass

## ??
class Salmon(Fish):
    pass

## ??
class Halibut(Fish):
    pass


## Rover is-a Dog
rover = Dog("Rover")

## ??
satan = Cat("Satan")

## ??
mary = Person("Mary")

## ??
mary.pet = satan

## ??
frank = Employee("Frank", 120000)

## ??
frank.pet = rover

## ??
flipper = Fish()

## ??
crouse = Salmon()

## ??
harry = Halibut()
```

<h3 align="center">Name(object) კლასის შესახებ</h3>

- Python 3-ში უკვე აღარაა საჭირო `(object)`-ის დამატება კლასის დასახელების შემდგომ, მაგრამ Python-ის საზოგადოებას სჯერა, რომ "აშკარა ჩაწერა ნაგულისხმევს (იმპლიცისტურს) ჯობია".


- თქვენ შეგიძლიათ გაუშვათ კოდი და არ მიუთითოთ `(object)` კლასის განსაზღვრის შემდეგ, სკრიპტი მშვენივრად იმუშავებს.
