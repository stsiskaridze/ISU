- ქვემოთ მოცემულია ფაილზე სამუშაო **ბრძანებების სია**:

- `close` - ხურავს ფაილს;
- `read` - ფაილის კონტენტს კითხულობს. შედეგი შეიძლება მიაწერო ცვლადს;
- `readline` - ფაილის მხოლოდ ერთ სტრიქონს კითხულობს;
- `truncate` - აცარიელებს ფაილს. ფრთხილად იყავით, თუ ფაილში მნიშვნელოვანი მონაცემები გაქვთ;
- `write('stuff')` - წერს 'stuff'-ს ფაილში;
- `seek(0)` - წაკითხვის/წერის თავაკს ფაილის დასაწყისში აყენებს.


- თქვენ შეგიძლიათ ეს ფუნქციები აღიქვით როგორც ლენტური კასეტები გაქვთ, როდესაც საჭიროა ზუსტი ადგილის პოვნა, რომ შემდგომ იმ ადგილიდან დაიწყო მუშაობა. 


- ზოგი ფუნქცია იყენებს პარამეტრებს, მაგრამ ამჯერად ჩვენ არ გვესაჭიროება ისინი. 

- ჯერჯერობით დავიმახსოვროთ, რომ ფუნქცია `write()` საჭიროებს ჩვენგან სტრიქონის პარამეტრს, რომელსაც შემდგომ ჩაწერს ფაილში.


- მოდით გამოვიყენოთ ზოგი ჩვენს მომდევნო კოდში:

```python
from sys import argv

script, filename = argv

print(f"მე ვაპირებ შემდეგი ფაილის წაშლას: {filename}.")
print("თუ თქვენ არ გინდათ ფაილის წაშლა, მაშინ დააჭირეთ CTRL-C (^C).")
print("თუ არა და, დაელოდეთ ოპერატორს.")

input("?")

print("მიმდინარეობს ზარების მონი... უკაცრავად, ფაილის გახსნა...")
target = open(filename, 'w', encoding="utf8")

print("იყო და არა იყო და აღარაა!")
target.truncate()

print("ახლა გისმენ. მითხარი რა ჩავწერო და ჩავწერ. 3 ხაზზე მეტს ვერ ვიზამ ოღონდ.")

line1 = input("ხაზი 1: ")
line2 = input("ხაზი 2: ")
line3 = input("ხაზი 3: ")

print("მიმდინარეობს ზარების მონი... უკაცრავად, ფაილში ჩაწერა...")

target.write(line1)
target.write("\n")
target.write(line2)
target.write("\n")
target.write(line3)
target.write("\n")

print("ჩაწერა წარმატებით დასრულდა!")
print("უკვე გვიანაა, ვიკეტებით! ხვალ სცადე კიდევ თუ გინდა! აბა კაქაათ!")
target.close()
```

- **ყურადღება**: კოდის გაშვებისას **ტერმინალში** აკრიფეთ არგუმენტი, მაგალითად: `python ex_16.py test.txt`


- ეს სავარჯიშო არის, ალბათ, ყველაზე დიდი ფაილი, რაც აქამდე დაგვიწერია, ამიტომ არ იჩქაროთ. 

  შეამოწმეთ თითოეული ხაზი. 


- **სხვათაშორის**, ერთ-ერთი კარგი ვარიანტია, ნაწილობრივ გაუშვათ თქვენი კოდი, რათა შეამოწმოთ მისი სისწორე. 

  მაგალითად, ჯერ ხაზები 1-8, მერე კიდევ 5, კიდევ ცოტა და ა.შ.

- ჩვენ ხშირად შევცვლით კოდს, რომ დავადგინოთ მისი გაშვების საზღვრები.


- ქვევით მოყვანილია **კოდის მაგალითები** რომლებიც **არ კომპილირდება** და **იძლევა შეცდომას**. 


- **გაარჩიეთ** თითოეული მათგანი და **გაასწორეთ შეცდომები**!

<h2 align="center">ხშირად დასმული კითხვები</h2>

- ქვევით მოყვანილია კითხვები, რომლებიც გაუჩნდათ რეალურ სტუდენტებს ამ სავარჯიშოს შესრულებისას:


1. **გვჭირდება თუ არა ფუნქცია `truncate()`, თუ ჩვენ `w` პრამეტრს ვიყენებთ?**

   იხილეთ სამუშაო დავალებაის მე-5 პუნქტი.
   
   
2. **რას ნიშნავს `w`?**

   `w` არის სტრიქონის ტიპის ცვლადი, რომელიც აღნიშნავს სიტყვას `write`, კიდევ არის `r` - `read`-სთვის და `a` - `append`-სთვის. 
   
   ასევე არსებობს მათი მოდიფიკატორებიც.
   

3. **რა მოდიფიკატორების გამოყენება შემიძლია ფაილთან მუშაობისას?**

   ჯერჯერობით ყველაზე მნიშვნელოვანი მოდიფიკატორი, რაც დაგჭირდებათ, არის `+`, ანუ შეგიძლიათ დაწეროთ `w+`, `r+` და `a+`. 
   
   ამის შედეგად სკრიპტი გახსნის ფაილს ორივე მეთოდით: წაკითხვისა და ჩაწერის, - და დააყენებს ფაილს განსხვავებულ პოზიციაში იმისდა მიხედვით, თუ რომელი მოდიფიკატორი აირჩიეთ.
   
   
4. **როდესაც ვიყენებთ ფუნქციას `open(filename)`, მასში იგულისხმება `r` მოდიფიკატორი?**

   კი, ის მასში ნაგულისხმევია, ფუნქციის განმარტების თანახმად. 

<h2 align="center">სამუშაო დავალება</h2>


1. თუ არ გესმით ეს კოდი, დაუბრუნდით თქვენს სკრიპტს და გამოიყენეთ კომენტარების ხრიკი: დაუწერეთ თითოეულ ხაზს, თუ რას ასრულებს ის, რა შედეგი მოჰყვება მას. ასეთი მიდგომით თქვენ მიხვდებით, თუ რაში კოჭლობთ და რა საკითხი გაქვთ უკეთესად გამოსაკვლევი.


2. დაწერეთ ამ სავარჯიშოს მსგავსი კოდი, გამოიყენეთ `read` და `argv`, რომ წაიკითხოთ ფაილი, რომელიც კოდის შედეგად იქმნება.


3. სავარჯიშოში მოყვანილი კოდი ძალიან გრძელია. გამოიყენეთ სტრიქონები (strings), ფორმატი (format) და ესქეიპები (escape), რომ დაწეროთ მხოლოდ ერთი ფუნქცია `target.write()` სამიმე სტრიქონისთვის: `line1`, `line2`, `line3`.


4. გამოიკვლიეთ, თუ რატომ მივაწოდეთ დამატებითი პარამეტრი `w` ფუნქციას `open`. **მინიშნება**: ფუნქცია `open` ძალიან ფრთხილია და თქვენგან მოითხოვს ზუსტ მინიშნებას, რომ ფაილში რამის ჩაწერას აპირებთ.
