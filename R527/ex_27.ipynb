<h3 align="center">ბულის ალგებრა</h3>

- **ვინ იყო ბული**?

   **ჯორჯ ბული** (1815 წლის 2 ნოემბერი - 1864 წლის 8 დეკემბერი) იყო **ინგლისელი მათემატიკოსი**, **ფილოსოფოსი** და **ლოგისტიკოსი**. 
   
   ის მუშაობდა **ალგებრული ლოგიკის** და **დიფერენციალური განტოლებების** ველებში. 
   
   ის ყველაზე უკეთ ცნობილია როგორც წიგნის ავტორი **"აზროვნების კანონები"** გამოშებული 1854 წელს, რომელიც შეიცავს **ბულის ალგებრას**. 
   
   **ბულის ლოგიკას** მიაწერენ **ინფორმაციის ეპოქის** დაწყებას.

- **ბულის ალგებრას** ეწოდება არაცარიელ სიმრავლეს ($A$) რომელსაც გააჩნიეა:
  - ორი **ბინარული** (ორცვლადიანი) ოპერაცია $\wedge$ (**კონიუნქცია**) და $\vee$ (**დიზუინქცია**);
  - ერთი **უნარული** (ერთცვლადიანი) ოპერაცია $\lnot$ (**უარყოფა**);
  - ორი **გამოყოფილი ელემენტები** $0$ (**მცდარი** ანუ **False**) და $1$ (**მართალი** ანუ **True**);
  
  და რომელიც ნებისმიერი $a$, $b$ და $c$ ელემენტებისთვის $A$-დან აკმაყოფიულებს პირობებს:
  
  $\begin{matrix}
\bullet \textbf{ კომუტატიურობა: } & a \wedge b = b \wedge a \text{ და } a \vee b = b \vee a \\ 
\bullet \textbf{ ასოციურობა: } & a \wedge (b \wedge c) = (a \wedge b) \wedge c \text{ და } a \vee (b \vee c) = (a \vee b) \vee c \\
\bullet \textbf{ დისტრიბუციულობა: } & a \wedge (b \vee c) = (a \wedge b) \vee (a \wedge c) \text{ და } a \vee (b \wedge c) = (a \vee b) \wedge (a \vee c)  \\
\bullet \textbf{ შთანთქმის კანონი: } & a \wedge (a \vee b) = a \text{ და } a \vee (a \wedge b) = a \vee b \\ 
\bullet \textbf{ დამატების კანონი: } & a \vee \lnot a  = 1 \text{ და } a \wedge \lnot a  = 0  \\ 
\end{matrix}$
  

<h3 align="center">ბულის ალგებრა ინფორმატიკაში</h3>

- **Python**-ში ჩვენ ვიყენებთ შემდეგ **ლოგიკურ ოპერატორებს**, რათა ვიმუშაოთ გამოსახულებაზე და დავადგინოთ, **მართალია** ის თუ **მცდარი**:

  - `False` - რაც აღნიშნავს $0$-ს, ანუ **მცდარს**;
    
  - `True` - რაც აღნიშნავს $1$-ს, ანუ **მართალს**;

  - `and` - რაც აღნიშნავს $\wedge$-ს, ანუ **დიზუინქციას**;

  - `or` - რაც აღნიშნავს $\vee$-ს, ანუ **დიზუინქციას**;
  
  - `not` - რაც აღნიშნავს $\lnot$-ს, ანუ **უარყოფას**.


- ამის გარდა, **Python**-ში ჩვენ ვიყენებთ **შედარების ოპერატორებს**:

  - `==` - ტოლია;
  
  - `!=` - არ არის ტოლი;

  - `>` - მეტია;
 
  - `<` - ნაკლებია;

  - `>=` - მეტია ან ტოლი;
  
  - `<=` - ნაკლებია ან ტოლი.

<h2 align="center">ხშირად დასმული კითხვები</h2>

- ქვევით მოყვანილია კითხვები, რომლებიც გაუჩნდათ რეალურ სტუდენტებს ამ სავარჯიშოს შესრულებისას:


1. **არ შეიძლება, რომ უბრალოდ ალგებრულად დავიმახსოვრო, ვიდრე ზეპირად ვიცოდე?**

   რა თქმა უნდა, შეიძლება, მაგრამ სულ ყოველთვის მოგიწევს დახარჯო შენი დრო და რესურსი, რათა დაუბრუნდე ამ ცოდნას ლოგიკურად. თუ ახლა ამას დაიმახსოვრებთ, არა მხოლოდ თუ გაავარჯიშებთ თქვენს ტვინს, არამედ ამ ცხრილების გამოყენებას ბუნებრივ დონემდე აიყვანთ.
   
   პირველ რიგში, ბაზისურ ცოდნას უნდა დავეუფლოთ. მინდა, რომ ამ დავალებას საკმარისი დრო დაუთმოთ და **დაიმახსოვროთ** ყველა ცხრილიю

   მე არ გირჩევთ ზედიზედ 2 და მეტი საათი დაახარჯოთ დაზუთხვას. 
   
   რამდენიმე მიდგომაში უკეთესად დაიმახსოვრებთ და არ გადაიღლებით.
